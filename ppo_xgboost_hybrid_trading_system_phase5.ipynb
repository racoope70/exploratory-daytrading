```python
from __future__ import annotations
# ====== ======================================================

# CONFIG FIRST (Phase 0: reproducible + provenance)
# ============================================================
USE_SENTIMENT = False
USE_REGIME    = True
TEST_MODE     = True

CONTEXT_INTERVAL   = "1h"
EXEC_INTERVAL      = "15m"

PERIOD_DAYS        = 720
EXEC_CHUNK_DAYS    = 45   # chunk size for Alpaca 15m history pulls
FWD_HORIZON        = 10
UP_THR             = 0.02
DN_THR             = -0.02
VAL_FRACTION       = 0.20

DATA_VENDOR  = "alpaca"
TIMEZONE_STR = "America/New_York"

import os, gc, time, json, logging, hashlib, sys
from datetime import datetime, timezone
import numpy as np
import pandas as pd

# Alpaca
from alpaca.data.historical import StockHistoricalDataClient
from alpaca.data.requests import StockBarsRequest
from alpaca.data.timeframe import TimeFrame, TimeFrameUnit
from alpaca.data.enums import DataFeed

# Wavelet
import pywt

# Plot
import matplotlib.pyplot as plt

# ============================================================
# PATHS (Colab-safe + STRICT Google Drive handling)
# ============================================================
IN_COLAB = False

try:
    import google.colab  # noqa: F401
    from google.colab import drive  # type: ignore
    IN_COLAB = True

    print("[DRIVE] Mounting Google Drive...")
    drive.mount("/content/drive", force_remount=True)

    if not os.path.exists("/content/drive/MyDrive"):
        raise RuntimeError("Drive mounted but MyDrive not found")

    DRIVE_BASE = "/content/drive/MyDrive"

except Exception as e:
    DRIVE_BASE = os.getcwd()
    print(f"[PATH WARNING] Using local path instead of Drive: {e}")

print(f"[PATH CHECK] DRIVE_BASE = {DRIVE_BASE}")

DRIVE_DIR = os.path.join(DRIVE_BASE, "trading_data")
os.makedirs(DRIVE_DIR, exist_ok=True)

print(f"[PATH CHECK] DRIVE_DIR  = {DRIVE_DIR}")

# 🚨 HARD FAIL if in Colab but not actually mounted
if IN_COLAB and not DRIVE_BASE.startswith("/content/drive"):
    raise RuntimeError(
        f"Google Drive NOT mounted correctly. Current DRIVE_BASE={DRIVE_BASE}"
    )

LOCAL_OUT   = "multi_stock_feature_engineered_dataset_mt.csv"
LOCAL_TRAIN = "train_mt.csv"
LOCAL_VAL   = "val_mt.csv"

PARQ_FULL   = "features_full_mt.parquet"
PARQ_TRAIN  = "train_mt.parquet"
PARQ_VAL    = "val_mt.parquet"

def upload_data_env_in_colab():
    try:
        import google.colab  # noqa: F401
        from google.colab import files
    except Exception:
        return None

    print("\n[Colab setup] Upload your Alpaca data keys env file now.")
    print("[Colab setup] Expected filename example: Alpaca_data_keys.env")

    uploaded = files.upload()
    if not uploaded:
        raise ValueError("No env file uploaded.")

    uploaded_name = next(iter(uploaded.keys()))
    print(f"[Colab setup] Current DRIVE_BASE -> {DRIVE_BASE}")
    save_path = os.path.join(DRIVE_BASE, "Alpaca_data_keys.env")

    with open(save_path, "wb") as f:
        f.write(uploaded[uploaded_name])

    print(f"[Colab setup] Saved data env file -> {save_path}")
    return save_path


def load_env_file(env_path: str):
    if not env_path or not os.path.exists(env_path):
        raise FileNotFoundError(f"Env file not found: {env_path}")

    with open(env_path, "r", encoding="utf-8") as f:
        for raw_line in f:
            line = raw_line.strip()

            if not line:
                continue
            if line.startswith("#"):
                continue
            if "=" not in line:
                continue

            key, value = line.split("=", 1)
            key = key.strip()
            value = value.strip().strip('"').strip("'")

            os.environ[key] = value


DATA_ENV_FILE_PATH = None

try:
    import google.colab  # noqa: F401
    DATA_ENV_FILE_PATH = upload_data_env_in_colab()
    if DATA_ENV_FILE_PATH:
        load_env_file(DATA_ENV_FILE_PATH)
except Exception as e:
    print(f"[Colab setup] Data env upload skipped or failed: {e}")

# ============================================================
# ALPACA CONFIG
# ============================================================
ALPACA_API_KEY = os.getenv("APCA_API_KEY_ID", os.getenv("ALPACA_API_KEY"))
ALPACA_SECRET_KEY = os.getenv("APCA_API_SECRET_KEY", os.getenv("ALPACA_SECRET_KEY"))
ALPACA_FEED = "iex" if TEST_MODE else "sip"

if not ALPACA_API_KEY or not ALPACA_SECRET_KEY:
    raise ValueError(
        "Missing Alpaca credentials in Alpaca_data_keys.env. "
        "Expected APCA_API_KEY_ID and APCA_API_SECRET_KEY."
    )

alpaca_data_client = StockHistoricalDataClient(
    api_key=ALPACA_API_KEY,
    secret_key=ALPACA_SECRET_KEY,
)

# ============================================================
# LOGGING
# ============================================================
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s", force=True)
logging.info("Data vendor: Alpaca")
logging.info(f"pandas version: {pd.__version__}")
logging.info(f"Alpaca feed: {ALPACA_FEED}")

# ============================================================
# MANIFEST (Phase 0: config.json-ish + ordered features + provenance)
#    IMPROVEMENT: define MANIFEST ONCE (no overwrites)
# ============================================================
MANIFEST_PATH = os.path.join(DRIVE_DIR, "artifact_manifest.json")

def _sha256_file(path: str, chunk_size: int = 1024 * 1024) -> str:
    h = hashlib.sha256()
    with open(path, "rb") as f:
        while True:
            b = f.read(chunk_size)
            if not b:
                break
            h.update(b)
    return h.hexdigest()

def _safe_json_dump(obj: dict, path: str):
    os.makedirs(os.path.dirname(path), exist_ok=True)
    tmp = path + ".tmp"
    with open(tmp, "w") as f:
        json.dump(obj, f, indent=2, sort_keys=True, default=str)
    os.replace(tmp, path)

MANIFEST = {
    "created_utc": datetime.now(timezone.utc).isoformat(),
    "config": {
        "CONTEXT_INTERVAL": CONTEXT_INTERVAL,
        "EXEC_INTERVAL": EXEC_INTERVAL,
        "PERIOD_DAYS": PERIOD_DAYS,
        "EXEC_CHUNK_DAYS": EXEC_CHUNK_DAYS,
        "FWD_HORIZON": FWD_HORIZON,
        "UP_THR": UP_THR,
        "DN_THR": DN_THR,
        "VAL_FRACTION": VAL_FRACTION,
        "USE_REGIME": USE_REGIME,
        "USE_SENTIMENT": USE_SENTIMENT,
        "TEST_MODE": TEST_MODE,
        "DATA_VENDOR": DATA_VENDOR,
        "TIMEZONE": TIMEZONE_STR,
    },
    "versions": {
    "python": sys.version,
    "pandas": pd.__version__,
    "numpy": np.__version__,
    "data_vendor": "alpaca",
    },
    "data_provenance": {},
    "feature_manifest": [],
    "artifacts": {},
}

def manifest_record(key: str, path: str):
    MANIFEST.setdefault("artifacts", {})
    MANIFEST["artifacts"][key] = {
        "path": os.path.abspath(path),
        "sha256": _sha256_file(path) if os.path.exists(path) else None,
        "bytes": int(os.path.getsize(path)) if os.path.exists(path) else None,
        "written_utc": datetime.now(timezone.utc).isoformat(),
    }
    _safe_json_dump(MANIFEST, MANIFEST_PATH)

# write initial manifest once
_safe_json_dump(MANIFEST, MANIFEST_PATH)

# ============================================================
# UNIVERSE
# ============================================================
ticker_list = [
    "AAPL","TSLA","MSFT","GOOGL","AMZN","NVDA","META","BRK-B","JPM","JNJ",
    "XOM","V","PG","UNH","MA","HD","LLY","MRK","PEP","KO",
    "BAC","ABBV","AVGO","PFE","COST","CSCO","TMO","ABT","ACN","WMT",
    "MCD","ADBE","DHR","CRM","NKE","INTC","QCOM","NEE","AMD","TXN",
    "AMGN","UPS","LIN","PM","UNP","BMY","LOW","RTX","CVX","IBM",
    "GE","SBUX","ORCL",
]
SYMBOLS = ["AAPL", "NVDA", "MSFT"] if TEST_MODE else ticker_list

# ============================================================
# OPTIONAL SENTIMENT MODEL
# ============================================================
if USE_SENTIMENT:
    try:
        import torch
        from transformers import pipeline

        device_id = 0 if torch.cuda.is_available() else -1
        sentiment_pipeline = pipeline("sentiment-analysis", model="ProsusAI/finbert", device=device_id)
        logging.info("FinBERT sentiment enabled.")
    except Exception as e:
        logging.warning(f"Could not init FinBERT; disabling sentiment. Err: {e}")
        USE_SENTIMENT = False
        sentiment_pipeline = None
else:
    sentiment_pipeline = None

# ============================================================
# HELPERS
# ============================================================

def _alpaca_timeframe_from_str(interval: str):
    interval = str(interval).lower().strip()

    if interval == "1h":
        return TimeFrame(1, TimeFrameUnit.Hour)
    if interval == "15m":
        return TimeFrame(15, TimeFrameUnit.Minute)
    if interval == "5m":
        return TimeFrame(5, TimeFrameUnit.Minute)
    if interval == "1d":
        return TimeFrame(1, TimeFrameUnit.Day)

    raise ValueError(f"Unsupported Alpaca interval: {interval}")


def _normalize_alpaca_bars_df(df_in: pd.DataFrame, ticker: str) -> pd.DataFrame:
    """
    Normalize Alpaca bars to:
    Datetime, Open, High, Low, Close, Adj Close, Volume, Symbol
    """
    if df_in is None or df_in.empty:
        return pd.DataFrame()

    df = df_in.copy()

    # Alpaca bars usually come with timestamp index or 'timestamp'
    if isinstance(df.index, pd.DatetimeIndex):
        df = df.reset_index()

    if "timestamp" in df.columns:
        df = df.rename(columns={"timestamp": "Datetime"})
    elif "Datetime" not in df.columns:
        first = df.columns[0]
        df = df.rename(columns={first: "Datetime"})

    rename_map = {
        "open": "Open",
        "high": "High",
        "low": "Low",
        "close": "Close",
        "volume": "Volume",
        "trade_count": "Trade_Count",
        "vwap": "VWAP",
    }
    df = df.rename(columns=rename_map)

    df["Datetime"] = pd.to_datetime(df["Datetime"], errors="coerce", utc=True)
    df = df.dropna(subset=["Datetime"]).sort_values("Datetime").drop_duplicates(subset=["Datetime"]).reset_index(drop=True)

    if "Adj Close" not in df.columns:
        df["Adj Close"] = df["Close"]

    df["Symbol"] = ticker

    needed = {"Datetime", "Open", "High", "Low", "Close", "Volume", "Symbol"}
    missing = needed - set(df.columns)
    if missing:
        raise ValueError(f"[{ticker}] Missing normalized Alpaca columns: {missing}")

    return df

def download_context_data(
    ticker: str,
    interval: str = "1h",
    period_days: int = 720,
    max_retries: int = 5,
    sleep_base: int = 3,
):
    start_ts = pd.Timestamp.now(tz="UTC") - pd.Timedelta(days=period_days)
    end_ts = pd.Timestamp.now(tz="UTC")
    timeframe = _alpaca_timeframe_from_str(interval)

    for attempt in range(1, max_retries + 1):
        try:
            logging.info(
                f"[{ticker}] Alpaca context attempt {attempt}: "
                f"{start_ts} -> {end_ts} | interval={interval}"
            )

            req = StockBarsRequest(
                symbol_or_symbols=ticker,
                timeframe=timeframe,
                start=start_ts.to_pydatetime(),
                end=end_ts.to_pydatetime(),
                feed=DataFeed.IEX if ALPACA_FEED.lower() == "iex" else DataFeed.SIP,
                adjustment="raw",
            )

            bars = alpaca_data_client.get_stock_bars(req)
            df = bars.df

            if df is None or df.empty:
                raise ValueError("Empty context bars from Alpaca")

            # If multi-index, isolate symbol
            if isinstance(df.index, pd.MultiIndex):
                try:
                    df = df.xs(ticker, level=0).copy()
                except Exception:
                    df = df.reset_index()
                    df = df[df["symbol"] == ticker].copy()

            df = _normalize_alpaca_bars_df(df, ticker)

            logging.info(
                f"[{ticker}] context rows: {len(df)} "
                f"from {df['Datetime'].min()} to {df['Datetime'].max()}"
            )
            return df

        except Exception as e:
            wait = sleep_base * attempt
            logging.warning(f"[{ticker}] context download failed: {e} | retrying in {wait}s")
            time.sleep(wait)

    logging.error(f"[{ticker}] Failed context download after {max_retries} attempts.")
    return None

def download_execution_data_15m(
    ticker: str,
    total_days: int = 720,
    chunk_days: int = 45,
    max_retries: int = 5,
    sleep_base: int = 3,
) -> pd.DataFrame | None:
    """
    Download 15m execution bars from Alpaca in chunks and stitch together.
    Uses the same normalization approach as the 1h context downloader.
    """
    end_ts = pd.Timestamp.now(tz="UTC").floor("min")
    start_ts = end_ts - pd.Timedelta(days=total_days)
    timeframe = _alpaca_timeframe_from_str("15m")

    chunk_frames = []
    chunk_start = start_ts

    while chunk_start < end_ts:
        chunk_end = min(chunk_start + pd.Timedelta(days=chunk_days), end_ts)
        got_chunk = False

        for attempt in range(1, max_retries + 1):
            try:
                logging.info(
                    f"[{ticker}] Alpaca 15m chunk attempt {attempt}: "
                    f"{chunk_start} -> {chunk_end}"
                )

                req = StockBarsRequest(
                    symbol_or_symbols=ticker,
                    timeframe=timeframe,
                    start=chunk_start.to_pydatetime(),
                    end=chunk_end.to_pydatetime(),
                    feed=DataFeed.IEX if ALPACA_FEED.lower() == "iex" else DataFeed.SIP,
                    adjustment="raw",
                )

                bars = alpaca_data_client.get_stock_bars(req)
                df = bars.df

                if df is None or df.empty:
                    raise ValueError("Empty 15m chunk from Alpaca")

                if isinstance(df.index, pd.MultiIndex):
                    try:
                        df = df.xs(ticker, level=0).copy()
                    except Exception:
                        df = df.reset_index()
                        if "symbol" in df.columns:
                            df = df[df["symbol"] == ticker].copy()

                df = _normalize_alpaca_bars_df(df, ticker)

                df = df[
                    (df["Datetime"] >= chunk_start) &
                    (df["Datetime"] < chunk_end)
                ].copy()

                if df.empty:
                    raise ValueError("15m chunk became empty after datetime filtering")

                chunk_frames.append(df)
                got_chunk = True
                break

            except Exception as e:
                wait = sleep_base * attempt
                logging.warning(
                    f"[{ticker}] Alpaca 15m chunk failed: {e} | retrying in {wait}s"
                )
                time.sleep(wait)

        if not got_chunk:
            logging.warning(
                f"[{ticker}] failed Alpaca 15m chunk "
                f"{chunk_start} -> {chunk_end}; continuing"
            )

        chunk_start = chunk_end

    if not chunk_frames:
        logging.error(f"[{ticker}] No 15m chunks downloaded from Alpaca.")
        return None

    out = pd.concat(chunk_frames, ignore_index=True)
    out["Datetime"] = pd.to_datetime(out["Datetime"], errors="coerce", utc=True)
    out = (
        out.dropna(subset=["Datetime"])
           .drop_duplicates(subset=["Datetime"])
           .sort_values("Datetime")
           .reset_index(drop=True)
    )

    logging.info(
        f"[{ticker}] final Alpaca 15m rows: {len(out)} "
        f"from {out['Datetime'].min()} to {out['Datetime'].max()}"
    )
    return out

def denoise_wavelet(series: pd.Series, wavelet: str = "db1", level: int = 2) -> pd.Series:
    s = pd.Series(series).astype(float).ffill().bfill().to_numpy()
    try:
        coeffs = pywt.wavedec(s, wavelet, mode="symmetric", level=level)
        for i in range(1, len(coeffs)):
            coeffs[i] = np.zeros_like(coeffs[i])
        rec = pywt.waverec(coeffs, wavelet, mode="symmetric")
        return pd.Series(rec[:len(s)], index=series.index)
    except Exception as e:
        logging.warning(f"Wavelet denoising failed ({e}); using raw Close.")
        return pd.Series(s, index=series.index)

def score_sentiment(texts: list[str]) -> list[float]:
    if not USE_SENTIMENT or sentiment_pipeline is None:
        return [0.0] * len(texts)
    try:
        outputs = sentiment_pipeline(texts, truncation=True, max_length=256, batch_size=32)
        scores = []
        for r in outputs:
            label = str(r["label"]).lower()
            if label == "positive":
                scores.append(+float(r["score"]))
            elif label == "negative":
                scores.append(-float(r["score"]))
            else:
                scores.append(0.0)
        return scores
    except Exception as e:
        logging.error(f"Sentiment scoring error: {e}")
        return [0.0] * len(texts)

def add_regime(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df["Vol20"] = df["Close"].pct_change().rolling(20).std()
    df["Ret20"] = df["Close"].pct_change(20)
    vol_hi   = (df["Vol20"] > df["Vol20"].median()).astype(int)
    trend_hi = (df["Ret20"].abs() > df["Ret20"].abs().median()).astype(int)
    df["Regime4"] = vol_hi * 2 + trend_hi
    return df

def compute_execution_features_15m(df: pd.DataFrame) -> pd.DataFrame:
    """
    Keep execution rows lighter for now.
    We can expand later if needed.
    """
    df = df.copy()

    if "Symbol" not in df.columns:
        df["Symbol"] = "UNKNOWN"

    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)
    df = df.loc[:, ~df.columns.duplicated()]

    req = {"Open", "High", "Low", "Close", "Volume", "Datetime", "Symbol"}
    missing = req - set(df.columns)
    if missing:
        raise ValueError(f"Missing required columns in 15m execution df: {missing}")

    df["Exec_Ret_1"] = df["Close"].pct_change(1)
    df["Exec_Ret_4"] = df["Close"].pct_change(4)
    df["Exec_Vol_20"] = df["Close"].pct_change().rolling(20).std()
    df["Exec_ATR_14"] = pd.concat(
        [
            (df["High"] - df["Low"]),
            (df["High"] - df["Close"].shift()).abs(),
            (df["Low"] - df["Close"].shift()).abs(),
        ],
        axis=1,
    ).max(axis=1).rolling(14).mean()

    df = df.reset_index(drop=True)
    cols = [c for c in df.columns if c != "Symbol"] + ["Symbol"]
    return df[cols]

def compute_context_features_1h(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()

    if "Symbol" not in df.columns:
        df["Symbol"] = "UNKNOWN"

    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)
    df = df.loc[:, ~df.columns.duplicated()]

    req = {"Open", "High", "Low", "Close", "Volume", "Datetime", "Symbol"}
    missing = req - set(df.columns)
    if missing:
        raise ValueError(f"Missing required columns: {missing}")

    # ---- Indicators ----
    df["SMA_20"] = df["Close"].rolling(20).mean()
    df["STD_20"] = df["Close"].rolling(20).std()
    df["Upper_Band"] = df["SMA_20"] + 2 * df["STD_20"]
    df["Lower_Band"] = df["SMA_20"] - 2 * df["STD_20"]

    df["Lowest_Low"]   = df["Low"].rolling(14).min()
    df["Highest_High"] = df["High"].rolling(14).max()
    denom = (df["Highest_High"] - df["Lowest_Low"]).replace(0, np.nan)
    df["Stoch"] = ((df["Close"] - df["Lowest_Low"]) / denom) * 100

    df["ROC"] = df["Close"].pct_change(10)
    df["OBV"] = (np.sign(df["Close"].diff()).fillna(0) * df["Volume"].fillna(0)).cumsum()

    tp = (df["High"] + df["Low"] + df["Close"]) / 3
    sma_tp = tp.rolling(20).mean()
    md = (tp - sma_tp).abs().rolling(20).mean()
    df["CCI"] = (tp - sma_tp) / (0.015 * md)

    df["EMA_10"] = df["Close"].ewm(span=10, adjust=False).mean()
    df["EMA_50"] = df["Close"].ewm(span=50, adjust=False).mean()
    ema12 = df["Close"].ewm(span=12, adjust=False).mean()
    ema26 = df["Close"].ewm(span=26, adjust=False).mean()
    df["MACD_Line"]   = ema12 - ema26
    df["MACD_Signal"] = df["MACD_Line"].ewm(span=9, adjust=False).mean()

    delta = df["Close"].diff()
    gain  = delta.clip(lower=0).rolling(14).mean()
    loss  = (-delta.clip(upper=0)).rolling(14).mean()
    rs = gain / (loss.replace(0, np.nan))
    df["RSI"] = 100 - (100 / (1 + rs))

    tr = pd.concat(
        [
            (df["High"] - df["Low"]),
            (df["High"] - df["Close"].shift()).abs(),
            (df["Low"] - df["Close"].shift()).abs(),
        ],
        axis=1,
    ).max(axis=1)
    df["ATR"] = tr.rolling(14).mean()

    df["Volatility"] = df["Close"].pct_change().rolling(20).std()
    df["Denoised_Close"] = denoise_wavelet(df["Close"].ffill())

    if USE_REGIME:
        df = add_regime(df)

    if USE_SENTIMENT and len(df):
        headline = f"{df['Symbol'].iloc[0]} is expected to perform well in the market."
        try:
            score = score_sentiment([headline])[0]
        except Exception as e:
            logging.warning(f"Sentiment scoring failed for {df['Symbol'].iloc[0]}: {e}")
            score = 0.0
        df["SentimentScore"] = float(score)
    else:
        df["SentimentScore"] = 0.0

    df["Delta"] = df["Close"].pct_change(1).fillna(0)
    df["Gamma"] = df["Delta"].diff().fillna(0)

    df = df.reset_index(drop=True)
    cols = [c for c in df.columns if c != "Symbol"] + ["Symbol"]
    return df[cols]


def merge_context_into_execution(
    context_features: pd.DataFrame,
    exec_features: pd.DataFrame,
) -> pd.DataFrame:
    """
    Merge 1h context features into 15m execution rows using backward merge_asof.
    Each 15m row only sees the most recent completed 1h context row.
    """
    if context_features is None or context_features.empty:
        raise ValueError("context_features is empty")
    if exec_features is None or exec_features.empty:
        raise ValueError("exec_features is empty")

    ctx = context_features.copy()
    exe = exec_features.copy()

    ctx["Datetime"] = pd.to_datetime(ctx["Datetime"], errors="coerce")
    exe["Datetime"] = pd.to_datetime(exe["Datetime"], errors="coerce")

    ctx = ctx.dropna(subset=["Datetime"]).sort_values("Datetime").reset_index(drop=True)
    exe = exe.dropna(subset=["Datetime"]).sort_values("Datetime").reset_index(drop=True)

    if "Symbol" in ctx.columns:
        ctx = ctx.drop(columns=["Symbol"])

    ctx = ctx.rename(columns={"Datetime": "Context_Datetime"})

    rename_map = {}
    for col in ctx.columns:
        if col != "Context_Datetime":
            rename_map[col] = f"CTX_{col}"
    ctx = ctx.rename(columns=rename_map)

    merged = pd.merge_asof(
        exe,
        ctx,
        left_on="Datetime",
        right_on="Context_Datetime",
        direction="backward",
        allow_exact_matches=False,
    )

    if "Symbol" not in merged.columns and "Symbol" in exec_features.columns:
        merged["Symbol"] = exec_features["Symbol"].values[:len(merged)]

    merged = merged.sort_values("Datetime").reset_index(drop=True)

    logging.info(
        f"[MERGE] merged rows={len(merged)} | "
        f"exec_rows={len(exec_features)} | context_rows={len(context_features)}"
    )

    return merged


def validate_no_lookahead_after_merge(
    merged_df: pd.DataFrame,
    context_df: pd.DataFrame,
    ticker: str,
    sample_rows: int = 12,
) -> None:
    if merged_df is None or merged_df.empty:
        raise ValueError(f"[{ticker}] merged_df is empty in leakage check.")

    if context_df is None or context_df.empty:
        raise ValueError(f"[{ticker}] context_df is empty in leakage check.")

    work_m = merged_df.copy()
    work_c = context_df.copy()

    work_m["Datetime"] = pd.to_datetime(work_m["Datetime"], errors="coerce")
    work_m["Context_Datetime"] = pd.to_datetime(work_m["Context_Datetime"], errors="coerce")
    work_c["Datetime"] = pd.to_datetime(work_c["Datetime"], errors="coerce")

    work_m = work_m.dropna(subset=["Datetime"]).sort_values("Datetime").reset_index(drop=True)
    work_c = work_c.dropna(subset=["Datetime"]).sort_values("Datetime").reset_index(drop=True)

    bad_future = work_m[
        work_m["Context_Datetime"].notna() &
        (work_m["Context_Datetime"] > work_m["Datetime"])
    ].copy()

    if not bad_future.empty:
        raise AssertionError(
            f"[{ticker}] LOOKAHEAD DETECTED: {len(bad_future)} merged rows use future context."
        )

    work_m["Context_Age_Min"] = (
        (work_m["Datetime"] - work_m["Context_Datetime"]).dt.total_seconds() / 60.0
    )

    negative_age = work_m[work_m["Context_Age_Min"] < 0].copy()
    if not negative_age.empty:
        raise AssertionError(f"[{ticker}] LOOKAHEAD DETECTED: negative context age found.")

    same_ts = work_m[
        work_m["Context_Datetime"].notna() &
        (work_m["Context_Datetime"] == work_m["Datetime"])
    ].copy()

    if not same_ts.empty:
        raise AssertionError(
            f"[{ticker}] exact timestamp context matches still found: {len(same_ts)} rows."
        )

    if "CTX_Close" in work_m.columns and "Close" in work_c.columns:
        ctx_compare = pd.merge(
            work_m[["Datetime", "Context_Datetime", "CTX_Close"]].copy(),
            work_c[["Datetime", "Close"]].rename(
                columns={"Datetime": "Context_Datetime", "Close": "Expected_CTX_Close"}
            ),
            on="Context_Datetime",
            how="left",
        )

        mismatch = ctx_compare[
            ~np.isclose(
                pd.to_numeric(ctx_compare["CTX_Close"], errors="coerce"),
                pd.to_numeric(ctx_compare["Expected_CTX_Close"], errors="coerce"),
                equal_nan=True,
            )
        ].copy()

        if not mismatch.empty:
            raise AssertionError(
                f"[{ticker}] CTX_Close mismatch detected on {len(mismatch)} rows."
            )

    diag_cols = ["Datetime", "Context_Datetime", "Context_Age_Min"]
    if "Close" in work_m.columns:
        diag_cols.append("Close")
    if "CTX_Close" in work_m.columns:
        diag_cols.append("CTX_Close")

    logging.info(f"[{ticker}] leakage check passed.")
    logging.info(
        f"[{ticker}] context age stats (minutes) | "
        f"min={work_m['Context_Age_Min'].min():.2f}, "
        f"median={work_m['Context_Age_Min'].median():.2f}, "
        f"max={work_m['Context_Age_Min'].max():.2f}"
    )

    logging.info(f"[{ticker}] leakage check sample:")
    logging.info("\n" + work_m[diag_cols].head(sample_rows).to_string(index=False))


def relabel(df_in: pd.DataFrame) -> pd.DataFrame:
    df = df_in.copy()
    df["Return"] = (
        df.groupby("Symbol")["Close"]
          .shift(-FWD_HORIZON)
          .sub(df["Close"])
          .div(df["Close"])
    )
    df["Target"] = np.select(
        [df["Return"] > UP_THR, df["Return"] < DN_THR],
        [1, -1],
        default=0,
    )
    return df

def summarize(df: pd.DataFrame):
    print("Combined dataset shape:", df.shape)
    print("Range:", df["Datetime"].min(), "-", df["Datetime"].max())
    print("Per-ticker counts:")
    print(df["Symbol"].value_counts().to_string())

    na_cols = df.columns[df.isna().any()]
    if len(na_cols):
        print("\nColumns with NaNs:")
        print(df[na_cols].isna().sum().sort_values(ascending=False).to_string())
    else:
        print("\nNo NaNs detected.")

    print("\nLabel counts:")
    print(df["Target"].value_counts().sort_index().to_string())
    print("\nLabel ratios (%):")
    print((df["Target"].value_counts(normalize=True) * 100).round(2).to_string())

# ============================================================
# BUILD DATASET
# ============================================================
all_dfs: list[pd.DataFrame] = []

for i, ticker in enumerate(SYMBOLS, 1):
    logging.info(f"[{i}/{len(SYMBOLS)}] Processing {ticker}")

    context_raw = download_context_data(
        ticker,
        interval=CONTEXT_INTERVAL,
        period_days=PERIOD_DAYS,
        max_retries=5,
        sleep_base=3,
    )

    exec_raw = download_execution_data_15m(
        ticker=ticker,
        total_days=PERIOD_DAYS,
        chunk_days=EXEC_CHUNK_DAYS,
        max_retries=5,
        sleep_base=3,
    )

    if context_raw is None or context_raw.empty:
        logging.warning(f"[{ticker}] No 1h context data")
        continue

    if exec_raw is None or exec_raw.empty:
        logging.warning(f"[{ticker}] No 15m execution data")
        continue

    MANIFEST.setdefault("data_provenance", {})
    MANIFEST["data_provenance"][ticker] = {
        "vendor": DATA_VENDOR,
        "feed": ALPACA_FEED,
        "context_interval": CONTEXT_INTERVAL,
        "execution_interval": EXEC_INTERVAL,
        "period_days": PERIOD_DAYS,
        "exec_chunk_days": EXEC_CHUNK_DAYS,
        "context_rows": int(len(context_raw)),
        "execution_rows": int(len(exec_raw)),
        "context_start": str(context_raw["Datetime"].min()),
        "context_end": str(context_raw["Datetime"].max()),
        "execution_start": str(exec_raw["Datetime"].min()),
        "execution_end": str(exec_raw["Datetime"].max()),
    }
    _safe_json_dump(MANIFEST, MANIFEST_PATH)

    try:
        logging.info(f"[{ticker}] raw 1h rows before features: {len(context_raw)}")
        logging.info(f"[{ticker}] raw 15m rows before features: {len(exec_raw)}")

        context_features = compute_context_features_1h(context_raw)
        exec_features = compute_execution_features_15m(exec_raw)

        logging.info(f"[{ticker}] 1h context rows after features: {len(context_features)}")
        logging.info(f"[{ticker}] 15m execution rows after features: {len(exec_features)}")

        merged = merge_context_into_execution(context_features, exec_features)

        validate_no_lookahead_after_merge(
            merged_df=merged,
            context_df=context_features,
            ticker=ticker,
            sample_rows=12,
        )

        if merged is not None and not merged.empty:
            logging.info(f"[{ticker}] merged rows before final relabel: {len(merged)}")
            all_dfs.append(merged)
        else:
            logging.warning(f"[{ticker}] merged dataset empty")

    except Exception as e:
        logging.error(f"Multi-timeframe feature build failed for {ticker}: {e}")

    finally:
        del context_raw, exec_raw
        try:
            del context_features
        except NameError:
            pass
        try:
            del exec_features
        except NameError:
            pass
        try:
            del merged
        except NameError:
            pass
        gc.collect()
        time.sleep(0.5)

if not all_dfs:
    logging.warning("No usable multi-timeframe data found for any ticker.")
    raise SystemExit

final_df = pd.concat(all_dfs, ignore_index=True)
logging.info(f"Combined multi-timeframe dataset shape: {final_df.shape}")

# ============================================================
# TIMEZONE HANDLING + RTH FILTER + TZ-NAIVE OUTPUT
# applies to merged 15m execution rows
# ============================================================

dt = pd.to_datetime(final_df["Datetime"], errors="coerce", utc=True)
ctx_dt = pd.to_datetime(final_df["Context_Datetime"], errors="coerce", utc=True)

logging.info(f"[POST-CONCAT] rows before timezone/RTH filter: {len(final_df)}")

final_df["Datetime"] = dt.dt.tz_convert(TIMEZONE_STR)
final_df["Context_Datetime"] = ctx_dt.dt.tz_convert(TIMEZONE_STR)

dt2 = final_df["Datetime"]
rth_mask = (
    (dt2.dt.weekday < 5) &
    (dt2.dt.time >= pd.to_datetime("09:30").time()) &
    (dt2.dt.time < pd.to_datetime("16:00").time())
)
final_df = final_df[rth_mask].reset_index(drop=True)

logging.info(f"[POST-CONCAT] rows after RTH filter: {len(final_df)}")

final_df["Datetime"] = final_df["Datetime"].dt.tz_localize(None)
final_df["Context_Datetime"] = final_df["Context_Datetime"].dt.tz_localize(None)

# ============================================================
# LABELING + HORIZON TRIM
# ============================================================
final_df = relabel(final_df)
final_df = final_df.sort_values(["Symbol", "Datetime"]).reset_index(drop=True)

logging.info(f"[POST-LABEL] rows before horizon trim: {len(final_df)}")

# Create __i/__n then drop last horizon rows per symbol
final_df["__i"] = final_df.groupby("Symbol").cumcount()
final_df["__n"] = final_df.groupby("Symbol")["__i"].transform("max") + 1
final_df = final_df[final_df["__i"] < (final_df["__n"] - FWD_HORIZON)].copy()
final_df = final_df.drop(columns=["__i", "__n"]).reset_index(drop=True)

logging.info(f"[POST-LABEL] rows after horizon trim: {len(final_df)}")

# ============================================================
# FEATURE MANIFEST (ordered) + NA CLEAN
# ============================================================
exclude = {"Target", "Return", "Symbol", "Datetime", "Context_Datetime"}
feature_cols = [c for c in final_df.columns if c not in exclude]

# stable/ordered feature manifest
feature_cols = sorted([str(c) for c in feature_cols])
MANIFEST["feature_manifest"] = feature_cols
_safe_json_dump(MANIFEST, MANIFEST_PATH)

logging.info(f"[FINAL CLEAN] rows before final feature dropna: {len(final_df)}")

na_counts = final_df[feature_cols].isna().sum()
na_counts = na_counts[na_counts > 0].sort_values(ascending=False)

if len(na_counts):
    logging.info("[FINAL CLEAN] feature NaN counts before dropna:")
    logging.info("\n" + na_counts.to_string())
else:
    logging.info("[FINAL CLEAN] no feature NaNs before dropna")

final_df = final_df.dropna(subset=feature_cols).reset_index(drop=True)

logging.info(f"[FINAL CLEAN] rows after final feature dropna: {len(final_df)}")

# Put Target/Return/Symbol at end (matches your training expectations)
ORDER_LAST = ["Target", "Return", "Symbol"]
cols = [c for c in final_df.columns if c not in ORDER_LAST] + ORDER_LAST
final_df = final_df[cols]

# ============================================================
# SUMMARY PRINT
# ============================================================
summarize(final_df)

logging.info("[FINAL DATASET] per-symbol row counts after all cleanup:")
logging.info("\n" + final_df["Symbol"].value_counts().to_string())

# ============================================================
# PER-SYMBOL TIME SPLIT (train/val)
# ============================================================
train_parts, val_parts = [], []

for sym, g in final_df.groupby("Symbol", sort=False):
    g = g.sort_values("Datetime").reset_index(drop=True)
    cut = int((1.0 - VAL_FRACTION) * len(g))

    if cut <= 0 or cut >= len(g):
        logging.warning(f"Skipping {sym} due to insufficient rows for split.")
        continue

    train_parts.append(g.iloc[:cut])
    val_parts.append(g.iloc[cut:])

if not train_parts or not val_parts:
    raise ValueError("No train/validation splits were created. Check final_df and split logic.")

train_df = pd.concat(train_parts, ignore_index=True)
val_df   = pd.concat(val_parts, ignore_index=True)

logging.info("[TRAIN SPLIT] per-symbol train counts:")
logging.info("\n" + train_df["Symbol"].value_counts().to_string())

logging.info("[VAL SPLIT] per-symbol val counts:")
logging.info("\n" + val_df["Symbol"].value_counts().to_string())

train_df = train_df.sort_values(["Symbol", "Datetime"]).reset_index(drop=True)
val_df   = val_df.sort_values(["Symbol", "Datetime"]).reset_index(drop=True)

print(f"\nTrain: {train_df.shape},  Val: {val_df.shape}")

# ============================================================
# GROUP B — GATE MODEL (trade / no-trade)
# ============================================================
import joblib
from sklearn.metrics import (
    precision_score,
    recall_score,
    f1_score,
    accuracy_score,
    confusion_matrix,
)
from sklearn.ensemble import RandomForestClassifier

GATE_DIR = os.path.join(DRIVE_DIR, "gate_outputs")
os.makedirs(GATE_DIR, exist_ok=True)

# 1) Binary gate target: trade (1) if Target != 0 else no-trade (0)
train_df = train_df.copy()
val_df = val_df.copy()

train_df["Gate_Target"] = (train_df["Target"] != 0).astype(int)
val_df["Gate_Target"] = (val_df["Target"] != 0).astype(int)

# 2) Gate feature set
gate_exclude = {"Target", "Return", "Symbol", "Datetime", "Context_Datetime", "Gate_Target"}

leakage_cols = {"Exec_Ret_1", "Exec_Ret_4"}

gate_features = [
    c for c in train_df.columns
    if c not in gate_exclude and c not in leakage_cols
]

gate_features = sorted(gate_features)

X_train = train_df[gate_features].copy()
y_train = train_df["Gate_Target"].copy()

X_val = val_df[gate_features].copy()
y_val = val_df["Gate_Target"].copy()

assert list(X_train.columns) == gate_features, "Train feature mismatch"
assert list(X_val.columns) == gate_features, "Validation feature mismatch"

logging.info(f"[GATE] X_train shape: {X_train.shape}, positive rate: {y_train.mean():.4f}")
logging.info(f"[GATE] X_val shape: {X_val.shape}, positive rate: {y_val.mean():.4f}")

baseline_trade_rate = y_val.mean()
logging.info(f"[GATE] Baseline trade rate (no model): {baseline_trade_rate:.4f}")

# 3) Train gate model
gate_model = RandomForestClassifier(
    n_estimators=300,
    max_depth=8,
    min_samples_leaf=10,
    min_samples_split=50,
    class_weight="balanced_subsample",
    random_state=42,
    n_jobs=-1,
)

gate_model.fit(X_train, y_train)

# 4) Predict probabilities
val_proba = gate_model.predict_proba(X_val)[:, 1]

# 5) Tune threshold
threshold_grid = np.round(np.arange(0.30, 0.81, 0.02), 2)

threshold_rows = []
best_threshold = 0.50
best_score = -np.inf

for thr in threshold_grid:
    pred = (val_proba >= thr).astype(int)

    trade_rate = float(pred.mean())
    precision = precision_score(y_val, pred, zero_division=0)
    recall = recall_score(y_val, pred, zero_division=0)
    f1 = f1_score(y_val, pred, zero_division=0)
    acc = accuracy_score(y_val, pred)

    # Choose your objective:
    # Here: favor F1, but lightly penalize too-few trades
    score = f1 - 0.10 * max(0.0, 0.05 - trade_rate)

    threshold_rows.append({
        "threshold": float(thr),
        "trade_rate": trade_rate,
        "precision": float(precision),
        "recall": float(recall),
        "f1": float(f1),
        "accuracy": float(acc),
        "score": float(score),
    })

    if score > best_score:
        best_score = score
        best_threshold = float(thr)

threshold_df = pd.DataFrame(threshold_rows).sort_values("threshold").reset_index(drop=True)

# 6) Final validation predictions with best threshold
val_gate_pred = (val_proba >= best_threshold).astype(int)

gate_precision = precision_score(y_val, val_gate_pred, zero_division=0)
gate_recall = recall_score(y_val, val_gate_pred, zero_division=0)
gate_f1 = f1_score(y_val, val_gate_pred, zero_division=0)
gate_acc = accuracy_score(y_val, val_gate_pred)
gate_cm = confusion_matrix(y_val, val_gate_pred)

logging.info(f"[GATE] Best threshold: {best_threshold:.2f}")
logging.info(
    f"[GATE] Val metrics | "
    f"acc={gate_acc:.4f}, precision={gate_precision:.4f}, "
    f"recall={gate_recall:.4f}, f1={gate_f1:.4f}"
)
logging.info(f"[GATE] Confusion matrix:\n{gate_cm}")

# 7) Save validation predictions
gate_val_preds = val_df[["Datetime", "Symbol", "Target", "Return"]].copy()
gate_val_preds["Gate_Target"] = y_val.values
gate_val_preds["Gate_Prob"] = val_proba
gate_val_preds["Gate_Pred"] = val_gate_pred

gate_val_preds_path = os.path.join(GATE_DIR, "gate_val_predictions.csv")
gate_val_preds.to_csv(gate_val_preds_path, index=False)

# 8) Save threshold sweep
thresholds_path = os.path.join(GATE_DIR, "gate_threshold_sweep.csv")
threshold_df.to_csv(thresholds_path, index=False)

# 9) Save config + metrics
gate_summary = {
    "model_type": "RandomForestClassifier",
    "best_threshold": best_threshold,
    "val_accuracy": float(gate_acc),
    "val_precision": float(gate_precision),
    "val_recall": float(gate_recall),
    "val_f1": float(gate_f1),
    "train_rows": int(len(train_df)),
    "val_rows": int(len(val_df)),
    "train_positive_rate": float(y_train.mean()),
    "val_positive_rate": float(y_val.mean()),
    "n_features": int(len(gate_features)),
    "features": gate_features,
    "confusion_matrix": gate_cm.tolist(),
}

gate_summary_path = os.path.join(GATE_DIR, "gate_summary.json")
with open(gate_summary_path, "w") as f:
    json.dump(gate_summary, f, indent=2)

# 10) Save model + feature list
gate_model_path = os.path.join(GATE_DIR, "gate_model.joblib")
joblib.dump(gate_model, gate_model_path)

gate_features_path = os.path.join(GATE_DIR, "gate_features.json")
with open(gate_features_path, "w") as f:
    json.dump(gate_features, f, indent=2)

gate_threshold_path = os.path.join(GATE_DIR, "gate_threshold.json")
with open(gate_threshold_path, "w") as f:
    json.dump({"best_threshold": best_threshold}, f, indent=2)

# 11) Record artifacts in manifest
manifest_record("gate_model", gate_model_path)
manifest_record("gate_features", gate_features_path)
manifest_record("gate_threshold", gate_threshold_path)
manifest_record("gate_summary", gate_summary_path)
manifest_record("gate_val_predictions", gate_val_preds_path)
manifest_record("gate_threshold_sweep", thresholds_path)

print("\n===== GATE SUMMARY =====")
print(f"Best threshold: {best_threshold:.2f}")
print(f"Accuracy:  {gate_acc:.4f}")
print(f"Precision: {gate_precision:.4f}")
print(f"Recall:    {gate_recall:.4f}")
print(f"F1:        {gate_f1:.4f}")
print(f"Saved gate outputs to: {GATE_DIR}")

print(f"[SAVE CHECK] DRIVE_BASE = {DRIVE_BASE}")
print(f"[SAVE CHECK] DRIVE_DIR  = {DRIVE_DIR}")

if "MyDrive" not in DRIVE_DIR and "content/drive" not in DRIVE_DIR:
    logging.warning(f"[SAVE CHECK] Not saving to Google Drive. DRIVE_DIR={DRIVE_DIR}")

# ============================================================
# SAVE LOCALLY (CSV + Parquet) + DRIVE (CSV)
# ============================================================
final_df.to_csv(LOCAL_OUT, index=False)
train_df.to_csv(LOCAL_TRAIN, index=False)
val_df.to_csv(LOCAL_VAL, index=False)

manifest_record("local_full_csv",  os.path.abspath(LOCAL_OUT))
manifest_record("local_train_csv", os.path.abspath(LOCAL_TRAIN))
manifest_record("local_val_csv",   os.path.abspath(LOCAL_VAL))

logging.info(f"Saved local CSVs: {LOCAL_OUT}, {LOCAL_TRAIN}, {LOCAL_VAL}")

final_df.to_parquet(PARQ_FULL, index=False)
train_df.to_parquet(PARQ_TRAIN, index=False)
val_df.to_parquet(PARQ_VAL, index=False)

manifest_record("local_full_parquet",  os.path.abspath(PARQ_FULL))
manifest_record("local_train_parquet", os.path.abspath(PARQ_TRAIN))
manifest_record("local_val_parquet",   os.path.abspath(PARQ_VAL))

logging.info(f"Saved local Parquets: {PARQ_FULL}, {PARQ_TRAIN}, {PARQ_VAL}")

# Save to Google Drive (CSV)
final_path = os.path.join(DRIVE_DIR, LOCAL_OUT)
train_path = os.path.join(DRIVE_DIR, LOCAL_TRAIN)
val_path   = os.path.join(DRIVE_DIR, LOCAL_VAL)

final_df.to_csv(final_path, index=False)
train_df.to_csv(train_path, index=False)
val_df.to_csv(val_path, index=False)

manifest_record("drive_full_csv",  final_path)
manifest_record("drive_train_csv", train_path)
manifest_record("drive_val_csv",   val_path)

logging.info(f"Saved to Google Drive:\n- {final_path}\n- {train_path}\n- {val_path}")

print(final_df[["Datetime", "Context_Datetime"]].head(10))
print(final_df.columns.tolist())

print("\n===== DATE RANGE =====")
print("Min:", final_df["Datetime"].min())
print("Max:", final_df["Datetime"].max())

print("\n===== SYMBOL COUNTS =====")
print(final_df["Symbol"].value_counts())

print("\n===== TARGET DISTRIBUTION =====")
print(final_df["Target"].value_counts(normalize=True))

print("\n===== MISSING VALUES =====")
na = final_df.isna().sum()
print(na[na > 0].sort_values(ascending=False))

print("\n===== SAMPLE CONTEXT MERGE CHECK =====")
print(final_df[["Datetime", "Context_Datetime"]].head(10))

sample = final_df[final_df["Symbol"] == "AAPL"].iloc[:500].copy()

plt.figure(figsize=(10,4))
plt.plot(sample["Datetime"], sample["Close"], label="Close")
plt.title("AAPL Sample Price")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ============================================================
# CLEANUP
# ============================================================
del all_dfs, final_df, train_df, val_df
gc.collect()
```

```python
from __future__ import annotations
from pathlib import Path

# -------------------------
# Imports
# -------------------------
import os
import gc
import time
import json
import logging
import glob
import shutil
import heapq
import random
import platform
import hashlib
import csv
import sys
from threading import Lock
from datetime import datetime, timezone

import numpy as np

import torch
import gymnasium as gym
from gymnasium.spaces import Box as GBox

from gym_anytrading.envs import StocksEnv

from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv, VecNormalize
from stable_baselines3.common.utils import set_random_seed

import joblib
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, precision_score, recall_score

from copy import deepcopy

print("\n🔍 Checking dataset paths...")
for path in DATA_CANDIDATES:
    print(path, "->", os.path.exists(path))

# -------------------------
# Basic Logging
# -------------------------
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    force=True
)

# Optional Phase 3 dependency
try:
    from arch import arch_model
    ARCH_AVAILABLE = True
except Exception:
    ARCH_AVAILABLE = False

# Optional Phase 5 dependency (lightweight VMD)
try:
    from vmdpy import VMD
    VMD_AVAILABLE = True
    logging.info("[VMD] vmdpy import OK")
except Exception as e:
    VMD_AVAILABLE = False
    logging.warning(f"[VMD] vmdpy import failed. VMD disabled. Err: {e}")

# -------------------------
# Repro
# -------------------------
SEED = 42


def seed_everything(seed: int = 42):
    random.seed(seed)
    np.random.seed(seed)
    os.environ["PYTHONHASHSEED"] = str(seed)

    try:
        torch.manual_seed(seed)
        if torch.cuda.is_available():
            torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False
    except Exception:
        pass

    try:
        set_random_seed(seed)
    except Exception:
        pass


seed_everything(SEED)


def now_utc() -> datetime:
    return datetime.now(timezone.utc)


# -------------------------
# Annualization heuristic (intraday: 6.5 hrs * 252)
# -------------------------
def _annualization_factor() -> float:
    return float(np.sqrt(252 * 6.5))


def maybe_annualize_vol(vol_s: pd.Series) -> pd.Series:
    vol_s = pd.to_numeric(vol_s, errors="coerce")
    if not VOL_FORECAST_ANNUALIZE:
        return vol_s
    return vol_s * _annualization_factor()


# -------------------------
# Feature engineering import (fallback)
# -------------------------
FEATURE_ENGINEERING_CANDIDATES = [
    os.getcwd(),
    "/content",
    "/content/drive/MyDrive",
    "/content/drive/MyDrive/Results_May_2025",
    "/content/drive/MyDrive/trading_data",
]

for p in FEATURE_ENGINEERING_CANDIDATES:
    if p not in sys.path and os.path.exists(p):
        sys.path.append(p)

FEATURES_FALLBACK_ACTIVE = False
FEATURES_IMPORT_ERROR = None

try:
    from feature_engineering import compute_enhanced_features
    logging.info("[FEATURES] feature_engineering import OK")
except Exception as e:
    FEATURES_FALLBACK_ACTIVE = True
    FEATURES_IMPORT_ERROR = str(e)

    def compute_enhanced_features(df_in: pd.DataFrame) -> pd.DataFrame:
        df = df_in.copy()
        if "Datetime" in df.columns:
            df["Datetime"] = pd.to_datetime(df["Datetime"], errors="coerce")
            df = df.sort_values("Datetime").reset_index(drop=True)
        if "Close" not in df.columns:
            raise ValueError("compute_enhanced_features fallback: missing 'Close'")
        return df

# -------------------------
# Flags / knobs
# -------------------------
ENABLE_VOL_CONTROLS = True
VOL_LOOKBACK = 20
VOL_EWM_SPAN = 20
VOL_GATE_Q = 0.995
VOL_SCALE_K = 0.75
VOL_MIN_SCALE = 0.65

ENABLE_RISK_FORECAST_UPGRADE = True
VOL_FORECAST_CANDIDATES = ["ewm", "rolling", "garch"]
VOL_FORECAST_BASELINE = "ewm"
VOL_FORECAST_SELECTION_METRIC = "qlike"
VOL_FORECAST_MIN_IMPROVEMENT = 0.00
VOL_FORECAST_TRAIN_FRAC = 0.70
VOL_FORECAST_GARCH_P = 1
VOL_FORECAST_GARCH_Q = 1
VOL_FORECAST_GARCH_DIST = "normal"
VOL_FORECAST_ANNUALIZE = False
VOL_FORECAST_LOG_EVAL = False
VOL_FORECAST_GARCH_REFIT_EVERY = 25
VOL_FORECAST_GARCH_MIN_TRAIN = 120

TP_BASE = 0.045
SL_BASE = 0.030
TP_MIN, TP_MAX = 0.015, 0.100
SL_MIN, SL_MAX = 0.010, 0.080

ENABLE_SENTIMENT = False
ENABLE_DRAWDOWN_CIRCUIT_BREAKER = True
ENABLE_WAVELET = True

TEST_MODE = True
ENABLE_PLOTS = False
LIVE_MODE = False
SIM_LATENCY_MS = 0
BROKER = "log"
MAX_WORKERS = 1
FORCE_RETRAIN = False

# -------------------------
# Phase 4 Config
# -------------------------
ENABLE_PHASE4_EVAL = True
WALKFORWARD_TRAIN_FRAC = 0.70
WALKFORWARD_MIN_TRAIN_ROWS = 600
WALKFORWARD_MIN_EVAL_ROWS = 250

ENABLE_BOOTSTRAP_CI = False
BOOTSTRAP_B = 500 if not TEST_MODE else 200
BOOTSTRAP_BLOCK_SIZE = 24
BOOTSTRAP_ALPHA = 0.05

ENABLE_ABLATION_TABLES = True
ENABLE_REGIME_BREAKDOWN = True

# Optional placeholder for VMD-aware datasets
ENABLE_VMD = False
logging.info("[VMD] Disabled for this run.")

VMD_SOURCE_COL = "Denoised_Close"
VMD_LOOKBACK = 128
VMD_K = 5
VMD_ALPHA = 2000
VMD_TAU = 0.0
VMD_DC = 0
VMD_INIT = 1
VMD_TOL = 1e-7

# Keep these defined even if ENABLE_VMD = False
VMD_TREND_MODES = [0, 1]
VMD_CYCLE_MODES = [2, 3]
VMD_NOISE_MODES = [4]

ENABLE_VMD_MODEL_SELECTION = True
VMD_MIN_SHARPE_IMPROVEMENT = 0.10
VMD_REQUIRE_FINAL_NOT_WORSE = True

# -------------------------
# Phase 2 Gate Config
# -------------------------
ENABLE_TRADE_GATE = True
GATE_MODEL_KIND = "rf"
GATE_LABEL_HORIZON = 5
GATE_LABEL_RETURN_THR = 0.005
GATE_THRESHOLD = 0.50
GATE_MIN_TRAIN_ROWS = 200
GATE_TRAIN_FRACTION = 0.70
GATE_TOP_EXPORT_TO_QC = True

# Back-compat alias
test_mode = TEST_MODE

if LIVE_MODE and FEATURES_FALLBACK_ACTIVE:
    raise RuntimeError("LIVE_MODE=True but compute_enhanced_features fallback active. Fix import first.")


# -------------------------
# Paths (Colab or local)
# -------------------------
def _detect_colab() -> bool:
    try:
        import google.colab  # noqa: F401
        return True
    except Exception:
        return False


IN_COLAB = _detect_colab()

try:
    if IN_COLAB:
        from google.colab import drive
        drive.mount("/content/drive")
        DRIVE_BASE = "/content/drive/MyDrive"
    else:
        DRIVE_BASE = os.getcwd()
except Exception:
    DRIVE_BASE = os.getcwd()

BASE_RESULTS_DIR = os.path.join(DRIVE_BASE, "Results_May_2025")
os.makedirs(BASE_RESULTS_DIR, exist_ok=True)

RUN_TAG = datetime.now().strftime("%Y%m%d_%H%M%S")
RUN_RESULTS_DIR = os.path.join(BASE_RESULTS_DIR, f"ppo_walkforward_results_{RUN_TAG}")
FINAL_MODEL_DIR = os.path.join(BASE_RESULTS_DIR, "ppo_models_master")
QC_TOP_DIR = os.path.join(BASE_RESULTS_DIR, "ppo_models_QC_TOP")

PHASE4_ABLATION_PATH = os.path.join(RUN_RESULTS_DIR, "phase4_ablation_all_windows.csv")
PHASE4_REGIME_PATH = os.path.join(RUN_RESULTS_DIR, "phase4_regime_all_windows.csv")
PHASE4_SUMMARY_PATH = os.path.join(RUN_RESULTS_DIR, "phase4_summary_by_ticker.csv")

os.makedirs(RUN_RESULTS_DIR, exist_ok=True)
os.makedirs(FINAL_MODEL_DIR, exist_ok=True)
os.makedirs(QC_TOP_DIR, exist_ok=True)


# -------------------------
# Force-retrain cleanup
# -------------------------
def safe_reset_dir(dir_path: str):
    if os.path.isdir(dir_path):
        shutil.rmtree(dir_path)
    os.makedirs(dir_path, exist_ok=True)


def cleanup_prefix_artifacts(base_dir: str, prefix: str):
    if not os.path.isdir(base_dir):
        return

    patterns = [
        f"{prefix}_model.zip",
        f"{prefix}_vecnorm.pkl",
        f"{prefix}_features.json",
        f"{prefix}_probability_config.json",
        f"{prefix}_model_info.json",
        f"{prefix}_predictions.csv",
        f"{prefix}_predictions_compat.csv",
        f"{prefix}_trades.csv",
        f"{prefix}_step_log.csv",
        f"{prefix}_eval.csv",
        f"{prefix}_metrics.json",
        f"{prefix}_ablation.csv",
        f"{prefix}_regime_breakdown.csv",
        f"{prefix}_bootstrap_ci.json",
        f"{prefix}_gate_model.pkl",
        f"{prefix}_gate_features.json",
        f"{prefix}_gate_config.json",
    ]

    for pat in patterns:
        for fp in glob.glob(os.path.join(base_dir, pat)):
            try:
                if os.path.isfile(fp):
                    os.remove(fp)
            except Exception as e:
                logging.warning(f"[FORCE_RETRAIN] Could not delete {fp}: {e}")


def cleanup_for_force_retrain(
    run_results_dir: str,
    final_model_dir: str,
    qc_top_dir: str,
    tickers: list[str] | None = None,
    max_windows: int = 50,
):
    logging.info("[FORCE_RETRAIN] Starting artifact cleanup...")
    safe_reset_dir(run_results_dir)

    if tickers:
        for ticker in tickers:
            for w_idx in range(1, max_windows + 1):
                prefix = f"ppo_{ticker}_window{w_idx}"
                cleanup_prefix_artifacts(final_model_dir, prefix)
                cleanup_prefix_artifacts(qc_top_dir, prefix)

    logging.info("[FORCE_RETRAIN] Artifact cleanup complete.")


SELECTOR_FULL_PATH = os.path.join(BASE_RESULTS_DIR, "ppo_model_selector_FULL.csv")
SELECTOR_JSON_PATH = os.path.join(BASE_RESULTS_DIR, "ppo_model_selector_final.json")

SKIP_AGG_PATH = os.path.join(RUN_RESULTS_DIR, "skipped_windows_global.csv")
SKIP_LOCK = Lock()


# -------------------------
# Device gate
# -------------------------
FORCE_PPO_CPU = True  # ✅ set to False later when stable

if FORCE_PPO_CPU:
    USE_GPU = False
    DEVICE_STR = "cpu"
else:
    USE_GPU = torch.cuda.is_available() and (MAX_WORKERS == 1)
    DEVICE_STR = "cuda" if USE_GPU else "cpu"

logging.info(
    f"[CONFIG] IN_COLAB={IN_COLAB} | MAX_WORKERS={MAX_WORKERS} | "
    f"cuda={torch.cuda.is_available()} | TEST_MODE={TEST_MODE} | DEVICE={DEVICE_STR}"
)


# -------------------------
# Run-level artifact manifest
# -------------------------
ENABLE_HASHES = False
ARTIFACT_MANIFEST_PATH = os.path.join(RUN_RESULTS_DIR, "artifact_manifest.json")


def _sha256_file(path: str, chunk_size: int = 1024 * 1024) -> str:
    h = hashlib.sha256()
    with open(path, "rb") as f:
        while True:
            b = f.read(chunk_size)
            if not b:
                break
            h.update(b)
    return h.hexdigest()


def _safe_json_dump(obj, path: str):
    os.makedirs(os.path.dirname(path), exist_ok=True)
    tmp = path + ".tmp"
    with open(tmp, "w") as f:
        json.dump(obj, f, indent=2, default=str)
    os.replace(tmp, path)


def load_manifest() -> dict:
    if os.path.exists(ARTIFACT_MANIFEST_PATH):
        try:
            with open(ARTIFACT_MANIFEST_PATH, "r") as f:
                return json.load(f)
        except Exception:
            pass

    return {
        "run_tag": RUN_TAG,
        "created_utc": now_utc().isoformat(),
        "run_results_dir": RUN_RESULTS_DIR,
        "final_model_dir": FINAL_MODEL_DIR,
        "qc_top_dir": QC_TOP_DIR,
        "environment": {
            "python": platform.python_version(),
            "platform": platform.platform(),
            "torch": getattr(torch, "__version__", None),
            "gymnasium": getattr(gym, "__version__", None),
            "stable_baselines3": None,
            "pandas": getattr(pd, "__version__", None),
            "numpy": getattr(np, "__version__", None),
            "yfinance": None,
        },
        "artifacts": {}
    }

MANIFEST = load_manifest()
try:
    import stable_baselines3
    MANIFEST["environment"]["stable_baselines3"] = stable_baselines3.__version__
except Exception:
    pass
_safe_json_dump(MANIFEST, ARTIFACT_MANIFEST_PATH)


def manifest_update(prefix: str, updates: dict):
    MANIFEST["artifacts"].setdefault(prefix, {})
    MANIFEST["artifacts"][prefix].update(updates)
    MANIFEST["artifacts"][prefix]["updated_utc"] = now_utc().isoformat()
    _safe_json_dump(MANIFEST, ARTIFACT_MANIFEST_PATH)


def manifest_record_paths(prefix: str, paths: dict, do_hashes: bool = False):
    rec = {}
    for k, p in paths.items():
        rec[k] = {"path": p, "exists": bool(p and os.path.exists(p))}
        if do_hashes and p and os.path.exists(p):
            try:
                rec[k]["sha256"] = _sha256_file(p)
            except Exception as e:
                rec[k]["sha256_error"] = str(e)
    manifest_update(prefix, {"paths": rec})


# ============================================================
# Bundle Writer (uses model_info passed in — CORRECT)
# ============================================================

def write_master_bundle(
    prefix: str,
    features: list[str],
    threshold: float,
    model_info: dict,
    gate_features: list[str] | None = None,
    gate_threshold: float | None = None,
):
    f_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_features.json")
    p_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_probability_config.json")
    i_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_model_info.json")

    _safe_json_dump({"features": list(features)}, f_path)

    _safe_json_dump(
        {
            "threshold": float(threshold),
            "use_confidence": True,
            "inference_mode": "deterministic",
        },
        p_path
    )

    _safe_json_dump(model_info, i_path)

    record_paths = {
        "master_features_json": f_path,
        "master_probability_config_json": p_path,
        "master_model_info_json": i_path,
    }

    # -------------------------
    # Gate artifacts (optional)
    # -------------------------
    if gate_features is not None:
        gf_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_gate_features.json")
        gc_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_gate_config.json")

        _safe_json_dump({"features": list(gate_features)}, gf_path)

        _safe_json_dump(
            {
                "gate_enabled": True,
                "gate_model_type": GATE_MODEL_KIND,
                "threshold": float(
                    gate_threshold if gate_threshold is not None else GATE_THRESHOLD
                ),
                "label_horizon": int(GATE_LABEL_HORIZON),
                "label_return_threshold": float(GATE_LABEL_RETURN_THR),
                "uses_scaler": False,
            },
            gc_path,
        )

        record_paths["master_gate_features_json"] = gf_path
        record_paths["master_gate_config_json"] = gc_path

    # Save all paths into manifest
    manifest_record_paths(prefix, record_paths, do_hashes=ENABLE_HASHES)

def save_gate_model(prefix: str, gate_model, gate_features: list[str], gate_config: dict):
    gm_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_gate_model.pkl")
    gf_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_gate_features.json")
    gc_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_gate_config.json")

    joblib.dump(gate_model, gm_path)
    _safe_json_dump({"features": list(gate_features)}, gf_path)
    _safe_json_dump(gate_config, gc_path)

    manifest_record_paths(
        prefix,
        {
            "gate_model_pkl": gm_path,
            "gate_features_json": gf_path,
            "gate_config_json": gc_path,
        },
        do_hashes=ENABLE_HASHES,
    )
    return gm_path, gf_path, gc_path


# -------------------------
# Health check helper
# -------------------------
def health_check_latest_run(base_dir: str) -> dict:
    run_dirs = sorted(glob.glob(os.path.join(base_dir, "ppo_walkforward_results_*")))[::-1]
    if not run_dirs:
        return {"ok": False, "reason": "No run folders found."}

    latest = run_dirs[0]
    summaries = sorted(glob.glob(os.path.join(latest, "summary*.csv")))
    out = {
        "ok": True,
        "latest_run_dir": latest,
        "summary_files": summaries,
        "rows_total": 0,
        "tickers": [],
    }

    frames = []
    for p in summaries:
        try:
            frames.append(pd.read_csv(p))
        except Exception:
            pass

    if frames:
        combo = pd.concat(frames, ignore_index=True)
        out["rows_total"] = int(len(combo))
        if "Ticker" in combo.columns:
            out["tickers"] = sorted(list(combo["Ticker"].dropna().astype(str).unique()))
    else:
        out["ok"] = False
        out["reason"] = "Latest run has no readable summary*.csv"

    return out



# -------------------------
# Data load (training uses prebuilt CSV)
# -------------------------
DATA_CANDIDATES = [
    "multi_stock_feature_engineered_dataset_mt.csv",
    os.path.join(DRIVE_BASE, "trading_data", "multi_stock_feature_engineered_dataset_mt.csv"),
]

print("\n🔍 Checking dataset paths...")
for path in DATA_CANDIDATES:
    print(path, "->", os.path.exists(path))

DATA_PATH = next((p for p in DATA_CANDIDATES if os.path.exists(p)), None)
print("Checked paths:", DATA_CANDIDATES)

if not DATA_PATH:
    raise FileNotFoundError(
        "Could not find multi_stock_feature_engineered_dataset_mt.csv or multi_stock_feature_engineered_dataset.csv in cwd or Drive/trading_data."
    )

df = pd.read_csv(DATA_PATH)

print("\n[DATA CHECK] Using dataset:", DATA_PATH)
print("[DATA CHECK] Shape:", df.shape)
print("[DATA CHECK] First 20 columns:", df.columns[:20].tolist())

if FEATURES_FALLBACK_ACTIVE:
    logging.warning(
        f"[FEATURES] feature_engineering import failed; fallback active. "
        f"This run uses a prebuilt CSV, so training can still proceed if engineered columns already exist. "
        f"Err: {FEATURES_IMPORT_ERROR}"
    )
else:
    logging.info("[FEATURES] Running with full feature engineering module.")

if "Datetime" not in df.columns:
    raise ValueError("Dataset missing 'Datetime' column.")
df["Datetime"] = pd.to_datetime(df["Datetime"], errors="coerce")
df = df.dropna(subset=["Datetime"]).reset_index(drop=True)

if "Symbol" not in df.columns:
    raise ValueError("Dataset missing 'Symbol' column.")
if "Close" not in df.columns:
    raise ValueError("Dataset missing 'Close' column.")

if ENABLE_WAVELET and "Denoised_Close" not in df.columns:
    logging.warning("ENABLE_WAVELET=True but 'Denoised_Close' missing; copying Close -> Denoised_Close.")
    df["Denoised_Close"] = df["Close"]

RUN_CONFIG_PATH = os.path.join(RUN_RESULTS_DIR, "config.json")
FEATURE_MANIFEST_PATH = os.path.join(RUN_RESULTS_DIR, "feature_manifest.json")
DATA_PROVENANCE_PATH = os.path.join(RUN_RESULTS_DIR, "data_provenance.json")


def build_ordered_feature_list(df_in: pd.DataFrame) -> list[str]:
    exclude = {"Datetime", "Symbol", "Target", "Return"}
    feats = [
        str(c) for c in df_in.columns
        if c not in exclude and pd.api.types.is_numeric_dtype(df_in[c])
    ]
    return sorted(feats)

def build_data_provenance_from_df(df_in: pd.DataFrame) -> dict:
    prov = {
        "source": "prebuilt_csv",
        "data_path": os.path.abspath(DATA_PATH),
        "data_sha256": _sha256_file(DATA_PATH) if ENABLE_HASHES else None,
        "rows_total": int(len(df_in)),
        "symbols": sorted(df_in["Symbol"].dropna().astype(str).unique().tolist()),
        "per_symbol": {},
    }
    for sym, g in df_in.groupby("Symbol", sort=False):
        g2 = g.sort_values("Datetime")
        prov["per_symbol"][str(sym)] = {
            "rows": int(len(g2)),
            "start": str(g2["Datetime"].iloc[0]),
            "end": str(g2["Datetime"].iloc[-1]),
        }
    return prov


RUN_CONFIG = {
    "run_tag": RUN_TAG,
    "created_utc": now_utc().isoformat(),
    "paths": {
        "run_results_dir": RUN_RESULTS_DIR,
        "final_model_dir": FINAL_MODEL_DIR,
        "qc_top_dir": QC_TOP_DIR,
        "base_results_dir": BASE_RESULTS_DIR,
        "dataset_csv": os.path.abspath(DATA_PATH),
    },
    "seed": SEED,
    "device": DEVICE_STR,
    "flags": {
        "ENABLE_RISK_FORECAST_UPGRADE": ENABLE_RISK_FORECAST_UPGRADE,
        "ENABLE_VOL_CONTROLS": ENABLE_VOL_CONTROLS,
        "ENABLE_SENTIMENT": ENABLE_SENTIMENT,
        "ENABLE_WAVELET": ENABLE_WAVELET,
        "ENABLE_DRAWDOWN_CIRCUIT_BREAKER": ENABLE_DRAWDOWN_CIRCUIT_BREAKER,
        "LIVE_MODE": LIVE_MODE,
        "TEST_MODE": TEST_MODE,
        "MAX_WORKERS": MAX_WORKERS,
        "FORCE_RETRAIN": FORCE_RETRAIN,
        "FEATURES_FALLBACK_ACTIVE": FEATURES_FALLBACK_ACTIVE,
        "ENABLE_TRADE_GATE": ENABLE_TRADE_GATE,
        "ENABLE_PHASE4_EVAL": ENABLE_PHASE4_EVAL,
        "ENABLE_BOOTSTRAP_CI": ENABLE_BOOTSTRAP_CI,
        "ENABLE_ABLATION_TABLES": ENABLE_ABLATION_TABLES,
        "ENABLE_REGIME_BREAKDOWN": ENABLE_REGIME_BREAKDOWN,
    },
    "phase4_config": {
        "walkforward_train_frac": WALKFORWARD_TRAIN_FRAC,
        "walkforward_min_train_rows": WALKFORWARD_MIN_TRAIN_ROWS,
        "walkforward_min_eval_rows": WALKFORWARD_MIN_EVAL_ROWS,
        "bootstrap_b": BOOTSTRAP_B,
        "bootstrap_block_size": BOOTSTRAP_BLOCK_SIZE,
        "bootstrap_alpha": BOOTSTRAP_ALPHA,
    },
    "feature_pipeline": {
        "fallback_active": FEATURES_FALLBACK_ACTIVE,
        "module_import_ok": not FEATURES_FALLBACK_ACTIVE,
    },
    "gate_config": {
        "enabled": ENABLE_TRADE_GATE,
        "model_kind": GATE_MODEL_KIND,
        "label_horizon": GATE_LABEL_HORIZON,
        "label_return_threshold": GATE_LABEL_RETURN_THR,
        "threshold": GATE_THRESHOLD,
    },
    "risk_forecast_config": {
        "candidates": VOL_FORECAST_CANDIDATES,
        "baseline": VOL_FORECAST_BASELINE,
        "selection_metric": VOL_FORECAST_SELECTION_METRIC,
        "min_improvement": VOL_FORECAST_MIN_IMPROVEMENT,
        "train_fraction": VOL_FORECAST_TRAIN_FRAC,
        "garch_p": VOL_FORECAST_GARCH_P,
        "garch_q": VOL_FORECAST_GARCH_Q,
        "garch_dist": VOL_FORECAST_GARCH_DIST,
        "arch_available": bool(ARCH_AVAILABLE),
    },
}
_safe_json_dump(RUN_CONFIG, RUN_CONFIG_PATH)

ORDERED_FEATURES = build_ordered_feature_list(df)
_safe_json_dump({"features_ordered": ORDERED_FEATURES}, FEATURE_MANIFEST_PATH)

PROVENANCE = build_data_provenance_from_df(df)
_safe_json_dump(PROVENANCE, DATA_PROVENANCE_PATH)

manifest_record_paths(
    "RUN_LEVEL",
    {
        "config_json": RUN_CONFIG_PATH,
        "feature_manifest_json": FEATURE_MANIFEST_PATH,
        "data_provenance_json": DATA_PROVENANCE_PATH,
        "dataset_csv": os.path.abspath(DATA_PATH),
    },
    do_hashes=ENABLE_HASHES,
)


# -------------------------
# Skip logging
# -------------------------
def record_skips_global(
    ticker: str,
    skipped_windows: list[str],
    total_windows: int | None = None,
    fully_skipped: bool = False,
):
    if not skipped_windows and not fully_skipped:
        return

    with SKIP_LOCK:
        new_file = not os.path.exists(SKIP_AGG_PATH)
        with open(SKIP_AGG_PATH, "a", newline="") as f:
            w = csv.writer(f)
            if new_file:
                w.writerow(["Ticker", "Window", "FullySkipped", "TotalWindows"])

            if fully_skipped:
                w.writerow([ticker, "ALL", True, total_windows if total_windows is not None else ""])
            else:
                for wname in skipped_windows:
                    win = ""
                    try:
                        _, win_str = wname.split("_window")
                        win = int(win_str)
                    except Exception:
                        pass
                    w.writerow([ticker, win, False, total_windows if total_windows is not None else ""])


# -------------------------
# Helpers: volatility forecasts
# -------------------------
def _safe_series(x, name: str) -> pd.Series:
    s = pd.Series(x, copy=True)
    s = pd.to_numeric(s, errors="coerce")
    s.name = name
    return s


def realized_vol_proxy(ret_s: pd.Series) -> pd.Series:
    r = _safe_series(ret_s, "ret").fillna(0.0)
    return (r ** 2).shift(-1)


def rolling_vol_forecast(ret_s: pd.Series, lookback: int) -> pd.Series:
    r = _safe_series(ret_s, "ret").fillna(0.0)
    return r.rolling(lookback).std()


def ewm_vol_forecast(ret_s: pd.Series, span: int) -> pd.Series:
    r = _safe_series(ret_s, "ret").fillna(0.0)
    return r.ewm(span=span, adjust=False).std()

def garch_vol_forecast(
    ret_s: pd.Series,
    p: int = 1,
    q: int = 1,
    dist: str = "normal",
    refit_every: int = 25,
    min_train_size: int = 120,
) -> pd.Series:
    r = _safe_series(ret_s, "ret")
    out = pd.Series(np.nan, index=r.index, name="garch_vol")

    if (not ARCH_AVAILABLE) or len(r.dropna()) < min_train_size:
        return out

    start_idx = max(min_train_size, p + q + 5)
    start_idx = min(start_idx, len(r) - 2)
    if start_idx < min_train_size:
        return out

    r_pct = (r.fillna(0.0) * 100.0).astype(float)

    last_vol = np.nan
    for i in range(start_idx, len(r_pct)):
        should_refit = ((i - start_idx) % refit_every == 0) or (not np.isfinite(last_vol))
        if should_refit:
            hist = r_pct.iloc[:i].dropna()
            if len(hist) < min_train_size:
                continue
            try:
                am = arch_model(
                    hist,
                    mean="Zero",
                    vol="GARCH",
                    p=p,
                    q=q,
                    dist=dist,
                    rescale=False,
                )
                res = am.fit(disp="off", show_warning=False)
                fcast = res.forecast(horizon=1, reindex=False)
                var_1 = float(fcast.variance.values[-1, 0]) / (100.0 ** 2)
                var_1 = max(var_1, 0.0)
                last_vol = np.sqrt(var_1)
            except Exception:
                last_vol = np.nan
        out.iloc[i] = last_vol
    return out


def qlike_loss(realized_var: pd.Series, forecast_var: pd.Series) -> float:
    rv = pd.to_numeric(realized_var, errors="coerce")
    fv = pd.to_numeric(forecast_var, errors="coerce")
    tmp = pd.DataFrame({"rv": rv, "fv": fv}).dropna()
    if tmp.empty:
        return np.inf
    eps = 1e-12
    rv_ = tmp["rv"].clip(lower=eps)
    fv_ = tmp["fv"].clip(lower=eps)
    return float(np.mean(np.log(fv_) + (rv_ / fv_)))


def mse_loss(realized_var: pd.Series, forecast_var: pd.Series) -> float:
    rv = pd.to_numeric(realized_var, errors="coerce")
    fv = pd.to_numeric(forecast_var, errors="coerce")
    tmp = pd.DataFrame({"rv": rv, "fv": fv}).dropna()
    if tmp.empty:
        return np.inf
    return float(np.mean((tmp["rv"] - tmp["fv"]) ** 2))


def build_vol_forecast_candidates(
    df_window: pd.DataFrame,
    lookback: int,
    ewm_span: int,
    train_frac: float,
    oos_start_idx: int | None = None,
) -> tuple[pd.DataFrame, dict]:
    """
    Phase 4 adjustment:
    choose the forecast method using ONLY the in-window training section,
    not the held-out OOS evaluation section.
    """
    df_local = df_window.copy()
    if "Close" not in df_local.columns:
        raise ValueError("build_vol_forecast_candidates: missing Close")

    df_local["Ret_1"] = pd.to_numeric(df_local["Close"], errors="coerce").pct_change().fillna(0.0)

    ret_s = df_local["Ret_1"].copy()
    realized_var = realized_vol_proxy(ret_s)

    candidate_map = {
        "rolling": rolling_vol_forecast(ret_s, lookback=lookback),
        "ewm": ewm_vol_forecast(ret_s, span=ewm_span),
    }

    if "garch" in VOL_FORECAST_CANDIDATES:
        candidate_map["garch"] = garch_vol_forecast(
            ret_s,
            p=VOL_FORECAST_GARCH_P,
            q=VOL_FORECAST_GARCH_Q,
            dist=VOL_FORECAST_GARCH_DIST,
            refit_every=VOL_FORECAST_GARCH_REFIT_EVERY,
            min_train_size=VOL_FORECAST_GARCH_MIN_TRAIN,
        )
    else:
        candidate_map["garch"] = pd.Series(np.nan, index=df_local.index)

    df_local["Vol_Roll"] = maybe_annualize_vol(candidate_map["rolling"])
    df_local["Vol_EWM"] = maybe_annualize_vol(candidate_map["ewm"])
    df_local["Vol_GARCH"] = maybe_annualize_vol(candidate_map["garch"])

    if oos_start_idx is None:
        selection_end = len(df_local)
    else:
        selection_end = max(min(int(oos_start_idx), len(df_local)), 61)

    split_idx = max(int(selection_end * train_frac), 60)
    split_idx = min(split_idx, selection_end - 1)

    score_fn = qlike_loss if VOL_FORECAST_SELECTION_METRIC.lower() == "qlike" else mse_loss

    scores = {}
    score_coverage = {}

    for name in VOL_FORECAST_CANDIDATES:
        vol_s = candidate_map.get(name, pd.Series(np.nan, index=df_local.index))
        var_s = (vol_s ** 2)

        rv_slice = pd.to_numeric(realized_var.iloc[split_idx:selection_end], errors="coerce")
        fv_slice = pd.to_numeric(var_s.iloc[split_idx:selection_end], errors="coerce")
        valid_mask = rv_slice.notna() & fv_slice.notna()

        coverage = int(valid_mask.sum())
        score_coverage[name] = coverage

        if coverage < 50:
            scores[name] = np.inf
        else:
            scores[name] = score_fn(
                rv_slice[valid_mask],
                fv_slice[valid_mask],
            )

    baseline_name = VOL_FORECAST_BASELINE if VOL_FORECAST_BASELINE in scores else "ewm"
    baseline_score = scores.get(baseline_name, np.inf)

    best_method = baseline_name
    best_score = baseline_score
    for name, sc in scores.items():
        if np.isfinite(sc) and (sc < best_score):
            best_method = name
            best_score = sc

    improvement = baseline_score - best_score if np.isfinite(baseline_score) and np.isfinite(best_score) else 0.0
    if best_method != baseline_name and improvement < VOL_FORECAST_MIN_IMPROVEMENT:
        best_method = baseline_name
        best_score = baseline_score

    selected_series = {
        "rolling": df_local["Vol_Roll"],
        "ewm": df_local["Vol_EWM"],
        "garch": df_local["Vol_GARCH"],
    }.get(best_method, df_local["Vol_EWM"])

    selected_series = pd.to_numeric(selected_series, errors="coerce")
    selected_series = selected_series.fillna(df_local["Vol_EWM"])
    selected_series = selected_series.fillna(df_local["Vol_Roll"])
    selected_series = selected_series.fillna(0.0)

    df_local["Vol_Forecast"] = selected_series
    df_local["Vol_Method_Selected"] = best_method

    vol_ref = float(df_local["Vol_Forecast"].median()) if len(df_local) else 0.0
    vol_gate = float(df_local["Vol_Forecast"].quantile(VOL_GATE_Q)) if len(df_local) else 0.0

    df_local["Vol_Ref"] = vol_ref
    df_local["Vol_Gate"] = vol_gate

    diagnostics = {
        "selected_method": best_method,
        "selection_metric": VOL_FORECAST_SELECTION_METRIC,
        "scores": {k: (None if not np.isfinite(v) else float(v)) for k, v in scores.items()},
        "score_coverage": score_coverage,
        "baseline_method": baseline_name,
        "baseline_score": None if not np.isfinite(baseline_score) else float(baseline_score),
        "best_score": None if not np.isfinite(best_score) else float(best_score),
        "improvement_vs_baseline": float(improvement) if np.isfinite(improvement) else None,
        "arch_available": bool(ARCH_AVAILABLE),
        "vol_ref": float(vol_ref),
        "vol_gate": float(vol_gate),
        "split_idx": int(split_idx),
        "selection_end": int(selection_end),
    }

    return df_local, diagnostics


# -------------------------
# Env kwargs
# -------------------------
ENV_KWARGS = dict(
    window_size=10,
    cost_rate=0.0002,
    slip_rate=0.0003,
    k_alpha=0.0,
    k_mom=0.15,
    k_sent=(0.01 if ENABLE_SENTIMENT else 0.0),
    mom_source="denoised",
    mom_lookback=20,
    min_trade_delta=0.06,
    cooldown=8,
    reward_clip=0.05,
    k_vol=0.00,
    k_dd=0.00,
    enable_slo=ENABLE_DRAWDOWN_CIRCUIT_BREAKER,
    slo_max_dd=0.15,
    slo_lock_steps=15,
    slo_hard_stop_enabled=False,
    slo_hard_stop_pct=0.01,
    slo_chop_enabled=True,
    slo_chop_col="Chop",
    slo_chop_lock_steps=3,
    enable_vol_controls=ENABLE_VOL_CONTROLS,
    vol_col="Vol_Forecast",
    vol_gate_quantile=VOL_GATE_Q,
    vol_scale_k=VOL_SCALE_K,
    vol_min_scale=VOL_MIN_SCALE,
    tp_base=TP_BASE,
    sl_base=SL_BASE,
    tp_min=TP_MIN,
    tp_max=TP_MAX,
    sl_min=SL_MIN,
    sl_max=SL_MAX,
)


# -------------------------
# ContinuousPositionEnv
# -------------------------
class ContinuousPositionEnv(StocksEnv):
    def __init__(self, df: pd.DataFrame, frame_bound: tuple[int, int], **kwargs):
        if "window_size" not in kwargs:
            raise ValueError("ContinuousPositionEnv requires window_size in kwargs.")
        window_size = int(kwargs.pop("window_size"))

        self.cost_rate = float(kwargs.pop("cost_rate", 0.0002))
        self.slip_rate = float(kwargs.pop("slip_rate", 0.0003))
        self.k_alpha = float(kwargs.pop("k_alpha", 0.0))
        self.k_mom = float(kwargs.pop("k_mom", 0.15))
        self.k_sent = float(kwargs.pop("k_sent", 0.0))
        self.mom_source = str(kwargs.pop("mom_source", "denoised"))
        self.mom_lookback = int(kwargs.pop("mom_lookback", 20))
        self.min_trade_delta = float(kwargs.pop("min_trade_delta", 0.04))
        self.cooldown = int(kwargs.pop("cooldown", 6))
        self.reward_clip = float(kwargs.pop("reward_clip", 0.05))
        self.k_vol = float(kwargs.pop("k_vol", 0.0))
        self.k_dd = float(kwargs.pop("k_dd", 0.0))

        self.enable_vol_controls = bool(kwargs.pop("enable_vol_controls", False))
        self.vol_col = str(kwargs.pop("vol_col", "Vol_EWM"))
        self.vol_gate_quantile = float(kwargs.pop("vol_gate_quantile", 0.95))
        self.vol_scale_k = float(kwargs.pop("vol_scale_k", 2.0))
        self.vol_min_scale = float(kwargs.pop("vol_min_scale", 0.25))

        self.tp_base = float(kwargs.pop("tp_base", 0.03))
        self.sl_base = float(kwargs.pop("sl_base", 0.02))
        self.tp_min = float(kwargs.pop("tp_min", 0.01))
        self.tp_max = float(kwargs.pop("tp_max", 0.08))
        self.sl_min = float(kwargs.pop("sl_min", 0.005))
        self.sl_max = float(kwargs.pop("sl_max", 0.06))

        self.enable_slo = bool(kwargs.pop("enable_slo", False))
        self.slo_max_dd = float(kwargs.pop("slo_max_dd", 0.15))
        self.slo_lock_steps = int(kwargs.pop("slo_lock_steps", 15))
        self.slo_hard_stop_enabled = bool(kwargs.pop("slo_hard_stop_enabled", False))
        self.slo_hard_stop_pct = float(kwargs.pop("slo_hard_stop_pct", 0.01))

        self.slo_chop_enabled = bool(kwargs.pop("slo_chop_enabled", False))
        self.slo_chop_col = str(kwargs.pop("slo_chop_col", "Chop"))
        self.slo_chop_lock_steps = int(kwargs.pop("slo_chop_lock_steps", 5))
        self._chop_lock_until = -1

        if kwargs:
            raise ValueError(f"Unexpected env kwargs: {list(kwargs.keys())}")

        super().__init__(df=df.reset_index(drop=True), frame_bound=frame_bound, window_size=window_size)

        self._parent_action_space = self.action_space

        if isinstance(self.observation_space, gym.spaces.Box):
            self.observation_space = GBox(
                low=self.observation_space.low,
                high=self.observation_space.high,
                shape=self.observation_space.shape,
                dtype=self.observation_space.dtype,
            )

        self.nav = 1.0
        self.pos = 0.0
        self.peak_nav = 1.0
        self.nav_history = []
        self.ret_history = []
        self.trade_count = 0
        self._last_trade_step = -self.cooldown
        self._slo_lock_until = -1

        self.entry_price = None
        self.entry_tick = None

        self.action_space = GBox(low=-1.0, high=1.0, shape=(1,), dtype=np.float32)

    def reset(self, **kwargs):
        out = super().reset(**kwargs)
        if isinstance(out, tuple) and len(out) == 2:
            obs, info = out
        else:
            obs, info = out, {}

        self.nav = 1.0
        self.pos = 0.0
        self.peak_nav = 1.0
        self.nav_history = [self.nav]
        self.ret_history = []
        self.trade_count = 0
        self._last_trade_step = -self.cooldown
        self._slo_lock_until = -1
        self.entry_price = None
        self.entry_tick = None
        self._chop_lock_until = -1

        self._vol_ref = float(self.df["Vol_Ref"].iloc[0]) if "Vol_Ref" in self.df.columns else 0.0
        self._vol_gate = float(self.df["Vol_Gate"].iloc[0]) if "Vol_Gate" in self.df.columns else 0.0

        info = info or {}
        info.update({"nav": self.nav, "pos": self.pos, "trade_count": int(self.trade_count)})
        return obs, info

    def _ret_t(self) -> float:
        cur = float(self.df.loc[self._current_tick, "Close"])
        prev = float(self.df.loc[max(self._current_tick - 1, 0), "Close"])
        return 0.0 if prev <= 0 else (cur - prev) / prev

    def _price_t(self) -> float:
        return float(self.df.loc[self._current_tick, "Close"])

    def _vol_t(self) -> float:
        if not self.enable_vol_controls:
            return 0.0
        if self.vol_col in self.df.columns:
            try:
                return float(self.df.loc[self._current_tick, self.vol_col])
            except Exception:
                return 0.0
        return 0.0

    def _mom_signal(self) -> float:
        if self.mom_source == "macd" and "MACD_Line" in self.df.columns:
            recent = self.df["MACD_Line"].iloc[max(self._current_tick - 200, 0):self._current_tick + 1]
            denom = 1e-6 + float(recent.std())
            return float(np.tanh(float(self.df.loc[self._current_tick, "MACD_Line"]) / denom))

        if "Denoised_Close" in self.df.columns and self._current_tick - self.mom_lookback >= 0:
            now = float(self.df.loc[self._current_tick, "Denoised_Close"])
            then = float(self.df.loc[self._current_tick - self.mom_lookback, "Denoised_Close"])
            base = float(self.df.loc[max(self._current_tick - 1, 0), "Close"])
            slope = (now - then) / max(self.mom_lookback, 1)
            return float(np.tanh(10.0 * (slope / max(abs(base), 1e-6))))

        return 0.0

    def _dynamic_tp_sl(self, vol_t: float) -> tuple[float, float]:
        vol_ref = float(getattr(self, "_vol_ref", 0.0))
        eps = 1e-9
        ratio = (vol_ref / max(vol_t, eps)) if vol_ref > 0 else 1.0
        tp = float(np.clip(self.tp_base * ratio, self.tp_min, self.tp_max))
        sl = float(np.clip(self.sl_base * ratio, self.sl_min, self.sl_max))
        return tp, sl

    def _base_ret_next(self) -> float:
        try:
            cur = float(self.df.loc[self._current_tick, "Close"])
            prev = float(self.df.loc[max(self._current_tick - 1, 0), "Close"])
            return 0.0 if prev <= 0 else (cur - prev) / prev
        except Exception:
            return 0.0

    def _step_parent_hold(self):
        cur_as = self.action_space
        try:
            self.action_space = self._parent_action_space
            out = super().step(0)
        finally:
            self.action_space = cur_as

        if len(out) == 5:
            obs, _, terminated, truncated, info = out
            return obs, bool(terminated), bool(truncated), info
        if len(out) == 4:
            obs, _, done, info = out
            return obs, bool(done), False, info
        raise ValueError(f"Unexpected parent step() return length: {len(out)}")

    def step(self, action):
        a = float(np.array(action).squeeze())
        next_pos = float(np.clip(a, -1.0, 1.0))

        forced_close = False
        tp_pct = 0.0
        sl_pct = 0.0
        scale = 1.0
        gated = False
        slo_triggered = False

        r_t = self._ret_t()
        vol_t = self._vol_t()

        if self.enable_vol_controls and (abs(self.pos) > 1e-9) and (self.entry_price is not None):
            tp_pct, sl_pct = self._dynamic_tp_sl(vol_t)
            px = self._price_t()
            pnl_dir = ((px - float(self.entry_price)) / max(float(self.entry_price), 1e-9)) * float(np.sign(self.pos))
            if pnl_dir >= tp_pct or pnl_dir <= -sl_pct:
                next_pos = 0.0
                forced_close = True

        locked = self.enable_slo and (self._current_tick < self._slo_lock_until)
        if locked and not forced_close:
            next_pos = float(self.pos)

        vol_ref = float(getattr(self, "_vol_ref", 0.0))
        vol_gate = float(getattr(self, "_vol_gate", 0.0))

        if self.enable_vol_controls and vol_ref > 0 and vol_t > vol_ref:
            scale = 1.0 / (1.0 + self.vol_scale_k * max(0.0, (vol_t / (vol_ref + 1e-9)) - 1.0))
            scale = float(np.clip(scale, self.vol_min_scale, 1.0))
            next_pos = float(np.clip(next_pos * scale, -1.0, 1.0))

        if self.enable_vol_controls and vol_gate > 0 and vol_t >= vol_gate and not forced_close:
            gated = True
            if abs(self.pos) < 1e-9:
                next_pos = 0.0
            else:
                if np.sign(next_pos) != np.sign(self.pos):
                    next_pos = 0.0
                elif abs(next_pos) > abs(self.pos):
                    next_pos = float(self.pos)

        chop_now = False
        if self.slo_chop_enabled and (self.slo_chop_col in self.df.columns):
            try:
                chop_now = bool(int(self.df.loc[self._current_tick, self.slo_chop_col]) == 1)
            except Exception:
                chop_now = False

        if chop_now:
            self._chop_lock_until = max(self._chop_lock_until, self._current_tick + int(self.slo_chop_lock_steps))
        chop_locked = self._current_tick < self._chop_lock_until

        if chop_locked and not forced_close:
            if abs(self.pos) < 1e-9:
                next_pos = 0.0
            else:
                if np.sign(next_pos) != np.sign(self.pos):
                    next_pos = 0.0
                elif abs(next_pos) > abs(self.pos):
                    next_pos = float(self.pos)

        if forced_close:
            changed = abs(self.pos) > 1e-9
        else:
            changed = (
                (abs(next_pos - self.pos) >= self.min_trade_delta)
                and ((self._current_tick - self._last_trade_step) >= self.cooldown)
            )

        if gated and not forced_close:
            changed = False

        delta_pos = (next_pos - self.pos) if changed else 0.0
        fee_cost = self.cost_rate * abs(delta_pos)
        slip_cost = self.slip_rate * abs(delta_pos)
        trade_cost = fee_cost + slip_cost
        turnover_step = abs(delta_pos)

        pnl_from_prev_pos = self.pos * r_t
        rel_alpha = pnl_from_prev_pos - r_t
        mom_term = self.pos * self._mom_signal()
        shaped = pnl_from_prev_pos + (self.k_mom * mom_term) - trade_cost
        reward = float(np.clip(shaped, -self.reward_clip, self.reward_clip))

        next_nav = self.nav * (1.0 + pnl_from_prev_pos - trade_cost)

        dd_now = 0.0
        if self.enable_slo:
            prev_peak = self.peak_nav
            dd_now = (prev_peak - next_nav) / max(prev_peak, 1e-9)
            if dd_now >= self.slo_max_dd:
                slo_triggered = True
                if abs(self.pos) > 1e-9:
                    trade_cost += (self.cost_rate * abs(self.pos)) + (self.slip_rate * abs(self.pos))
                    next_nav = self.nav * (1.0 + pnl_from_prev_pos - trade_cost)
                self.pos = 0.0
                next_pos = 0.0
                self._slo_lock_until = self._current_tick + int(self.slo_lock_steps)
                self._last_trade_step = self._current_tick
                changed = False
                delta_pos = 0.0
                dd_now = (self.peak_nav - next_nav) / max(self.peak_nav, 1e-9)

        self.nav = float(next_nav)
        self.nav_history.append(self.nav)
        self.peak_nav = max(self.peak_nav, self.nav)

        executed_trade = False
        if changed:
            prev_pos = float(self.pos)
            self.pos = float(next_pos)
            self._last_trade_step = self._current_tick
            self.trade_count += 1
            executed_trade = True

            if abs(self.pos) < 1e-9:
                self.entry_price = None
                self.entry_tick = None
            else:
                if (abs(prev_pos) < 1e-9) or (np.sign(prev_pos) != np.sign(self.pos)):
                    self.entry_price = self._price_t()
                    self.entry_tick = int(self._current_tick)

        obs, terminated, truncated, info = self._step_parent_hold()
        base_ret = float(self._base_ret_next())

        info = info or {}
        info.update({
            "ret_t": float(r_t),
            "base_ret": float(base_ret),
            "fee_cost": float(fee_cost),
            "slip_cost": float(slip_cost),
            "turnover_step": float(turnover_step),
            "delta_pos": float(delta_pos),
            "nav": float(self.nav),
            "pos": float(self.pos),
            "trade_cost": float(trade_cost),
            "pnl_from_prev_pos": float(pnl_from_prev_pos),
            "rel_alpha": float(rel_alpha),
            "mom": float(mom_term),
            "changed": bool(changed),
            "executed_trade": bool(executed_trade),
            "trade_count": int(self.trade_count),
            "drawdown": float(dd_now),
            "slo_triggered": bool(slo_triggered),
            "slo_locked": bool(self.enable_slo and (self._current_tick < self._slo_lock_until)),
            "slo_lock_until": int(self._slo_lock_until),
            "vol_t": float(vol_t),
            "vol_ref": float(vol_ref),
            "vol_gate": float(vol_gate),
            "vol_scale": float(scale),
            "vol_gated": bool(gated),
            "tp_pct": float(tp_pct),
            "sl_pct": float(sl_pct),
            "forced_close": bool(forced_close),
            "chop_now": bool(chop_now),
            "chop_locked": bool(chop_locked),
            "chop_lock_until": int(self._chop_lock_until),
        })
        return obs, reward, terminated, truncated, info


# -------------------------
# PPO policy stats helper
# -------------------------
def get_mu_sigma(model: PPO, obs):
    with torch.no_grad():
        obs_t, _ = model.policy.obs_to_tensor(obs)
        features = model.policy.extract_features(obs_t)
        latent_pi, _ = model.policy.mlp_extractor(features)
        mean_actions = model.policy.action_net(latent_pi)
        mu = float(mean_actions.detach().cpu().numpy().reshape(-1).mean())
        sigma_vec = model.policy.log_std.exp().detach().cpu().numpy().reshape(-1)
        sigma = float(sigma_vec.mean())
    return mu, sigma


# -------------------------
# Windows
# -------------------------
def get_walk_forward_windows(df_in: pd.DataFrame, window_size=3500, step_size=500, min_len=1200):
    return [
        (start, start + window_size)
        for start in range(0, len(df_in) - min_len, step_size)
        if start + window_size <= len(df_in)
    ]


def get_train_eval_split_idx(n_rows: int) -> int:
    split_idx = int(n_rows * WALKFORWARD_TRAIN_FRAC)
    split_idx = max(split_idx, WALKFORWARD_MIN_TRAIN_ROWS)
    split_idx = min(split_idx, n_rows - WALKFORWARD_MIN_EVAL_ROWS)
    return split_idx


# -------------------------
# Gate Helpers
# -------------------------
def build_trade_gate_label(df_in: pd.DataFrame, horizon: int, ret_thr: float) -> pd.DataFrame:
    df_local = df_in.copy()

    future_closes = pd.concat(
        [df_local["Close"].shift(-k) for k in range(1, horizon + 1)],
        axis=1
    )

    df_local["Gate_Future_Max_Close"] = future_closes.max(axis=1)
    df_local["Gate_Future_Max_Return"] = (
        df_local["Gate_Future_Max_Close"].sub(df_local["Close"]).div(df_local["Close"])
    )
    df_local["Trade_Label"] = (df_local["Gate_Future_Max_Return"] > ret_thr).astype(int)

    return df_local


def choose_gate_features(df_in: pd.DataFrame) -> list[str]:
    priority = [
        "RSI", "MACD_Line", "MACD_Signal", "ATR", "Volatility",
        "Vol_Forecast", "Vol_EWM", "Vol_Roll", "Vol_GARCH", "Vol_Ref", "Vol_Gate",
        "Ret_1", "Ret1", "Ret20", "FlipRate20", "Chop",
        "Denoised_Close", "SMA_20", "STD_20", "Upper_Band", "Lower_Band",
        "Stoch", "ROC", "OBV", "CCI", "EMA_10", "EMA_50",
        "VMD_Trend", "VMD_Cycle", "VMD_Noise",
        "VMD_Trend_Slope", "VMD_Cycle_Slope", "VMD_Noise_Ratio",
        "Regime4", "SentimentScore", "Delta", "Gamma",
    ]
    feats = [c for c in priority if c in df_in.columns and pd.api.types.is_numeric_dtype(df_in[c])]

    if not feats:
        exclude = {
            "Datetime", "Symbol", "Target", "Return",
            "Trade_Label", "Gate_Future_Max_Return", "Gate_Future_Max_Close",
        }
        feats = [
            c for c in df_in.columns
            if c not in exclude and pd.api.types.is_numeric_dtype(df_in[c])
        ]
        feats = sorted(feats)

    return feats


def train_trade_gate(
    df_gate: pd.DataFrame,
    gate_features: list[str],
    threshold: float,
) -> dict:
    work = df_gate.copy()
    work = work.dropna(subset=gate_features + ["Trade_Label"]).reset_index(drop=True)

    if len(work) < GATE_MIN_TRAIN_ROWS:
        raise ValueError(f"Gate training skipped: only {len(work)} rows after dropna.")

    split_idx = max(int(len(work) * GATE_TRAIN_FRACTION), 1)
    if split_idx >= len(work):
        split_idx = len(work) - 1

    train_part = work.iloc[:split_idx].copy()
    eval_part = work.iloc[split_idx:].copy()

    X_train = train_part[gate_features].astype(float)
    y_train = train_part["Trade_Label"].astype(int)
    X_eval = eval_part[gate_features].astype(float)
    y_eval = eval_part["Trade_Label"].astype(int)

    if y_train.nunique() < 2:
        raise ValueError("Gate training skipped: training labels have only one class.")

    gate_model = RandomForestClassifier(
        n_estimators=300,
        max_depth=6,
        min_samples_leaf=10,
        random_state=SEED,
        n_jobs=-1,
        class_weight="balanced_subsample",
    )
    gate_model.fit(X_train, y_train)

    eval_probs = gate_model.predict_proba(X_eval)[:, 1] if len(X_eval) else np.array([])
    eval_pred = (eval_probs > threshold).astype(int) if len(eval_probs) else np.array([])

    auc = np.nan
    precision = np.nan
    recall = np.nan
    if len(eval_probs) and len(np.unique(y_eval)) > 1:
        try:
            auc = float(roc_auc_score(y_eval, eval_probs))
        except Exception:
            pass
    if len(eval_probs):
        try:
            precision = float(precision_score(y_eval, eval_pred, zero_division=0))
        except Exception:
            pass
        try:
            recall = float(recall_score(y_eval, eval_pred, zero_division=0))
        except Exception:
            pass

    return {
        "model": gate_model,
        "features": gate_features,
        "threshold": float(threshold),
        "train_rows": int(len(train_part)),
        "eval_rows": int(len(eval_part)),
        "auc": auc,
        "precision": precision,
        "recall": recall,
        "train_pos_rate": float(y_train.mean()),
        "eval_pos_rate": float(y_eval.mean()) if len(y_eval) else np.nan,
    }


def gate_predict_row(gate_model, feature_row_df: pd.DataFrame) -> float:
    try:
        proba = gate_model.predict_proba(feature_row_df.astype(float))[0, 1]
        return float(proba)
    except Exception:
        return 0.0


# -------------------------
# QC Save
# -------------------------
def save_quantconnect_model(artifact: dict, prefix: str, save_dir: str):
    os.makedirs(save_dir, exist_ok=True)

    model_dst = os.path.join(save_dir, f"{prefix}_model.zip")
    model_obj = artifact.get("model", None)
    model_src = artifact.get("model_path", None)

    try:
        if model_obj is not None:
            model_obj.save(model_dst)
        else:
            if model_src and os.path.exists(model_src):
                if os.path.abspath(model_src) != os.path.abspath(model_dst):
                    shutil.copyfile(model_src, model_dst)
    except Exception as e:
        logging.warning(f"[QC SAVE] model issue for {prefix}: {e}")

    vecnorm_src = artifact.get("vecnorm_path")
    if vecnorm_src and os.path.exists(vecnorm_src):
        try:
            vecnorm_dst = os.path.join(save_dir, f"{prefix}_vecnorm.pkl")
            if os.path.abspath(vecnorm_src) != os.path.abspath(vecnorm_dst):
                shutil.copyfile(vecnorm_src, vecnorm_dst)
        except Exception as e:
            logging.warning(f"[QC SAVE] vecnorm issue for {prefix}: {e}")
    else:
        logging.warning(f"[QC SAVE] vecnorm missing for {prefix}")

    try:
        with open(os.path.join(save_dir, f"{prefix}_features.json"), "w") as f:
            json.dump({"features": artifact.get("features", [])}, f)
    except Exception as e:
        logging.warning(f"[QC SAVE] features.json issue for {prefix}: {e}")

    try:
        thr = float(artifact.get("result", {}).get("Action_Threshold", 0.2))
        with open(os.path.join(save_dir, f"{prefix}_probability_config.json"), "w") as f:
            json.dump(
                {
                    "threshold": thr,
                    "use_confidence": True,
                    "inference_mode": "deterministic",
                },
                f,
            )
    except Exception as e:
        logging.warning(f"[QC SAVE] prob_config issue for {prefix}: {e}")

    gate_model_src = artifact.get("gate_model_path")
    gate_features_src = artifact.get("gate_features_path")
    gate_config_src = artifact.get("gate_config_path")

    if gate_model_src and os.path.exists(gate_model_src):
        try:
            shutil.copyfile(gate_model_src, os.path.join(save_dir, f"{prefix}_gate_model.pkl"))
        except Exception as e:
            logging.warning(f"[QC SAVE] gate model issue for {prefix}: {e}")

    if gate_features_src and os.path.exists(gate_features_src):
        try:
            shutil.copyfile(gate_features_src, os.path.join(save_dir, f"{prefix}_gate_features.json"))
        except Exception as e:
            logging.warning(f"[QC SAVE] gate features issue for {prefix}: {e}")

    if gate_config_src and os.path.exists(gate_config_src):
        try:
            shutil.copyfile(gate_config_src, os.path.join(save_dir, f"{prefix}_gate_config.json"))
        except Exception as e:
            logging.warning(f"[QC SAVE] gate config issue for {prefix}: {e}")

    try:
        provided_model_info = artifact.get("model_info")
        if provided_model_info is not None:
            qc_model_info = provided_model_info
        else:
            r = artifact.get("result", {})
            qc_model_info = {
                "model": "PPO",
                "ticker": r.get("Ticker"),
                "window": r.get("Window"),
                "window_idx": r.get("WindowIdx"),
                "prefix": prefix,
                "date_trained_utc": now_utc().strftime("%Y-%m-%dT%H:%M:%SZ"),
                "framework": "stable-baselines3",
                "input_features": artifact.get("features", []),
                "final_portfolio": r.get("PPO_Portfolio"),
                "buy_hold": r.get("BuyHold"),
                "sharpe": r.get("Sharpe"),
                "drawdown_pct": r.get("Drawdown_%"),

                "trade_count": r.get("Executed_Trade_Count", r.get("Trade_Count")),
                "executed_trade_count": r.get("Executed_Trade_Count", r.get("Trade_Count")),
                "threshold_signal_count": r.get("Signal_Trade_Count"),
                "raw_signal_count": r.get("Raw_Signal_Count"),
                "signal_trade_count": r.get("Signal_Trade_Count"),  # back-compat alias

                "accuracy": r.get("Accuracy"),
                "action_threshold": r.get("Action_Threshold"),
                "gate_model": r.get("Gate_Model"),
                "gate_threshold": r.get("Gate_Threshold"),
                "gate_pass_count": r.get("Gate_Pass_Count"),
                "gate_block_count": r.get("Gate_Block_Count"),
                "gate_pass_rate": r.get("Gate_Pass_Rate"),
                "gate_auc_train": r.get("Gate_AUC_Train"),
                "gate_precision_train": r.get("Gate_Precision_Train"),
                "gate_recall_train": r.get("Gate_Recall_Train"),
                "gate_auc_eval": r.get("Gate_AUC_Eval"),
                "gate_precision_eval": r.get("Gate_Precision_Eval"),
                "gate_recall_eval": r.get("Gate_Recall_Eval"),
                "phase4_oos_only": artifact.get("phase4_oos_only", True),
                "risk_forecast_info": {
                    "enabled": artifact.get("risk_forecast_enabled"),
                    "selected_method": artifact.get("vol_method_selected"),
                    "selection_metric": artifact.get("vol_selection_metric"),
                    "scores": artifact.get("vol_scores"),
                    "score_coverage": artifact.get("vol_score_coverage"),
                    "baseline_method": artifact.get("vol_baseline_method"),
                    "baseline_score": artifact.get("vol_baseline_score"),
                    "best_score": artifact.get("vol_best_score"),
                    "improvement_vs_baseline": artifact.get("vol_improve_vs_baseline"),
                    "arch_available": artifact.get("vol_arch_available"),
                    "vol_ref": artifact.get("vol_ref"),
                    "vol_gate": artifact.get("vol_gate"),
                    "split_idx": artifact.get("vol_split_idx"),
                    "selection_end": artifact.get("vol_selection_end"),
                },
            }

        with open(os.path.join(save_dir, f"{prefix}_model_info.json"), "w") as f:
            json.dump(qc_model_info, f, indent=2, default=str)

    except Exception as e:
        logging.warning(f"[QC SAVE] model_info issue for {prefix}: {e}")

    logging.info(f"[QC SAVE] Saved QC artifacts for {prefix} -> {save_dir}")

# -------------------------
# Phase 4 helpers
# -------------------------
def sharpe_from_returns(ret_s: pd.Series) -> float:
    r = pd.Series(ret_s).replace([np.inf, -np.inf], np.nan).fillna(0.0)
    if len(r) == 0:
        return 0.0
    return float((r.mean() / (r.std() + 1e-9)) * _annualization_factor())


def max_drawdown_from_nav(nav_track: list[float]) -> float:
    if len(nav_track) == 0:
        return 0.0
    nav_s = pd.Series(nav_track, dtype=float)
    dd = ((nav_s.cummax() - nav_s) / nav_s.cummax().replace(0, np.nan)).fillna(0.0)
    return float(dd.max() * 100.0)


def block_bootstrap_indices(n: int, block_size: int) -> np.ndarray:
    if n <= 0:
        return np.array([], dtype=int)
    idx = []
    while len(idx) < n:
        start = np.random.randint(0, max(1, n))
        block = [(start + i) % n for i in range(block_size)]
        idx.extend(block)
    return np.array(idx[:n], dtype=int)


def bootstrap_ci_from_returns(
    ret_s: pd.Series,
    b: int = 500,
    alpha: float = 0.05,
    block_size: int = 24,
) -> dict:
    r = pd.Series(ret_s).replace([np.inf, -np.inf], np.nan).fillna(0.0).values.astype(float)
    n = len(r)

    if n < 10:
        total_ret = float(np.prod(1.0 + r) - 1.0) if n else 0.0
        shp = sharpe_from_returns(pd.Series(r))
        return {
            "n_obs": int(n),
            "bootstrap_b": int(b),
            "alpha": float(alpha),
            "block_size": int(block_size),
            "return_mean_boot": total_ret,
            "return_ci_low": total_ret,
            "return_ci_high": total_ret,
            "sharpe_mean_boot": shp,
            "sharpe_ci_low": shp,
            "sharpe_ci_high": shp,
        }

    total_returns = []
    sharpes = []
    for _ in range(int(b)):
        idx = block_bootstrap_indices(n, block_size)
        rs = r[idx]
        total_returns.append(float(np.prod(1.0 + rs) - 1.0))
        sharpes.append(float(sharpe_from_returns(pd.Series(rs))))

    lo = alpha / 2.0
    hi = 1.0 - (alpha / 2.0)

    return {
        "n_obs": int(n),
        "bootstrap_b": int(b),
        "alpha": float(alpha),
        "block_size": int(block_size),
        "return_mean_boot": float(np.mean(total_returns)),
        "return_ci_low": float(np.quantile(total_returns, lo)),
        "return_ci_high": float(np.quantile(total_returns, hi)),
        "sharpe_mean_boot": float(np.mean(sharpes)),
        "sharpe_ci_low": float(np.quantile(sharpes, lo)),
        "sharpe_ci_high": float(np.quantile(sharpes, hi)),
    }


def clone_vecnormalize_for_eval(
    df_eval_env: pd.DataFrame,
    src_vecnorm: VecNormalize,
    env_kwargs: dict,
):
    base = DummyVecEnv([lambda: ContinuousPositionEnv(df=df_eval_env, frame_bound=(50, len(df_eval_env) - 3), **env_kwargs)])
    vn = VecNormalize(base, norm_obs=True, norm_reward=True, clip_obs=10.0)

    try:
        if hasattr(src_vecnorm, "obs_rms"):
            vn.obs_rms = deepcopy(src_vecnorm.obs_rms)
        if hasattr(src_vecnorm, "ret_rms"):
            vn.ret_rms = deepcopy(src_vecnorm.ret_rms)
    except Exception as e:
        logging.warning(f"[VECNORM CLONE] stat copy issue: {e}")

    vn.training = False
    vn.norm_reward = False
    return vn


def get_vmd_columns(df_in: pd.DataFrame) -> list[str]:
    patterns = ("VMD_", "vmd_", "Mode_", "mode_", "Trend_Mode", "Cycle_Mode", "Noise_Mode")
    cols = [c for c in df_in.columns if any(p in str(c) for p in patterns)]
    return cols


def neutralize_vmd_columns(df_in: pd.DataFrame) -> tuple[pd.DataFrame, list[str]]:
    df_out = df_in.copy()
    vmd_cols = get_vmd_columns(df_out)
    for c in vmd_cols:
        try:
            df_out[c] = 0.0
        except Exception:
            pass
    return df_out, vmd_cols


def _safe_vmd_mode_sum(modes: np.ndarray, idxs: list[int]) -> np.ndarray:
    valid = [i for i in idxs if 0 <= i < modes.shape[0]]
    if not valid:
        return np.zeros(modes.shape[1], dtype=float)
    return np.sum(modes[valid, :], axis=0)


def compute_vmd_features_rolling(
    series: pd.Series,
    lookback: int = 128,
    K: int = 5,
    alpha: float = 2000,
    tau: float = 0.0,
    DC: int = 0,
    init: int = 1,
    tol: float = 1e-7,
) -> pd.DataFrame:
    """
    Leakage-safe VMD features:
    for each row t, decompose only the trailing window [t-lookback+1 : t].
    """
    s = pd.to_numeric(series, errors="coerce").astype(float).copy()
    n = len(s)

    out = pd.DataFrame(index=series.index)
    out["VMD_Trend"] = np.nan
    out["VMD_Cycle"] = np.nan
    out["VMD_Noise"] = np.nan

    if (not VMD_AVAILABLE) or n < lookback:
        return out

    for t in range(lookback - 1, n):
        window = s.iloc[t - lookback + 1 : t + 1].copy()
        if window.isna().any():
            window = window.ffill().bfill()
        if window.isna().any():
            continue

        x = window.values.astype(float)

        try:
            modes, _, _ = VMD(
                x,
                alpha=alpha,
                tau=tau,
                K=K,
                DC=DC,
                init=init,
                tol=tol,
            )

            trend_sig = _safe_vmd_mode_sum(modes, VMD_TREND_MODES)
            cycle_sig = _safe_vmd_mode_sum(modes, VMD_CYCLE_MODES)
            noise_sig = _safe_vmd_mode_sum(modes, VMD_NOISE_MODES)

            out.iloc[t, out.columns.get_loc("VMD_Trend")] = float(trend_sig[-1])
            out.iloc[t, out.columns.get_loc("VMD_Cycle")] = float(cycle_sig[-1])
            out.iloc[t, out.columns.get_loc("VMD_Noise")] = float(noise_sig[-1])

        except Exception:
            continue

    out["VMD_Trend"] = pd.to_numeric(out["VMD_Trend"], errors="coerce")
    out["VMD_Cycle"] = pd.to_numeric(out["VMD_Cycle"], errors="coerce")
    out["VMD_Noise"] = pd.to_numeric(out["VMD_Noise"], errors="coerce")

    out["VMD_Trend_Slope"] = out["VMD_Trend"].diff(3)
    out["VMD_Cycle_Slope"] = out["VMD_Cycle"].diff(3)
    out["VMD_Noise_Ratio"] = out["VMD_Noise"].abs() / (series.abs().replace(0, np.nan))

    out = out.replace([np.inf, -np.inf], np.nan)
    return out


def add_vmd_feature_block(df_in: pd.DataFrame) -> pd.DataFrame:
    df_out = df_in.copy()

    source_col = VMD_SOURCE_COL if VMD_SOURCE_COL in df_out.columns else "Close"
    if source_col not in df_out.columns:
        return df_out

    vmd_block = compute_vmd_features_rolling(
        series=df_out[source_col],
        lookback=VMD_LOOKBACK,
        K=VMD_K,
        alpha=VMD_ALPHA,
        tau=VMD_TAU,
        DC=VMD_DC,
        init=VMD_INIT,
        tol=VMD_TOL,
    )

    for c in vmd_block.columns:
        df_out[c] = vmd_block[c]

    return df_out


def build_regime_labels(step_df: pd.DataFrame, raw_eval_df: pd.DataFrame | None = None) -> pd.Series:
    if raw_eval_df is not None and "Regime4" in raw_eval_df.columns:
        reg = raw_eval_df["Regime4"].reset_index(drop=True)
        reg = reg.iloc[:len(step_df)].astype(str)
        return reg.rename("Regime")

    out = []
    vol_q75 = float(step_df["vol_t"].quantile(0.75)) if ("vol_t" in step_df.columns and len(step_df)) else np.nan

    for _, row in step_df.iterrows():
        if bool(row.get("chop_locked", False)):
            out.append("chop")
        elif pd.notna(vol_q75) and float(row.get("vol_t", 0.0)) >= vol_q75:
            out.append("high_vol")
        else:
            out.append("trend")
    return pd.Series(out, name="Regime")


def build_regime_breakdown(
    step_df: pd.DataFrame,
    ticker: str,
    prefix: str,
    window_label: str,
    raw_eval_df: pd.DataFrame | None = None,
) -> pd.DataFrame:
    if len(step_df) == 0:
        return pd.DataFrame()

    work = step_df.copy().reset_index(drop=True)
    work["Regime"] = build_regime_labels(work, raw_eval_df=raw_eval_df)

    rows = []
    for regime, g in work.groupby("Regime", dropna=False):
        ret_s = pd.to_numeric(g["ret_t"], errors="coerce").fillna(0.0) if "ret_t" in g.columns else pd.Series(dtype=float)
        nav_s = pd.to_numeric(g["nav"], errors="coerce").dropna() if "nav" in g.columns else pd.Series(dtype=float)
        rows.append({
            "Ticker": ticker,
            "Prefix": prefix,
            "Window": window_label,
            "Regime": regime,
            "Bars": int(len(g)),
            "Mean_Ret": float(ret_s.mean()) if len(ret_s) else np.nan,
            "Sharpe": sharpe_from_returns(ret_s) if len(ret_s) else np.nan,
            "Avg_Abs_Exposure": float(pd.to_numeric(g["pos"], errors="coerce").abs().mean()) if "pos" in g.columns else np.nan,
            "Trade_Steps": int(pd.Series(g.get("executed_trade", pd.Series(dtype=bool))).fillna(False).astype(bool).sum()) if "executed_trade" in g.columns else np.nan,
            "Forced_Close_Count": int(pd.to_numeric(g.get("forced_close", 0), errors="coerce").fillna(0).sum()) if "forced_close" in g.columns else 0,
            "Vol_Gated_Count": int(pd.to_numeric(g.get("vol_gated", 0), errors="coerce").fillna(0).sum()) if "vol_gated" in g.columns else 0,
            "Mean_Vol": float(pd.to_numeric(g["vol_t"], errors="coerce").mean()) if "vol_t" in g.columns else np.nan,
            "Nav_Last": float(nav_s.iloc[-1]) if len(nav_s) else np.nan,
        })
    return pd.DataFrame(rows)


def save_phase4_window_outputs(prefix: str, ablation_df: pd.DataFrame, regime_df: pd.DataFrame, bootstrap_ci: dict):
    if len(ablation_df):
        ablation_path = os.path.join(RUN_RESULTS_DIR, f"{prefix}_ablation.csv")
        ablation_df.to_csv(ablation_path, index=False)
    else:
        ablation_path = None

    if len(regime_df):
        regime_path = os.path.join(RUN_RESULTS_DIR, f"{prefix}_regime_breakdown.csv")
        regime_df.to_csv(regime_path, index=False)
    else:
        regime_path = None

    ci_path = os.path.join(RUN_RESULTS_DIR, f"{prefix}_bootstrap_ci.json")
    _safe_json_dump(bootstrap_ci, ci_path)

    manifest_record_paths(
        prefix,
        {
            "phase4_ablation_csv": ablation_path,
            "phase4_regime_csv": regime_path,
            "phase4_bootstrap_ci_json": ci_path,
        },
        do_hashes=ENABLE_HASHES,
    )


def apply_ablation_variant(df_eval_env: pd.DataFrame, variant: str) -> tuple[pd.DataFrame, dict, list[str]]:
    """
    Returns:
      modified_df_eval_env, env_kwargs_override, neutralized_cols
    """
    df_mod = df_eval_env.copy()
    env_over = {}
    neutralized_cols = []

    if variant == "full":
        return df_mod, env_over, neutralized_cols

    if variant == "no_vol_controls":
        env_over["enable_vol_controls"] = False
        return df_mod, env_over, neutralized_cols

    if variant == "no_slo":
        env_over["enable_slo"] = False
        env_over["slo_chop_enabled"] = False
        return df_mod, env_over, neutralized_cols

    if variant == "no_chop":
        env_over["slo_chop_enabled"] = False
        return df_mod, env_over, neutralized_cols

    if variant == "no_vmd":
        df_mod, neutralized_cols = neutralize_vmd_columns(df_mod)
        return df_mod, env_over, neutralized_cols

    return df_mod, env_over, neutralized_cols


def summarize_eval_metrics(
    nav_track: list[float],
    bh_track: list[float],
    step_df: pd.DataFrame,
    trades: list[dict],
    thr_window: float,
    gate_pass_count: int,
    gate_block_count: int,
    gate_auc_eval: float,
    gate_precision_eval: float,
    gate_recall_eval: float,
) -> dict:
    avg_abs_exposure = float(step_df["pos"].abs().mean()) if len(step_df) and "pos" in step_df.columns else 0.0
    turnover = float(step_df["turnover_step"].sum()) if len(step_df) and "turnover_step" in step_df.columns else 0.0
    avg_bars_held = float(
        pd.to_numeric(pd.DataFrame(trades).get("BarsHeld", pd.Series(dtype=float)), errors="coerce").dropna().mean()
    ) if len(trades) else 0.0

    vol_gated_count = int(step_df["vol_gated"].sum()) if "vol_gated" in step_df.columns else 0
    forced_close_count = int(step_df["forced_close"].sum()) if "forced_close" in step_df.columns else 0
    slo_trigger_count = int(step_df["slo_triggered"].sum()) if "slo_triggered" in step_df.columns else 0
    chop_locked_count = int(step_df["chop_locked"].sum()) if "chop_locked" in step_df.columns else 0

    correct = 0
    total = 0
    tp_buy = fp_buy = tp_sell = fp_sell = 0
    for _, r in step_df.iterrows():
        a = float(r["Action"])
        ret_t = float(r.get("next_ret", 0.0))
        if a > thr_window:
            total += 1
            if ret_t > 0:
                tp_buy += 1
                correct += 1
            else:
                fp_buy += 1
        elif a < -thr_window:
            total += 1
            if ret_t < 0:
                tp_sell += 1
                correct += 1
            else:
                fp_sell += 1

    precision_long = tp_buy / (tp_buy + fp_buy + 1e-9)
    precision_short = tp_sell / (tp_sell + fp_sell + 1e-9)
    precision_trades = (tp_buy + tp_sell) / ((tp_buy + tp_sell) + (fp_buy + fp_sell) + 1e-9)
    step_accuracy = round(correct / total, 4) if total > 0 else 0.0

    final_value = float(nav_track[-1]) * 100_000.0 if len(nav_track) else 100_000.0
    hold_value = float(bh_track[-1]) * 100_000.0 if len(bh_track) else 100_000.0
    returns = pd.Series(nav_track).pct_change().fillna(0.0) if len(nav_track) else pd.Series(dtype=float)
    sharpe = sharpe_from_returns(returns)
    drawdown = max_drawdown_from_nav(nav_track)
    gate_pass_rate = float(gate_pass_count / max(gate_pass_count + gate_block_count, 1)) if (gate_pass_count + gate_block_count) > 0 else np.nan

    return {
        "final_value": round(final_value, 2),
        "hold_value": round(hold_value, 2),
        "returns": returns,
        "sharpe": float(sharpe),
        "drawdown": float(drawdown),
        "accuracy": float(step_accuracy),
        "precision_long": float(precision_long),
        "precision_short": float(precision_short),
        "precision_trades": float(precision_trades),
        "avg_abs_exposure": round(avg_abs_exposure, 6),
        "turnover": round(turnover, 6),
        "avg_bars_held": round(avg_bars_held, 3),
        "vol_gated_count": int(vol_gated_count),
        "forced_close_count": int(forced_close_count),
        "slo_trigger_count": int(slo_trigger_count),
        "chop_locked_count": int(chop_locked_count),
        "gate_pass_count": int(gate_pass_count),
        "gate_block_count": int(gate_block_count),
        "gate_pass_rate": gate_pass_rate,
        "gate_auc_eval": gate_auc_eval,
        "gate_precision_eval": gate_precision_eval,
        "gate_recall_eval": gate_recall_eval,
    }


def evaluate_trained_policy_oos(
    model: PPO,
    src_vecnorm: VecNormalize,
    df_eval_env: pd.DataFrame,
    df_align_eval: pd.DataFrame,
    df_gate_live_eval: pd.DataFrame,
    df_window_eval_full: pd.DataFrame,
    gate_model,
    prefix: str,
    ticker: str,
    window_label: str,
    gate_enabled: bool,
    env_kwargs_override: dict | None = None,
    variant_name: str = "full",
):
    env_kwargs_local = dict(ENV_KWARGS)
    if env_kwargs_override:
        env_kwargs_local.update(env_kwargs_override)

    eval_env = clone_vecnormalize_for_eval(df_eval_env=df_eval_env, src_vecnorm=src_vecnorm, env_kwargs=env_kwargs_local)
    obs = eval_env.reset()

    nav_track = [1.0]
    bh_track = [1.0]
    step_log = []

    trades = []
    open_trade = None
    prev_pos = 0.0
    eps = 1e-9
    START_EQUITY = 100_000.0

    executed_trade_count = 0
    raw_signal_count = 0
    threshold_signal_count = 0
    DIAG_THR = 0.2
    last_eval_i = 0

    gate_pass_count = 0
    gate_block_count = 0
    gate_probs_eval = []
    gate_true_eval = []

    def _safe_float(x, default=0.0):
        try:
            if x is None:
                return float(default)
            return float(x)
        except Exception:
            return float(default)

    def close_trade(reason: str, exit_idx: int, exit_dt, exit_px: float, exit_nav: float):
        nonlocal open_trade, trades
        if open_trade is None:
            return

        entry_px = float(open_trade["EntryPrice"])
        entry_pos = float(open_trade["EntryPos"])
        side = open_trade["Side"]
        dir_sign = 1.0 if side == "LONG" else -1.0

        gross_ret = dir_sign * abs(entry_pos) * ((exit_px / max(entry_px, 1e-9)) - 1.0)
        costs = float(open_trade["Costs"])
        net_ret = gross_ret - costs
        pnl_usd = (float(exit_nav) - float(open_trade["EntryNAV"])) * START_EQUITY

        open_trade.update({
            "Variant": variant_name,
            "ExitIndex": int(exit_idx),
            "ExitDatetime": exit_dt,
            "ExitPrice": float(exit_px),
            "ExitNAV": float(exit_nav),
            "BarsHeld": int(exit_idx - int(open_trade["EntryIndex"])),
            "GrossReturnPct": float(gross_ret * 100.0),
            "CostsPct": float(costs * 100.0),
            "NetReturnPct": float(net_ret * 100.0),
            "PnL_Usd": float(pnl_usd),
            "ExitReason": str(reason),
            "MFE_Pct": float(open_trade.get("MFE", 0.0) * 100.0),
            "MAE_Pct": float(open_trade.get("MAE", 0.0) * 100.0),
        })
        trades.append(open_trade)
        open_trade = None

    if len(df_align_eval) < 2:
        raise ValueError(f"{prefix}: df_align_eval too short ({len(df_align_eval)})")

    for i in range(len(df_align_eval) - 1):
        raw_action, _ = model.predict(obs, deterministic=True)
        mu, sigma = get_mu_sigma(model, obs)

        action = raw_action
        gate_prob = np.nan
        gate_passed = True

        if gate_enabled and gate_model is not None and (i + 1) < len(df_gate_live_eval):
            feature_row = df_gate_live_eval.iloc[[i + 1]].copy()

            if not feature_row.isna().any(axis=None):
                gate_prob = gate_predict_row(gate_model, feature_row)

                if gate_prob <= GATE_THRESHOLD:
                    gate_passed = False
                    action = np.array([0.0], dtype=np.float32)
                    gate_block_count += 1
                else:
                    gate_passed = True
                    gate_pass_count += 1
                    raw_a = float(np.array(raw_action).squeeze())
                    gate_strength = (gate_prob - GATE_THRESHOLD) / max(1.0 - GATE_THRESHOLD, 1e-9)
                    gate_strength = float(np.clip(gate_strength, 0.0, 1.0))
                    sized_action = raw_a * gate_strength
                    action = np.array([sized_action], dtype=np.float32)
            else:
                gate_prob = np.nan
                gate_passed = False
                action = np.array([0.0], dtype=np.float32)
                gate_block_count += 1

            gate_true = (
                int(df_window_eval_full["Trade_Label"].iloc[i + 1])
                if "Trade_Label" in df_window_eval_full.columns and (i + 1) < len(df_window_eval_full)
                else np.nan
            )
            if pd.notna(gate_prob) and pd.notna(gate_true):
                gate_probs_eval.append(float(gate_prob))
                gate_true_eval.append(int(gate_true))

        obs, rewards, dones, infos = eval_env.step(action)
        rew0 = float(rewards[0]) if isinstance(rewards, np.ndarray) else float(rewards)
        done0 = bool(dones[0]) if isinstance(dones, np.ndarray) else bool(dones)
        info0 = infos[0] if isinstance(infos, (list, tuple)) and len(infos) else (infos or {})
        info0 = info0 or {}

        nav_now = _safe_float(info0.get("nav", nav_track[-1]), nav_track[-1])
        base_ret = _safe_float(info0.get("base_ret", 0.0), 0.0)
        pos_now = _safe_float(info0.get("pos", 0.0), 0.0)
        trade_cost_step = _safe_float(info0.get("trade_cost", 0.0), 0.0)
        turnover_step = _safe_float(
            info0.get("turnover_step", abs(_safe_float(info0.get("delta_pos", 0.0), 0.0))),
            0.0,
        )
        delta_pos = _safe_float(info0.get("delta_pos", 0.0), 0.0)

        nav_track.append(nav_now)
        bh_track.append(bh_track[-1] * (1.0 + base_ret))

        a = float(np.array(action).squeeze())
        raw_a = float(np.array(raw_action).squeeze())

        dt_val = df_align_eval["Datetime"].iloc[i + 1] if "Datetime" in df_align_eval.columns else pd.NaT
        px = float(df_align_eval["Close"].iloc[i + 1])

        if open_trade is not None:
            open_trade["Costs"] += trade_cost_step

            entry_px = float(open_trade["EntryPrice"])
            entry_pos = float(open_trade["EntryPos"])
            side = str(open_trade["Side"])
            dir_sign = 1.0 if side == "LONG" else -1.0

            gross_ret_now = dir_sign * abs(entry_pos) * ((px / max(entry_px, 1e-9)) - 1.0)
            open_trade["MFE"] = max(float(open_trade.get("MFE", gross_ret_now)), gross_ret_now)
            open_trade["MAE"] = min(float(open_trade.get("MAE", gross_ret_now)), gross_ret_now)

        was_flat = abs(prev_pos) < eps
        now_flat = abs(pos_now) < eps
        sign_flip = (abs(prev_pos) >= eps) and (abs(pos_now) >= eps) and (np.sign(prev_pos) != np.sign(pos_now))

        if (not was_flat) and now_flat:
            close_trade(
                reason=("TP/SL" if bool(info0.get("forced_close", False)) else "SLO" if bool(info0.get("slo_triggered", False)) else "FLAT"),
                exit_idx=i + 1,
                exit_dt=dt_val,
                exit_px=px,
                exit_nav=nav_now,
            )
        elif sign_flip:
            close_trade(
                reason="SIGN_FLIP",
                exit_idx=i + 1,
                exit_dt=dt_val,
                exit_px=px,
                exit_nav=nav_now,
            )
            side = "LONG" if pos_now > 0 else "SHORT"
            open_trade = {
                "Ticker": ticker,
                "Window": window_label,
                "Prefix": prefix,
                "Variant": variant_name,
                "Side": side,
                "EntryIndex": int(i + 1),
                "EntryDatetime": dt_val,
                "EntryPrice": float(px),
                "EntryPos": float(pos_now),
                "EntryNAV": float(nav_now),
                "Costs": 0.0,
                "MFE": 0.0,
                "MAE": 0.0,
            }
        elif was_flat and (not now_flat):
            side = "LONG" if pos_now > 0 else "SHORT"
            open_trade = {
                "Ticker": ticker,
                "Window": window_label,
                "Prefix": prefix,
                "Variant": variant_name,
                "Side": side,
                "EntryIndex": int(i + 1),
                "EntryDatetime": dt_val,
                "EntryPrice": float(px),
                "EntryPos": float(pos_now),
                "EntryNAV": float(nav_now),
                "Costs": 0.0,
                "MFE": 0.0,
                "MAE": 0.0,
            }

        prev_pos = pos_now

        if abs(raw_a) > DIAG_THR:
            raw_signal_count += 1
        if bool(info0.get("executed_trade", False)):
            executed_trade_count += 1

        if i + 2 < len(df_align_eval):
            p0 = float(df_align_eval["Close"].iloc[i + 1])
            p1 = float(df_align_eval["Close"].iloc[i + 2])
            next_ret = 0.0 if p0 <= 0 else (p1 - p0) / p0
        else:
            next_ret = 0.0

        step_log.append({
            "Variant": variant_name,
            "Index": i + 1,
            "Datetime": dt_val,
            "Close": px,
            "Raw_Action": raw_a,
            "Action": a,
            "mu": mu,
            "sigma": sigma,
            "nav": nav_now,
            "ret_t": float(info0.get("ret_t", 0.0)),
            "next_ret": float(next_ret),
            "reward": float(rew0),
            "pos": float(pos_now),
            "trade_cost": float(trade_cost_step),
            "turnover_step": float(turnover_step),
            "delta_pos": float(delta_pos),
            "base_ret": float(base_ret),
            "rel_alpha": float(info0.get("rel_alpha", 0.0)),
            "mom": float(info0.get("mom", 0.0)),
            "vol_t": float(info0.get("vol_t", 0.0)),
            "vol_ref": float(info0.get("vol_ref", 0.0)),
            "vol_gate": float(info0.get("vol_gate", 0.0)),
            "vol_scale": float(info0.get("vol_scale", 1.0)),
            "vol_gated": bool(info0.get("vol_gated", False)),
            "tp_pct": float(info0.get("tp_pct", 0.0)),
            "sl_pct": float(info0.get("sl_pct", 0.0)),
            "forced_close": bool(info0.get("forced_close", False)),
            "slo_triggered": bool(info0.get("slo_triggered", False)),
            "slo_locked": bool(info0.get("slo_locked", False)),
            "slo_lock_until": int(info0.get("slo_lock_until", -1)),
            "drawdown": float(info0.get("drawdown", 0.0)),
            "chop_locked": bool(info0.get("chop_locked", False)),
            "executed_trade": bool(info0.get("executed_trade", False)),
            "Gate_Prob": gate_prob,
            "Gate_Passed": gate_passed,
            "Gate_Blocked": (not gate_passed) if gate_enabled else False,
            "Raw_Signal_Flag": bool(abs(raw_a) > DIAG_THR),
        })

        last_eval_i = i + 1
        if done0:
            break

    if open_trade is not None:
        last_idx = int(min(last_eval_i, len(df_align_eval) - 1))
        last_dt = df_align_eval["Datetime"].iloc[last_idx] if "Datetime" in df_align_eval.columns else pd.NaT
        last_px = float(df_align_eval["Close"].iloc[last_idx])
        last_nav = float(nav_track[-1])
        close_trade(reason="EOD", exit_idx=last_idx, exit_dt=last_dt, exit_px=last_px, exit_nav=last_nav)

    step_df = pd.DataFrame(step_log)

    abs_actions = np.array([abs(float(r["Action"])) for r in step_log], dtype=float) if len(step_log) else np.array([])
    if len(abs_actions) > 0:
        thr_window = float(np.quantile(abs_actions, 0.60))
        thr_window = float(np.clip(thr_window, 0.08, 0.25))
    else:
        thr_window = 0.2

    if len(step_df):
        step_df["Threshold_Signal_Flag"] = step_df["Action"].abs() > thr_window
        step_df["Executed_Trade_Flag"] = step_df["executed_trade"].fillna(False).astype(bool)

        raw_threshold_signal_count = int(step_df["Threshold_Signal_Flag"].sum())
        executed_trade_count_from_df = int(step_df["Executed_Trade_Flag"].sum())

        threshold_signal_count = raw_threshold_signal_count
    else:
        raw_threshold_signal_count = 0
        executed_trade_count_from_df = 0
        threshold_signal_count = 0

    gate_auc_eval = np.nan
    gate_precision_eval = np.nan
    gate_recall_eval = np.nan

    if gate_enabled and len(gate_probs_eval) > 0:
        gp = np.array(gate_probs_eval, dtype=float)
        gy = np.array(gate_true_eval, dtype=int)
        gpred = (gp > GATE_THRESHOLD).astype(int)
        if len(np.unique(gy)) > 1:
            try:
                gate_auc_eval = float(roc_auc_score(gy, gp))
            except Exception:
                pass
        try:
            gate_precision_eval = float(precision_score(gy, gpred, zero_division=0))
        except Exception:
            pass
        try:
            gate_recall_eval = float(recall_score(gy, gpred, zero_division=0))
        except Exception:
            pass

    metrics = summarize_eval_metrics(
        nav_track=nav_track,
        bh_track=bh_track,
        step_df=step_df,
        trades=trades,
        thr_window=thr_window,
        gate_pass_count=gate_pass_count,
        gate_block_count=gate_block_count,
        gate_auc_eval=gate_auc_eval,
        gate_precision_eval=gate_precision_eval,
        gate_recall_eval=gate_recall_eval,
    )

    metrics["executed_trade_count"] = int(executed_trade_count_from_df)
    metrics["raw_signal_count"] = int(raw_signal_count)
    metrics["threshold_signal_count"] = int(raw_threshold_signal_count)
    metrics["signal_trade_count_fixed"] = int(executed_trade_count_from_df)
    metrics["thr_window"] = float(thr_window)

    try:
        eval_env.close()
    except Exception:
        pass

    return {
        "step_df": step_df,
        "trades": trades,
        "nav_track": nav_track,
        "bh_track": bh_track,
        "metrics": metrics,
        "gate_auc_eval": gate_auc_eval,
        "gate_precision_eval": gate_precision_eval,
        "gate_recall_eval": gate_recall_eval,
    }


# -------------------------
# Parameter profiles
# -------------------------
TOP_N_WINDOWS = 3

FAST = {"lr": 8e-5, "n_steps": 3072, "batch": 512, "clip": 0.2, "ent": 0.01}
SLOW = {"lr": 3e-5, "n_steps": 3072, "batch": 512, "clip": 0.16, "ent": 0.005}

fast_names = {
    "TSLA", "NVDA", "AMD", "AVGO", "AAPL", "MSFT", "AMZN", "GOOGL", "META", "ADBE", "CRM",
    "INTC", "QCOM", "TXN", "ORCL", "NEE", "GE", "XOM", "CVX", "LLY", "NKE", "SBUX"
}
slow_names = {
    "BRK-B", "JPM", "BAC", "JNJ", "UNH", "MRK", "PFE", "ABBV", "ABT", "AMGN", "PG", "PEP", "KO",
    "V", "MA", "WMT", "MCD", "TMO", "DHR", "ACN", "IBM", "LIN", "PM", "RTX", "UPS", "UNP", "COST", "HD", "LOW"
}


def pick_params(symbol: str) -> dict:
    return FAST if symbol in fast_names else SLOW


# -------------------------
# QC_TOP export from existing runs
# -------------------------
def _safe_read_json(path: str, default=None):
    if default is None:
        default = {}
    try:
        if path and os.path.exists(path):
            with open(path, "r") as f:
                return json.load(f)
    except Exception as e:
        logging.warning(f"[JSON READ] Could not read {path}: {e}")
    return default

def export_qc_top_from_existing(ticker: str, top_n: int = 3):
    run_dirs = sorted(glob.glob(os.path.join(BASE_RESULTS_DIR, "ppo_walkforward_results_*")))[::-1]
    summary_files = []

    for rd in run_dirs:
        cand = glob.glob(os.path.join(rd, "summary*.csv"))
        if not cand:
            continue

        ok = False
        for p in cand:
            try:
                tmp = pd.read_csv(p, usecols=["Ticker"])
                if (tmp["Ticker"] == ticker).any():
                    ok = True
                    break
            except Exception:
                pass

        if ok:
            summary_files = cand
            break

    if not summary_files:
        logging.warning(f"[QC_TOP] No summaries containing {ticker} found.")
        return

    frames = []
    for p in summary_files:
        try:
            frames.append(pd.read_csv(p))
        except Exception as e:
            logging.warning(f"[QC_TOP] Could not read {p}: {e}")

    if not frames:
        logging.warning(f"[QC_TOP] No readable summaries for {ticker}.")
        return

    combo = pd.concat(frames, ignore_index=True)
    if "Ticker" not in combo.columns or "Sharpe" not in combo.columns:
        logging.warning("[QC_TOP] Missing required columns in summaries.")
        return

    combo = combo[combo["Ticker"] == ticker].copy()
    combo["Sharpe"] = pd.to_numeric(combo["Sharpe"], errors="coerce")
    combo = combo.dropna(subset=["Sharpe"])
    if combo.empty:
        logging.warning(f"[QC_TOP] No numeric Sharpe rows for {ticker}.")
        return

    use_prefix = ("Prefix" in combo.columns) and combo["Prefix"].notna().any()
    if use_prefix:
        top = combo.sort_values("Sharpe", ascending=False).head(top_n).copy()
        top["__prefix__"] = top["Prefix"].astype(str)
    else:
        def _window_start(w):
            try:
                s = str(w)
                return int(s.split("-")[0]) if "-" in s else np.nan
            except Exception:
                return np.nan

        combo["WindowStart"] = combo["Window"].apply(_window_start)
        combo = combo.sort_values(["WindowStart"]).reset_index(drop=True)
        combo["WindowIdx"] = combo.groupby("Ticker").cumcount() + 1
        top = combo.sort_values("Sharpe", ascending=False).head(top_n).copy()
        top["__prefix__"] = top["WindowIdx"].apply(lambda widx: f"ppo_{ticker}_window{int(widx)}")

    exported = 0
    for _, r in top.iterrows():
        prefix = str(r["__prefix__"])
        model_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_model.zip")
        vec_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_vecnorm.pkl")
        gate_model_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_gate_model.pkl")
        gate_features_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_gate_features.json")
        gate_config_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_gate_config.json")

        features_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_features.json")
        model_info_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_model_info.json")

        features_blob = _safe_read_json(features_path, default={})
        model_info_blob = _safe_read_json(model_info_path, default={})

        master_features = features_blob.get("features", [])
        risk_info = model_info_blob.get("risk_forecast_info", {})

        if not (os.path.exists(model_path) and os.path.exists(vec_path)):
            logging.warning(f"[QC_TOP] Missing model/vecnorm for {prefix}; skip export.")
            continue

        artifact_for_save = {
            "model": None,
            "model_path": model_path,
            "vecnorm_path": vec_path,
            "features": master_features,
            "result": r.to_dict(),
            "prefix": prefix,
            "gate_model_path": gate_model_path if (gate_model_path and os.path.exists(gate_model_path)) else None,
            "gate_features_path": gate_features_path if (gate_features_path and os.path.exists(gate_features_path)) else None,
            "gate_config_path": gate_config_path if (gate_config_path and os.path.exists(gate_config_path)) else None,

            "phase4_oos_only": model_info_blob.get("phase4_oos_only", True),
            "risk_forecast_enabled": risk_info.get("enabled"),
            "vol_method_selected": risk_info.get("selected_method"),
            "vol_selection_metric": risk_info.get("selection_metric"),
            "vol_scores": risk_info.get("scores"),
            "vol_score_coverage": risk_info.get("score_coverage"),
            "vol_baseline_method": risk_info.get("baseline_method"),
            "vol_baseline_score": risk_info.get("baseline_score"),
            "vol_best_score": risk_info.get("best_score"),
            "vol_improve_vs_baseline": risk_info.get("improvement_vs_baseline"),
            "vol_arch_available": risk_info.get("arch_available"),
            "vol_ref": risk_info.get("vol_ref"),
            "vol_gate": risk_info.get("vol_gate"),
            "vol_split_idx": risk_info.get("split_idx"),
            "vol_selection_end": risk_info.get("selection_end"),
            "model_info": model_info_blob if model_info_blob else None,
        }
        save_quantconnect_model(artifact_for_save, prefix, QC_TOP_DIR)
        exported += 1

    logging.info(f"[QC_TOP] Exported {exported}/{len(top)} for {ticker}")


def make_env_df(df_window: pd.DataFrame) -> pd.DataFrame:
    if "Close" not in df_window.columns:
        raise ValueError("make_env_df: missing Close")

    numeric_cols = [c for c in df_window.columns if pd.api.types.is_numeric_dtype(df_window[c])]
    cols = ["Close"] + [c for c in numeric_cols if c != "Close"]

    out = df_window[cols].copy()
    for c in out.columns:
        out[c] = pd.to_numeric(out[c], errors="coerce")

    out = out.replace([np.inf, -np.inf], np.nan)

    logging.info(f"[make_env_df] rows before selective cleanup: {len(out)}")

    na_counts = out.isna().sum()
    na_counts = na_counts[na_counts > 0].sort_values(ascending=False)
    if len(na_counts):
        logging.info("[make_env_df] numeric column NaN counts:")
        logging.info("\n" + na_counts.to_string())

    # Only require Close to exist
    out = out.dropna(subset=["Close"]).reset_index(drop=True)

    # Fill feature NaNs instead of dropping rows
    feature_cols = [c for c in out.columns if c != "Close"]
    if feature_cols:
        out[feature_cols] = out[feature_cols].ffill().bfill().fillna(0.0)

    logging.info(f"[make_env_df] rows after selective cleanup: {len(out)}")

    if len(out) < 60:
        raise ValueError(f"make_env_df: too few rows after cleanup ({len(out)})")

    return out

def build_and_train_ppo(train_df, frame_bound_train, ppo_overrides, timesteps):
    env_local = DummyVecEnv([
        lambda: ContinuousPositionEnv(
            df=train_df,
            frame_bound=frame_bound_train,
            **ENV_KWARGS
        )
    ])
    env_local = VecNormalize(env_local, norm_obs=True, norm_reward=True, clip_obs=10.0)

    model_local = PPO(
        "MlpPolicy",
        env_local,
        verbose=0,
        device=DEVICE_STR,
        learning_rate=float(ppo_overrides.get("lr", 1e-4)),
        n_steps=int(ppo_overrides.get("n_steps", 256)),
        batch_size=int(ppo_overrides.get("batch", 64)),
        n_epochs=5,
        gamma=0.99,
        gae_lambda=0.95,
        clip_range=float(ppo_overrides.get("clip", 0.2)),
        ent_coef=float(ppo_overrides.get("ent", 0.005)),
        policy_kwargs=dict(net_arch=[64, 64]),
    )

    model_local.learn(total_timesteps=int(timesteps))
    return model_local, env_local
# -------------------------
# Walkforward PPO + Gate + Phase 4 OOS Eval
# -------------------------
WINDOW_SIZE = 3500
STEP_SIZE = 500
TIMESTEPS = 150_000

if TEST_MODE:
    TIMESTEPS = 20_000
    WINDOW_SIZE = 2000
    STEP_SIZE = 750

def walkforward_ppo(
    df_sym: pd.DataFrame,
    ticker: str,
    window_size: int,
    step_size: int,
    timesteps: int,
    ppo_overrides: dict,
):
    if len(df_sym) < window_size:
        logging.warning(f"Skipping {ticker}: only {len(df_sym)} rows (<{window_size}).")
        return []

    df_sym = df_sym.sort_values("Datetime").reset_index(drop=True)

    windows = get_walk_forward_windows(df_sym, window_size=window_size, step_size=step_size)
    if TEST_MODE:
        windows = windows[:3]

    results = []
    top_heap = []
    skipped_windows = []

    all_done = True
    for idx in range(len(windows)):
        prefix = f"ppo_{ticker}_window{idx + 1}"
        model_ok = os.path.exists(os.path.join(FINAL_MODEL_DIR, f"{prefix}_model.zip"))
        vec_ok = os.path.exists(os.path.join(FINAL_MODEL_DIR, f"{prefix}_vecnorm.pkl"))
        feat_ok = os.path.exists(os.path.join(FINAL_MODEL_DIR, f"{prefix}_features.json"))
        prob_ok = os.path.exists(os.path.join(FINAL_MODEL_DIR, f"{prefix}_probability_config.json"))
        info_ok = os.path.exists(os.path.join(FINAL_MODEL_DIR, f"{prefix}_model_info.json"))

        if ENABLE_TRADE_GATE:
            gate_ok = (
                os.path.exists(os.path.join(FINAL_MODEL_DIR, f"{prefix}_gate_model.pkl")) and
                os.path.exists(os.path.join(FINAL_MODEL_DIR, f"{prefix}_gate_features.json")) and
                os.path.exists(os.path.join(FINAL_MODEL_DIR, f"{prefix}_gate_config.json"))
            )
        else:
            gate_ok = True

        if not (model_ok and vec_ok and feat_ok and prob_ok and info_ok and gate_ok):
            all_done = False
            break

    if all_done and len(windows) > 0 and (not FORCE_RETRAIN):
        logging.info(f"{ticker} fully skipped (all {len(windows)} windows already complete).")
        record_skips_global(ticker, [], total_windows=len(windows), fully_skipped=True)
        export_qc_top_from_existing(ticker, top_n=TOP_N_WINDOWS)
        return []

    for w_idx, (start, end) in enumerate(windows):
        try:
            window_start_time = time.time()
            gc.collect()
            try:
                torch.cuda.empty_cache()
            except Exception:
                pass

            prefix = f"ppo_{ticker}_window{w_idx + 1}"
            model_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_model.zip")
            vecnorm_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_vecnorm.pkl")

            feat_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_features.json")
            prob_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_probability_config.json")
            info_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_model_info.json")

            gate_model_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_gate_model.pkl")
            gate_features_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_gate_features.json")
            gate_config_path = os.path.join(FINAL_MODEL_DIR, f"{prefix}_gate_config.json")

            core_ok = os.path.exists(model_path) and os.path.exists(vecnorm_path)
            master_ok = os.path.exists(feat_path) and os.path.exists(prob_path) and os.path.exists(info_path)

            if ENABLE_TRADE_GATE:
                gate_ok = (
                    os.path.exists(gate_model_path) and
                    os.path.exists(gate_features_path) and
                    os.path.exists(gate_config_path)
                )
            else:
                gate_ok = True

            if core_ok and master_ok and gate_ok and (not FORCE_RETRAIN):
                logging.info(f"Skipping {ticker} window {w_idx + 1}: already trained + required artifacts complete.")
                skipped_windows.append(f"{ticker}_window{w_idx + 1}")
                manifest_update(prefix, {"status": "skipped_complete", "ticker": ticker, "window_idx": int(w_idx + 1)})
                continue

            if FORCE_RETRAIN and (
                core_ok or master_ok or
                (ENABLE_TRADE_GATE and (
                    os.path.exists(gate_model_path) or
                    os.path.exists(gate_features_path) or
                    os.path.exists(gate_config_path)
                ))
            ):
                logging.info(f"[FORCE_RETRAIN] Rebuilding {prefix} despite existing artifacts.")

            if core_ok and (not master_ok or not gate_ok) and (not FORCE_RETRAIN):
                logging.warning(f"[MASTER/OPTIONAL_GATE] {prefix} core exists but required deploy artifacts incomplete; will rebuild after eval.")

            df_window = df_sym.iloc[start:end].reset_index(drop=True)

            prelim_split_idx = get_train_eval_split_idx(len(df_window))
            if prelim_split_idx <= 60 or (len(df_window) - prelim_split_idx) < WALKFORWARD_MIN_EVAL_ROWS:
                logging.warning(f"{prefix}: skipping because train/eval split invalid. len={len(df_window)} split={prelim_split_idx}")
                continue

            if ENABLE_RISK_FORECAST_UPGRADE:
                df_window, vol_diag = build_vol_forecast_candidates(
                    df_window=df_window,
                    lookback=VOL_LOOKBACK,
                    ewm_span=VOL_EWM_SPAN,
                    train_frac=VOL_FORECAST_TRAIN_FRAC,
                    oos_start_idx=prelim_split_idx,
                )

                if VOL_FORECAST_LOG_EVAL:
                    logging.info(
                        f"[VOL FORECAST] {prefix} | selected={vol_diag['selected_method']} | "
                        f"metric={vol_diag['selection_metric']} | "
                        f"scores={vol_diag['scores']} | "
                        f"coverage={vol_diag.get('score_coverage')} | "
                        f"split_idx={vol_diag.get('split_idx')} | "
                        f"selection_end={vol_diag.get('selection_end')} | "
                        f"improvement_vs_{vol_diag['baseline_method']}={vol_diag['improvement_vs_baseline']}"
                    )
            else:
                df_window["Ret_1"] = df_window["Close"].pct_change().fillna(0.0)
                df_window["Vol_Roll"] = df_window["Ret_1"].rolling(VOL_LOOKBACK).std().fillna(0.0)
                df_window["Vol_EWM"] = df_window["Ret_1"].ewm(span=VOL_EWM_SPAN, adjust=False).std().fillna(0.0)
                df_window["Vol_GARCH"] = np.nan
                df_window["Vol_Forecast"] = df_window["Vol_EWM"]
                vol_ref = float(df_window["Vol_Forecast"].median()) if len(df_window) else 0.0
                vol_gate = float(df_window["Vol_Forecast"].quantile(VOL_GATE_Q)) if len(df_window) else 0.0
                df_window["Vol_Ref"] = vol_ref
                df_window["Vol_Gate"] = vol_gate

                vol_diag = {
                    "selected_method": "ewm",
                    "selection_metric": VOL_FORECAST_SELECTION_METRIC,
                    "scores": {"ewm": None, "rolling": None, "garch": None},
                    "score_coverage": {},
                    "baseline_method": "ewm",
                    "baseline_score": None,
                    "best_score": None,
                    "improvement_vs_baseline": None,
                    "arch_available": bool(ARCH_AVAILABLE),
                    "vol_ref": float(vol_ref),
                    "vol_gate": float(vol_gate),
                    "split_idx": None,
                    "selection_end": None,
                }

            df_window["Ret20"] = df_window["Close"].pct_change(20).fillna(0.0)
            df_window["Ret1"] = df_window["Close"].pct_change(1).fillna(0.0)
            flip = (np.sign(df_window["Ret1"]) != np.sign(df_window["Ret1"].shift(1))).astype(float)
            df_window["FlipRate20"] = flip.rolling(20).mean().fillna(0.0)

            trend_thresh = (
                df_window["Ret20"]
                .abs()
                .rolling(200)
                .quantile(0.35)
                .fillna(df_window["Ret20"].abs().quantile(0.35))
            )

            flip_thresh = (
                df_window["FlipRate20"]
                .rolling(200)
                .quantile(0.75)
                .fillna(df_window["FlipRate20"].quantile(0.75))
            )

            trend_low = df_window["Ret20"].abs() < trend_thresh
            flip_high = df_window["FlipRate20"] > flip_thresh
            df_window["Chop"] = (trend_low & flip_high).astype(int)

            if ENABLE_VMD:
                if VMD_AVAILABLE:
                    logging.info(f"[VMD] {prefix} | building rolling VMD features")
                    df_window = add_vmd_feature_block(df_window)
                else:
                    logging.warning(f"[VMD] {prefix} | ENABLE_VMD=True but vmdpy not available; skipping VMD features")

            df_window = build_trade_gate_label(
                df_window,
                horizon=GATE_LABEL_HORIZON,
                ret_thr=GATE_LABEL_RETURN_THR,
            )

            if len(df_window) <= 52 or (len(df_window) % 2 != 0):
                df_window = df_window.iloc[:-1].reset_index(drop=True)

            gate_model = None
            gate_train_stats = {
                "auc": np.nan,
                "precision": np.nan,
                "recall": np.nan,
                "train_rows": 0,
                "eval_rows": 0,
                "train_pos_rate": np.nan,
                "eval_pos_rate": np.nan,
            }
            gate_features = choose_gate_features(df_window)
            gate_config = {
                "gate_enabled": bool(ENABLE_TRADE_GATE),
                "gate_model_type": GATE_MODEL_KIND,
                "threshold": float(GATE_THRESHOLD),
                "label_horizon": int(GATE_LABEL_HORIZON),
                "label_return_threshold": float(GATE_LABEL_RETURN_THR),
                "uses_scaler": False,
            }

            gate_model_path = None
            gate_features_path = None
            gate_config_path = None

            if ENABLE_TRADE_GATE:
                try:
                    split_for_gate = get_train_eval_split_idx(len(df_window))
                    df_gate_train = df_window.iloc[:split_for_gate].copy()
                    gate_fit = train_trade_gate(
                        df_gate=df_gate_train,
                        gate_features=gate_features,
                        threshold=GATE_THRESHOLD,
                    )
                    gate_model = gate_fit["model"]
                    gate_train_stats = {
                        "auc": gate_fit["auc"],
                        "precision": gate_fit["precision"],
                        "recall": gate_fit["recall"],
                        "train_rows": gate_fit["train_rows"],
                        "eval_rows": gate_fit["eval_rows"],
                        "train_pos_rate": gate_fit["train_pos_rate"],
                        "eval_pos_rate": gate_fit["eval_pos_rate"],
                    }
                    gate_config.update({
                        "train_rows": int(gate_fit["train_rows"]),
                        "eval_rows": int(gate_fit["eval_rows"]),
                        "auc": gate_fit["auc"],
                        "precision": gate_fit["precision"],
                        "recall": gate_fit["recall"],
                        "train_pos_rate": gate_fit["train_pos_rate"],
                        "eval_pos_rate": gate_fit["eval_pos_rate"],
                    })
                    gate_model_path, gate_features_path, gate_config_path = save_gate_model(
                        prefix=prefix,
                        gate_model=gate_model,
                        gate_features=gate_features,
                        gate_config=gate_config,
                    )
                except Exception as e:
                    logging.warning(f"[GATE] {prefix} gate training failed; gating disabled for this window. Err: {e}")
                    gate_model = None

            ppo_leak_cols = ["Trade_Label", "Gate_Future_Max_Return", "Gate_Future_Max_Close"]
            df_window_base = df_window.copy()
            if ENABLE_VMD:
                df_window_base, _ = neutralize_vmd_columns(df_window_base)

            df_window_base_ppo = df_window_base.drop(
                columns=[c for c in ppo_leak_cols if c in df_window_base.columns],
                errors="ignore"
            ).copy()
            df_env_base = make_env_df(df_window_base_ppo)

            use_vmd_candidate = ENABLE_VMD and VMD_AVAILABLE and (len(get_vmd_columns(df_window)) > 0)

            if use_vmd_candidate:
                df_window_vmd = df_window.copy()
                df_window_vmd_ppo = df_window_vmd.drop(
                    columns=[c for c in ppo_leak_cols if c in df_window_vmd.columns],
                    errors="ignore"
                ).copy()
                df_env_vmd = make_env_df(df_window_vmd_ppo)

                if len(df_env_vmd) < (WALKFORWARD_MIN_TRAIN_ROWS + WALKFORWARD_MIN_EVAL_ROWS + 5):
                    logging.warning(
                        f"{prefix}: VMD candidate disabled after env cleanup because df_env_vmd too short "
                        f"({len(df_env_vmd)} rows)."
                    )
                    df_env_vmd = None
                    use_vmd_candidate = False
            else:
                df_env_vmd = None

            if "Datetime" in df_window.columns:
                df_align = df_window[["Datetime", "Close"]].copy()
                df_align["Datetime"] = pd.to_datetime(df_align["Datetime"], errors="coerce")
            else:
                df_align = pd.DataFrame({"Datetime": [pd.NaT] * len(df_window), "Close": df_window["Close"].values})

            df_align["Close"] = pd.to_numeric(df_align["Close"], errors="coerce")
            df_align = df_align.dropna(subset=["Close"]).reset_index(drop=True)

            df_gate_live = df_window[gate_features].copy()
            for c in gate_features:
                df_gate_live[c] = pd.to_numeric(df_gate_live[c], errors="coerce")
            df_gate_live = df_gate_live.reset_index(drop=True)

            logging.info(
                f"[ALIGN PRE] {prefix} | "
                f"len_df_env_base={len(df_env_base)} | "
                f"len_df_env_vmd={len(df_env_vmd) if df_env_vmd is not None else 0} | "
                f"len_df_align={len(df_align)} | "
                f"len_df_gate_live={len(df_gate_live)} | "
                f"len_df_window={len(df_window)}"
            )

            n_base = min(len(df_env_base), len(df_align), len(df_gate_live), len(df_window))
            df_env_base = df_env_base.iloc[:n_base].reset_index(drop=True)

            if df_env_vmd is not None:
                n_vmd = min(len(df_env_vmd), len(df_align), len(df_gate_live), len(df_window))
                n = min(n_base, n_vmd)
                df_env_vmd = df_env_vmd.iloc[:n].reset_index(drop=True)
            else:
                n = n_base

            df_align = df_align.iloc[:n].reset_index(drop=True)
            df_gate_live = df_gate_live.iloc[:n].reset_index(drop=True)
            df_window = df_window.iloc[:n].reset_index(drop=True)
            df_env_base = df_env_base.iloc[:n].reset_index(drop=True)

            if len(df_env_base) < (WALKFORWARD_MIN_TRAIN_ROWS + WALKFORWARD_MIN_EVAL_ROWS + 5):
                logging.warning(
                    f"{prefix}: skipping after env cleanup because df_env_base too short "
                    f"({len(df_env_base)} rows)."
                )
                continue

            logging.info(
                f"[ALIGN POST] {prefix} | n={n} | "
                f"len_df_env_base={len(df_env_base)} | "
                f"len_df_env_vmd={len(df_env_vmd) if df_env_vmd is not None else 0} | "
                f"len_df_align={len(df_align)} | "
                f"len_df_gate_live={len(df_gate_live)} | "
                f"len_df_window={len(df_window)}"
            )

            split_idx_base = get_train_eval_split_idx(len(df_env_base))
            df_train_env_base = df_env_base.iloc[:split_idx_base].reset_index(drop=True)
            df_eval_env_base = df_env_base.iloc[split_idx_base:].reset_index(drop=True)

            df_align_base_eval = df_align.iloc[split_idx_base:].reset_index(drop=True)
            df_gate_live_base_eval = df_gate_live.iloc[split_idx_base:].reset_index(drop=True)
            df_window_base_eval_full = df_window.iloc[split_idx_base:].reset_index(drop=True)

            frame_bound_train_base = (50, len(df_train_env_base) - 3)
            if frame_bound_train_base[1] <= frame_bound_train_base[0]:
                raise ValueError(
                    f"{prefix}: bad BASE frame_bound={frame_bound_train_base} "
                    f"with len(df_train_env_base)={len(df_train_env_base)}"
                )

            model_base, env_base = build_and_train_ppo(
                train_df=df_train_env_base,
                frame_bound_train=frame_bound_train_base,
                ppo_overrides=ppo_overrides,
                timesteps=timesteps,
            )

            base_eval = evaluate_trained_policy_oos(
                model=model_base,
                src_vecnorm=env_base,
                df_eval_env=df_eval_env_base,
                df_align_eval=df_align_base_eval,
                df_gate_live_eval=df_gate_live_base_eval,
                df_window_eval_full=df_window_base_eval_full,
                gate_model=gate_model,
                prefix=prefix,
                ticker=ticker,
                window_label=f"{start}-{end}",
                gate_enabled=bool(ENABLE_TRADE_GATE and gate_model is not None),
                env_kwargs_override=None,
                variant_name="base",
            )
            base_metrics = base_eval["metrics"]

            use_vmd_final = False
            selected_feature_source = "BASE"

            if use_vmd_candidate and df_env_vmd is not None:
                split_idx_vmd = get_train_eval_split_idx(len(df_env_vmd))
                df_train_env_vmd = df_env_vmd.iloc[:split_idx_vmd].reset_index(drop=True)
                df_eval_env_vmd = df_env_vmd.iloc[split_idx_vmd:].reset_index(drop=True)
                df_align_vmd_eval = df_align.iloc[split_idx_vmd:].reset_index(drop=True)
                df_gate_live_vmd_eval = df_gate_live.iloc[split_idx_vmd:].reset_index(drop=True)
                df_window_vmd_eval_full = df_window.iloc[split_idx_vmd:].reset_index(drop=True)

                frame_bound_train_vmd = (50, len(df_train_env_vmd) - 3)
                if frame_bound_train_vmd[1] <= frame_bound_train_vmd[0]:
                    raise ValueError(
                        f"{prefix}: bad VMD frame_bound={frame_bound_train_vmd} "
                        f"with len(df_train_env_vmd)={len(df_train_env_vmd)}"
                    )

                logging.info(
                    f"Training VMD {ticker} window {w_idx + 1}/{len(windows)} | "
                    f"timesteps={timesteps} | train_rows={len(df_train_env_vmd)} | eval_rows={len(df_eval_env_vmd)}"
                )
                model_vmd, env_vmd = build_and_train_ppo(
                    train_df=df_train_env_vmd,
                    frame_bound_train=frame_bound_train_vmd,
                    ppo_overrides=ppo_overrides,
                    timesteps=timesteps,
                )

                vmd_eval = evaluate_trained_policy_oos(
                    model=model_vmd,
                    src_vecnorm=env_vmd,
                    df_eval_env=df_eval_env_vmd,
                    df_align_eval=df_align_vmd_eval,
                    df_gate_live_eval=df_gate_live_vmd_eval,
                    df_window_eval_full=df_window_vmd_eval_full,
                    gate_model=gate_model,
                    prefix=prefix,
                    ticker=ticker,
                    window_label=f"{start}-{end}",
                    gate_enabled=bool(ENABLE_TRADE_GATE and gate_model is not None),
                    env_kwargs_override=None,
                    variant_name="vmd",
                )
                vmd_metrics = vmd_eval["metrics"]

                sharpe_gain = vmd_metrics["sharpe"] - base_metrics["sharpe"]
                final_ok = (
                    vmd_metrics["final_value"] >= base_metrics["final_value"]
                    if VMD_REQUIRE_FINAL_NOT_WORSE else True
                )

                logging.info(
                    f"[VMD CHECK] {prefix} | "
                    f"Sharpe_Base={base_metrics['sharpe']:.3f} "
                    f"Sharpe_VMD={vmd_metrics['sharpe']:.3f} "
                    f"Gain={sharpe_gain:.3f} "
                    f"| Final_Base={base_metrics['final_value']:.2f} "
                    f"Final_VMD={vmd_metrics['final_value']:.2f}"
                )

                use_vmd_final = (
                    ENABLE_VMD_MODEL_SELECTION
                    and (sharpe_gain >= VMD_MIN_SHARPE_IMPROVEMENT)
                    and final_ok
                )

                if use_vmd_final:
                    m = vmd_metrics
                    step_df = vmd_eval["step_df"]
                    trades = vmd_eval["trades"]
                    nav_track = vmd_eval["nav_track"]
                    bh_track = vmd_eval["bh_track"]
                    model = model_vmd
                    env = env_vmd
                    selected_feature_source = "VMD"

                    try:
                        env_base.close()
                    except Exception:
                        pass
                    del model_base, env_base
                else:
                    m = base_metrics
                    step_df = base_eval["step_df"]
                    trades = base_eval["trades"]
                    nav_track = base_eval["nav_track"]
                    bh_track = base_eval["bh_track"]
                    model = model_base
                    env = env_base
                    selected_feature_source = "BASE"

                    try:
                        env_vmd.close()
                    except Exception:
                        pass
                    del model_vmd, env_vmd
            else:
                m = base_metrics
                step_df = base_eval["step_df"]
                trades = base_eval["trades"]
                nav_track = base_eval["nav_track"]
                bh_track = base_eval["bh_track"]
                model = model_base
                env = env_base
                selected_feature_source = "BASE"

            if int(m["executed_trade_count"]) != int(step_df["Executed_Trade_Flag"].sum()):
                logging.warning(
                    f"[SANITY] {prefix} executed_trade_count metric does not match step_df executed trades"
                )

            if int(m["threshold_signal_count"]) != int(step_df["Threshold_Signal_Flag"].sum()):
                logging.warning(
                    f"[SANITY] {prefix} threshold_signal_count metric does not match threshold signal flags"
                )

            if ENABLE_BOOTSTRAP_CI:
                bootstrap_ci = bootstrap_ci_from_returns(
                    ret_s=m["returns"],
                    b=BOOTSTRAP_B,
                    alpha=BOOTSTRAP_ALPHA,
                    block_size=BOOTSTRAP_BLOCK_SIZE,
                )
            else:
                bootstrap_ci = {
                    "n_obs": int(len(m["returns"])),
                    "bootstrap_b": 0,
                    "alpha": BOOTSTRAP_ALPHA,
                    "block_size": BOOTSTRAP_BLOCK_SIZE,
                    "return_mean_boot": float(np.prod(1.0 + m["returns"]) - 1.0) if len(m["returns"]) else 0.0,
                    "return_ci_low": np.nan,
                    "return_ci_high": np.nan,
                    "sharpe_mean_boot": float(m["sharpe"]),
                    "sharpe_ci_low": np.nan,
                    "sharpe_ci_high": np.nan,
                }

            selected_df_eval_env = df_eval_env_vmd if use_vmd_final else df_eval_env_base
            selected_df_align_eval = df_align_vmd_eval if use_vmd_final else df_align_base_eval
            selected_df_gate_live_eval = df_gate_live_vmd_eval if use_vmd_final else df_gate_live_base_eval
            selected_df_window_eval_full = df_window_vmd_eval_full if use_vmd_final else df_window_base_eval_full
            selected_train_rows = len(df_train_env_vmd) if use_vmd_final else len(df_train_env_base)
            selected_eval_rows = len(df_eval_env_vmd) if use_vmd_final else len(df_eval_env_base)

            regime_df = pd.DataFrame()
            if ENABLE_REGIME_BREAKDOWN:
                regime_df = build_regime_breakdown(
                    step_df=step_df,
                    ticker=ticker,
                    prefix=prefix,
                    window_label=f"{start}-{end}",
                    raw_eval_df=selected_df_window_eval_full,
                )

            ablation_rows = []
            if ENABLE_ABLATION_TABLES:
                full_final = float(m["final_value"])
                full_sharpe = float(m["sharpe"])
                full_drawdown = float(m["drawdown"])

                ablation_variants = ["full", "no_vol_controls", "no_slo", "no_chop", "no_vmd"]
                for variant in ablation_variants:
                    df_eval_variant, env_over, neutralized_cols = apply_ablation_variant(selected_df_eval_env, variant)

                    if variant == "no_vmd" and len(neutralized_cols) == 0:
                        ablation_rows.append({
                            "Ticker": ticker,
                            "Prefix": prefix,
                            "Window": f"{start}-{end}",
                            "Variant": variant,
                            "Status": "not_applicable",
                            "Neutralized_Cols": "",
                            "Final_Portfolio": np.nan,
                            "Sharpe": np.nan,
                            "Drawdown": np.nan,
                            "Delta_Final_vs_Full": np.nan,
                            "Delta_Sharpe_vs_Full": np.nan,
                            "Delta_Drawdown_vs_Full": np.nan,
                        })
                        continue

                    if variant == "full":
                        ab_m = m
                    else:
                        ab_eval = evaluate_trained_policy_oos(
                            model=model,
                            src_vecnorm=env,
                            df_eval_env=df_eval_variant,
                            df_align_eval=selected_df_align_eval,
                            df_gate_live_eval=selected_df_gate_live_eval,
                            df_window_eval_full=selected_df_window_eval_full,
                            gate_model=gate_model,
                            prefix=prefix,
                            ticker=ticker,
                            window_label=f"{start}-{end}",
                            gate_enabled=bool(ENABLE_TRADE_GATE and gate_model is not None),
                            env_kwargs_override=env_over,
                            variant_name=variant,
                        )
                        ab_m = ab_eval["metrics"]

                    ablation_rows.append({
                        "Ticker": ticker,
                        "Prefix": prefix,
                        "Window": f"{start}-{end}",
                        "Variant": variant,
                        "Status": "ok",
                        "Neutralized_Cols": ",".join(neutralized_cols),
                        "Final_Portfolio": float(ab_m["final_value"]),
                        "Sharpe": float(ab_m["sharpe"]),
                        "Drawdown": float(ab_m["drawdown"]),
                        "Delta_Final_vs_Full": float(ab_m["final_value"]) - full_final,
                        "Delta_Sharpe_vs_Full": float(ab_m["sharpe"]) - full_sharpe,
                        "Delta_Drawdown_vs_Full": float(ab_m["drawdown"]) - full_drawdown,
                    })

            ablation_df = pd.DataFrame(ablation_rows)
            save_phase4_window_outputs(prefix=prefix, ablation_df=ablation_df, regime_df=regime_df, bootstrap_ci=bootstrap_ci)

            logging.info(
                f"[PHASE4 OOS] {prefix} | sharpe={m['sharpe']:.4f} | "
                f"final={m['final_value']:.2f} | buyhold={m['hold_value']:.2f} | "
                f"dd={m['drawdown']:.2f} | rows_eval={len(selected_df_eval_env)}"
            )

            trades_path = os.path.join(RUN_RESULTS_DIR, f"{prefix}_trades.csv")
            pd.DataFrame(trades).to_csv(trades_path, index=False)

            model.save(model_path)
            try:
                env.save(vecnorm_path)
            except Exception as e:
                logging.warning(f"VecNormalize save failed for {prefix}: {e}")
                try:
                    if os.path.exists(model_path):
                        os.remove(model_path)
                except Exception:
                    pass
                raise

            exclude_feats = {
                "Datetime", "Symbol", "Target", "Return",
                "Trade_Label", "Gate_Future_Max_Return", "Gate_Future_Max_Close"
            }
            feature_cols = [
                c for c in df_window.columns
                if c not in exclude_feats and pd.api.types.is_numeric_dtype(df_window[c])
            ]

            model_info = {
                "phase4_oos_only": True,
                "risk_forecast_info": {
                    "enabled": bool(ENABLE_RISK_FORECAST_UPGRADE),
                    "selected_method": vol_diag.get("selected_method"),
                    "selection_metric": vol_diag.get("selection_metric"),
                    "scores": vol_diag.get("scores"),
                    "score_coverage": vol_diag.get("score_coverage"),
                    "baseline_method": vol_diag.get("baseline_method"),
                    "baseline_score": vol_diag.get("baseline_score"),
                    "best_score": vol_diag.get("best_score"),
                    "improvement_vs_baseline": vol_diag.get("improvement_vs_baseline"),
                    "arch_available": vol_diag.get("arch_available"),
                    "vol_ref": vol_diag.get("vol_ref"),
                    "vol_gate": vol_diag.get("vol_gate"),
                    "split_idx": vol_diag.get("split_idx"),
                    "selection_end": vol_diag.get("selection_end"),
                },
                "model": "PPO",
                "ticker": ticker,
                "window": f"{start}-{end}",
                "window_idx": int(w_idx + 1),
                "prefix": prefix,
                "date_trained_utc": now_utc().strftime("%Y-%m-%dT%H:%M:%SZ"),
                "framework": "stable-baselines3",
                "device": DEVICE_STR,
                "seed": SEED,
                "timesteps": int(timesteps),
                "train_rows": int(selected_train_rows),
                "eval_rows": int(selected_eval_rows),
                "vmd_info": {
                    "enabled": bool(ENABLE_VMD),
                    "available": bool(VMD_AVAILABLE),
                    "source_col": VMD_SOURCE_COL,
                    "lookback": int(VMD_LOOKBACK),
                    "K": int(VMD_K),
                    "trend_modes": VMD_TREND_MODES if "VMD_TREND_MODES" in globals() else [],
                    "cycle_modes": VMD_CYCLE_MODES if "VMD_CYCLE_MODES" in globals() else [],
                    "noise_modes": VMD_NOISE_MODES if "VMD_NOISE_MODES" in globals() else [],
                    "selected": bool(use_vmd_final),
                    "candidate_available": bool(use_vmd_candidate),
                    "selected_feature_source": selected_feature_source,
                },

                "metrics": {
                    "final_portfolio": round(m["final_value"], 2),
                    "buy_hold": round(m["hold_value"], 2),
                    "sharpe": round(m["sharpe"], 6),
                    "drawdown_pct": round(m["drawdown"], 6),
                    "executed_trade_count": int(m["executed_trade_count"]),
                    "threshold_signal_count": int(m["threshold_signal_count"]),
                    "raw_signal_count": int(m["raw_signal_count"]),
                    "signal_trade_count": int(m["threshold_signal_count"]),
                    "action_threshold": float(m["thr_window"]),
                    "accuracy": float(m["accuracy"]),
                }
            }

            write_master_bundle(
                prefix=prefix,
                features=feature_cols,
                threshold=float(m["thr_window"]),
                model_info=model_info,
                gate_features=gate_features if gate_model is not None else None,
                gate_threshold=float(GATE_THRESHOLD) if gate_model is not None else None,
            )

            pred_path = os.path.join(RUN_RESULTS_DIR, f"{prefix}_predictions.csv")
            step_df.to_csv(pred_path, index=False)

            compat_rows = []
            for _, r in step_df.iterrows():
                a = float(r["Action"])
                if a > float(m["thr_window"]):
                    signal = "BUY"
                elif a < -float(m["thr_window"]):
                    signal = "SELL"
                else:
                    signal = "HOLD"

                compat_rows.append({
                    "Index": r["Index"],
                    "Datetime": r["Datetime"],
                    "Close": r["Close"],
                    "Raw_Action": r["Raw_Action"],
                    "Action": a,
                    "Signal": signal,
                    "PortfolioValue": r["nav"],
                    "Reward": r.get("reward", np.nan),
                    "Gate_Prob": r.get("Gate_Prob", np.nan),
                    "Gate_Passed": r.get("Gate_Passed", np.nan),
                    "Gate_Blocked": r.get("Gate_Blocked", np.nan),
                    "Variant": r.get("Variant", "full"),
                })

            compat_path = os.path.join(RUN_RESULTS_DIR, f"{prefix}_predictions_compat.csv")
            pd.DataFrame(compat_rows).to_csv(compat_path, index=False)

            manifest_record_paths(
                prefix,
                {
                    "model_zip": model_path,
                    "vecnorm_pkl": vecnorm_path,
                    "predictions_csv": pred_path,
                    "predictions_compat_csv": compat_path,
                    "trades_csv": trades_path,
                    "gate_model_pkl": gate_model_path if (gate_model_path and os.path.exists(gate_model_path)) else None,
                    "gate_features_json": gate_features_path if (gate_features_path and os.path.exists(gate_features_path)) else None,
                    "gate_config_json": gate_config_path if (gate_config_path and os.path.exists(gate_config_path)) else None,
                },
                do_hashes=ENABLE_HASHES,
            )

            manifest_update(
                prefix,
                {
                    "status": "trained",
                    "ticker": ticker,
                    "window": f"{start}-{end}",
                    "window_idx": int(w_idx + 1),
                    "sharpe": float(m["sharpe"]),
                    "gate_enabled": bool(ENABLE_TRADE_GATE and gate_model is not None),
                    "phase4_oos_only": True,
                }
            )

            result_row = {
                "Vol_Method_Selected": vol_diag.get("selected_method"),
                "Vol_Selection_Metric": vol_diag.get("selection_metric"),
                "Vol_Baseline_Method": vol_diag.get("baseline_method"),
                "Vol_Improve_vs_Baseline": vol_diag.get("improvement_vs_baseline"),
                "Vol_Score_EWM": (vol_diag.get("scores") or {}).get("ewm"),
                "Vol_Score_Rolling": (vol_diag.get("scores") or {}).get("rolling"),
                "Vol_Score_GARCH": (vol_diag.get("scores") or {}).get("garch"),
                "Ticker": ticker,
                "Window": f"{start}-{end}",
                "WindowIdx": int(w_idx + 1),
                "Prefix": prefix,
                "Train_Rows": int(selected_train_rows),
                "Eval_Rows": int(selected_eval_rows),
                "Eval_Mode": "OOS_ONLY",
                "PPO_Portfolio": round(m["final_value"], 2),
                "BuyHold": round(m["hold_value"], 2),
                "Return_Pct": round((m["final_value"] / 100000.0) - 1.0, 6),
                "Sharpe": round(m["sharpe"], 3),
                "Drawdown_%": round(m["drawdown"], 2),
                "Winner": "PPO" if m["final_value"] > m["hold_value"] else "Buy & Hold",
                "Action_Threshold": round(m["thr_window"], 4),
                "Accuracy": float(m["accuracy"]),
                "Trade_Count": int(m["executed_trade_count"]),
                "Signal_Trade_Count": int(m["threshold_signal_count"]),
                "Raw_Signal_Count": int(m["raw_signal_count"]),
                "Executed_Trade_Count": int(m["executed_trade_count"]),
                "Vol_Gated_Count": int(m["vol_gated_count"]),
                "Forced_Close_Count": int(m["forced_close_count"]),
                "SLO_Trigger_Count": int(m["slo_trigger_count"]),
                "Chop_Locked_Count": int(m["chop_locked_count"]),
                "Precision_Long": round(m["precision_long"], 4),
                "Precision_Short": round(m["precision_short"], 4),
                "Precision_Trades": round(m["precision_trades"], 4),
                "Bootstrap_Sharpe_CI_Low": bootstrap_ci.get("sharpe_ci_low", np.nan),
                "Bootstrap_Sharpe_CI_High": bootstrap_ci.get("sharpe_ci_high", np.nan),
                "Bootstrap_Return_CI_Low": bootstrap_ci.get("return_ci_low", np.nan),
                "Bootstrap_Return_CI_High": bootstrap_ci.get("return_ci_high", np.nan),
                "Gate_Model": GATE_MODEL_KIND if gate_model is not None else "disabled",
                "Gate_Threshold": float(GATE_THRESHOLD) if gate_model is not None else np.nan,
                "Gate_Pass_Count": int(m["gate_pass_count"]),
                "Gate_Block_Count": int(m["gate_block_count"]),
                "Gate_Pass_Rate": m["gate_pass_rate"],
                "Gate_AUC_Train": gate_train_stats["auc"],
                "Gate_Precision_Train": gate_train_stats["precision"],
                "Gate_Recall_Train": gate_train_stats["recall"],
                "Gate_AUC_Eval": m["gate_auc_eval"],
                "Gate_Precision_Eval": m["gate_precision_eval"],
                "Gate_Recall_Eval": m["gate_recall_eval"],
            }
            results.append(result_row)

            meta = {
                "result": result_row,
                "features": feature_cols,
                "prefix": prefix,
                "model_path": model_path,
                "vecnorm_path": vecnorm_path,
                "gate_model_path": gate_model_path if (gate_model_path and os.path.exists(gate_model_path)) else None,
                "gate_features_path": gate_features_path if (gate_features_path and os.path.exists(gate_features_path)) else None,
                "gate_config_path": gate_config_path if (gate_config_path and os.path.exists(gate_config_path)) else None,
                "phase4_oos_only": True,
                "risk_forecast_enabled": bool(ENABLE_RISK_FORECAST_UPGRADE),
                "vol_method_selected": vol_diag.get("selected_method"),
                "vol_selection_metric": vol_diag.get("selection_metric"),
                "vol_scores": vol_diag.get("scores"),
                "vol_score_coverage": vol_diag.get("score_coverage"),
                "vol_baseline_method": vol_diag.get("baseline_method"),
                "vol_baseline_score": vol_diag.get("baseline_score"),
                "vol_best_score": vol_diag.get("best_score"),
                "vol_improve_vs_baseline": vol_diag.get("improvement_vs_baseline"),
                "vol_arch_available": vol_diag.get("arch_available"),
                "vol_ref": vol_diag.get("vol_ref"),
                "vol_gate": vol_diag.get("vol_gate"),
                "vol_split_idx": vol_diag.get("split_idx"),
                "vol_selection_end": vol_diag.get("selection_end"),
            }

            item = (float(result_row["Sharpe"]), prefix, meta)
            if len(top_heap) < TOP_N_WINDOWS:
                heapq.heappush(top_heap, item)
            else:
                if item[0] > top_heap[0][0]:
                    heapq.heapreplace(top_heap, item)

            logging.info(
                f"{ticker} | window {w_idx + 1} done | "
                f"OOS sharpe={result_row['Sharpe']} | runtime={round(time.time() - window_start_time, 2)}s"
            )

            top_list = sorted(top_heap, key=lambda t: t[0], reverse=True)

            for _, _, m_top in top_list:
                artifact_for_save = {
                    "model": None,
                    "model_path": m_top["model_path"],
                    "vecnorm_path": m_top["vecnorm_path"],
                    "features": m_top["features"],
                    "result": m_top["result"],
                    "prefix": m_top["prefix"],
                    "gate_model_path": m_top.get("gate_model_path"),
                    "gate_features_path": m_top.get("gate_features_path"),
                    "gate_config_path": m_top.get("gate_config_path"),
                    "phase4_oos_only": m_top.get("phase4_oos_only", True),
                    "risk_forecast_enabled": m_top.get("risk_forecast_enabled"),
                    "vol_method_selected": m_top.get("vol_method_selected"),
                    "vol_selection_metric": m_top.get("vol_selection_metric"),
                    "vol_scores": m_top.get("vol_scores"),
                    "vol_score_coverage": m_top.get("vol_score_coverage"),
                    "vol_baseline_method": m_top.get("vol_baseline_method"),
                    "vol_baseline_score": m_top.get("vol_baseline_score"),
                    "vol_best_score": m_top.get("vol_best_score"),
                    "vol_improve_vs_baseline": m_top.get("vol_improve_vs_baseline"),
                    "vol_arch_available": m_top.get("vol_arch_available"),
                    "vol_ref": m_top.get("vol_ref"),
                    "vol_gate": m_top.get("vol_gate"),
                    "vol_split_idx": m_top.get("vol_split_idx"),
                    "vol_selection_end": m_top.get("vol_selection_end"),
                }
                if GATE_TOP_EXPORT_TO_QC:
                    save_quantconnect_model(artifact_for_save, m_top["prefix"], QC_TOP_DIR)

            try:
                env.close()
            except Exception:
                pass

            del env, model
            gc.collect()
            try:
                torch.cuda.empty_cache()
            except Exception:
                pass

        except Exception as e:
            logging.error(f"{prefix}: window failed: {e}", exc_info=True)
            continue

    if skipped_windows:
        record_skips_global(ticker, skipped_windows, total_windows=len(windows), fully_skipped=False)

    return results

# -------------------------
# Parallel runner
# -------------------------
from concurrent.futures import ThreadPoolExecutor


def process_ticker(ticker: str) -> list[dict]:
    try:
        hp = pick_params(ticker)
        df_sym = df[df["Symbol"] == ticker].copy()
        return walkforward_ppo(
            df_sym=df_sym,
            ticker=ticker,
            window_size=WINDOW_SIZE,
            step_size=STEP_SIZE,
            timesteps=TIMESTEPS,
            ppo_overrides=hp,
        )
    except Exception as e:
        logging.error(f"{ticker}: training failed: {e}")
        return []


def run_parallel_tickers(tickers: list[str], out_path: str, max_workers: int):
    logging.info(f"[PARALLEL] max_workers={max_workers}")
    results = []
    with ThreadPoolExecutor(max_workers=max_workers) as ex:
        for res in ex.map(process_ticker, tickers):
            if res:
                results.extend(res)

    if results:
        pd.DataFrame(results).to_csv(out_path, index=False)
        logging.info(f"Saved summary -> {out_path}")
    else:
        logging.warning("No results produced; summary not written.")

    return results


# -------------------------
# Phase 4 aggregate reports
# -------------------------
def build_phase4_reports():
    abl_files = sorted(glob.glob(os.path.join(RUN_RESULTS_DIR, "*_ablation.csv")))
    reg_files = sorted(glob.glob(os.path.join(RUN_RESULTS_DIR, "*_regime_breakdown.csv")))
    sum_files = sorted(glob.glob(os.path.join(RUN_RESULTS_DIR, "summary*.csv")))

    if abl_files:
        abl_frames = []
        for p in abl_files:
            try:
                abl_frames.append(pd.read_csv(p))
            except Exception as e:
                logging.warning(f"[PHASE4] failed to read ablation file {p}: {e}")
        if abl_frames:
            abl_all = pd.concat(abl_frames, ignore_index=True)
            abl_all.to_csv(PHASE4_ABLATION_PATH, index=False)
            logging.info(f"[PHASE4] Saved ablation aggregate -> {PHASE4_ABLATION_PATH}")

    if reg_files:
        reg_frames = []
        for p in reg_files:
            try:
                reg_frames.append(pd.read_csv(p))
            except Exception as e:
                logging.warning(f"[PHASE4] failed to read regime file {p}: {e}")
        if reg_frames:
            reg_all = pd.concat(reg_frames, ignore_index=True)
            reg_all.to_csv(PHASE4_REGIME_PATH, index=False)
            logging.info(f"[PHASE4] Saved regime aggregate -> {PHASE4_REGIME_PATH}")

    if sum_files:
        frames = []
        for p in sum_files:
            try:
                frames.append(pd.read_csv(p))
            except Exception:
                pass

        if frames:
            combo = pd.concat(frames, ignore_index=True)

            if "Executed_Trade_Count" in combo.columns:
                trade_count_col = "Executed_Trade_Count"
            elif "Trade_Count" in combo.columns:
                trade_count_col = "Trade_Count"
            else:
                trade_count_col = None

            if "Ticker" in combo.columns:
                agg_map = {
                    "Sharpe": "mean",
                    "PPO_Portfolio": "mean",
                    "BuyHold": "mean",
                    "Drawdown_%": "mean",
                    "Bootstrap_Sharpe_CI_Low": "mean",
                    "Bootstrap_Sharpe_CI_High": "mean",
                    "Bootstrap_Return_CI_Low": "mean",
                    "Bootstrap_Return_CI_High": "mean",
                    "Eval_Rows": "sum",
                }
                if trade_count_col is not None:
                    agg_map[trade_count_col] = "sum"

                agg = (
                    combo.groupby("Ticker", as_index=False)
                    .agg(agg_map)
                    .sort_values("Sharpe", ascending=False)
                )

                if trade_count_col is not None:
                    agg = agg.rename(columns={trade_count_col: "Executed_Trade_Count"})
                else:
                    agg["Executed_Trade_Count"] = np.nan

                agg.to_csv(PHASE4_SUMMARY_PATH, index=False)
                logging.info(f"[PHASE4] Saved ticker aggregate -> {PHASE4_SUMMARY_PATH}")


# -------------------------
# Selector builder
# -------------------------
MODEL_NAME = "PPO"


def build_ppo_selector():
    summary_files = glob.glob(os.path.join(BASE_RESULTS_DIR, "ppo_walkforward_results_*", "summary*.csv"))
    frames = []

    for p in summary_files:
        try:
            tmp = pd.read_csv(p)
            tmp["RunFolder"] = os.path.dirname(p)
            frames.append(tmp)
        except Exception as e:
            logging.warning(f"Skipping summary {p}: {e}")

    if not frames:
        logging.warning("No PPO summaries found across run folders.")
        return

    combo = pd.concat(frames, ignore_index=True)

    if "Sharpe" not in combo.columns:
        logging.warning("Combined summaries missing 'Sharpe'. Selector will be empty.")
        return

    combo["Sharpe"] = pd.to_numeric(combo["Sharpe"], errors="coerce")
    combo = combo.dropna(subset=["Sharpe"])
    if combo.empty:
        logging.warning("No numeric Sharpe rows after coercion.")
        return
    if "BuyHold" not in combo.columns:
        combo["BuyHold"] = np.nan
    if "PPO_Portfolio" not in combo.columns:
        combo["PPO_Portfolio"] = np.nan
    if "Drawdown_%" not in combo.columns:
        combo["Drawdown_%"] = np.nan
    if "Vol_Improve_vs_Baseline" not in combo.columns:
        combo["Vol_Improve_vs_Baseline"] = np.nan
    if "Vol_Method_Selected" not in combo.columns:
        combo["Vol_Method_Selected"] = np.nan

    combo["Vol_Improve_vs_Baseline"] = pd.to_numeric(
        combo["Vol_Improve_vs_Baseline"], errors="coerce"
    )

    combo["Rel_vs_BH"] = np.where(
        pd.to_numeric(combo["BuyHold"], errors="coerce").fillna(0.0) != 0.0,
        pd.to_numeric(combo["PPO_Portfolio"], errors="coerce")
        / pd.to_numeric(combo["BuyHold"], errors="coerce"),
        np.nan
    )

    sharpe_pos_summary = (
        combo.groupby("Ticker", as_index=False)
        .agg(
            Positive_Sharpe_Windows=("Sharpe", lambda s: int((pd.to_numeric(s, errors="coerce") > 0).sum())),
            Total_Windows=("Sharpe", "count"),
        )
    )

    best_by_symbol = (
        combo.sort_values(
            ["Ticker", "Sharpe", "Vol_Improve_vs_Baseline", "Rel_vs_BH"],
            ascending=[True, False, False, False]
        )
        .groupby("Ticker")
        .first()
        .reset_index()
    )

    best_by_symbol = best_by_symbol.rename(
        columns={"Drawdown_%": "Drawdown", "PPO_Portfolio": "Final_Portfolio"}
    )

    best_by_symbol = best_by_symbol.merge(
        sharpe_pos_summary,
        on="Ticker",
        how="left"
    )

    best_by_symbol["Abs_Return_Pct"] = np.where(
        pd.to_numeric(best_by_symbol["Final_Portfolio"], errors="coerce").notna(),
        (pd.to_numeric(best_by_symbol["Final_Portfolio"], errors="coerce") / 100000.0) - 1.0,
        np.nan
    )

    best_by_symbol["Model"] = MODEL_NAME

    best_by_symbol.to_csv(SELECTOR_FULL_PATH, index=False)
    logging.info(f"Selector FULL saved -> {SELECTOR_FULL_PATH}")

    df_sel = best_by_symbol.copy()

    phase3_ok = (
        df_sel["Vol_Improve_vs_Baseline"].fillna(0.0) >= VOL_FORECAST_MIN_IMPROVEMENT
    )

    MIN_ABS_RETURN_PCT = 0.03
    DEPLOY_EXCLUDED_SYMBOLS = {"IBM"}

    if "Executed_Trade_Count" in df_sel.columns:
        trade_count_filter = pd.to_numeric(df_sel["Executed_Trade_Count"], errors="coerce").fillna(0)
    elif "Trade_Count" in df_sel.columns:
        trade_count_filter = pd.to_numeric(df_sel["Trade_Count"], errors="coerce").fillna(0)
    else:
        trade_count_filter = pd.Series(0, index=df_sel.index, dtype=float)

    ppo_beats_bh = (
        pd.to_numeric(df_sel["Final_Portfolio"], errors="coerce").fillna(0)
        >= pd.to_numeric(df_sel["BuyHold"], errors="coerce").fillna(np.inf)
    )

    abs_return_ok = (
        pd.to_numeric(df_sel["Abs_Return_Pct"], errors="coerce").fillna(-999)
        >= MIN_ABS_RETURN_PCT
    )

    base_gates = (
        (df_sel["Sharpe"].fillna(-999) > 0.0)
        & (df_sel["Drawdown"].fillna(999) < 50.0)
        & (df_sel["Final_Portfolio"].fillna(0) > 80_000)
        & ppo_beats_bh
        & abs_return_ok
        & (df_sel["Positive_Sharpe_Windows"].fillna(0) >= 2)
        & (trade_count_filter >= 10)
    )

    logging.info(
        "[SELECTOR FILTER] Applying strict rules: "
        "PPO >= BuyHold, Abs_Return_Pct >= 0.03, Positive_Sharpe_Windows >= 2, Trade_Count >= 10"
    )

    if ENABLE_RISK_FORECAST_UPGRADE:
        df_sel = df_sel[base_gates & phase3_ok].copy()
    else:
        df_sel = df_sel[base_gates].copy()

    df_sel = df_sel[~df_sel["Ticker"].isin(DEPLOY_EXCLUDED_SYMBOLS)].copy()

    df_sel["prefix"] = (
        "ppo_" + df_sel["Ticker"].astype(str) + "_window"
        + df_sel["WindowIdx"].fillna(1).astype(int).astype(str)
    )
    df_sel["artifact_path"] = df_sel["prefix"].apply(lambda p: os.path.join(FINAL_MODEL_DIR, f"{p}_model.zip"))
    df_sel["vecnorm_path"] = df_sel["prefix"].apply(lambda p: os.path.join(FINAL_MODEL_DIR, f"{p}_vecnorm.pkl"))
    df_sel["gate_model_path"] = df_sel["prefix"].apply(lambda p: os.path.join(FINAL_MODEL_DIR, f"{p}_gate_model.pkl"))
    df_sel["gate_features_path"] = df_sel["prefix"].apply(lambda p: os.path.join(FINAL_MODEL_DIR, f"{p}_gate_features.json"))
    df_sel["gate_config_path"] = df_sel["prefix"].apply(lambda p: os.path.join(FINAL_MODEL_DIR, f"{p}_gate_config.json"))

    selected_models = {}
    for _, row in df_sel.iterrows():
        tkr = str(row["Ticker"])
        final_port = float(row["Final_Portfolio"]) if pd.notna(row["Final_Portfolio"]) else np.nan
        buyhold = float(row["BuyHold"]) if pd.notna(row["BuyHold"]) else np.nan
        abs_ret = float(row["Abs_Return_Pct"]) if pd.notna(row["Abs_Return_Pct"]) else np.nan

        logging.info(
            f"[SELECTOR] {tkr} | Sharpe={float(row['Sharpe']):.3f} "
            f"| Final={final_port:.2f} | BuyHold={buyhold:.2f} | AbsRet={abs_ret:.4f} "
            f"| PosWin={int(row['Positive_Sharpe_Windows']) if pd.notna(row['Positive_Sharpe_Windows']) else 0} "
            f"| TotalWin={int(row['Total_Windows']) if pd.notna(row['Total_Windows']) else 0}"
        )
        selected_models[tkr] = {
            "vol_method_selected": row.get("Vol_Method_Selected", None),
            "vol_improve_vs_baseline": (
                float(row["Vol_Improve_vs_Baseline"])
                if "Vol_Improve_vs_Baseline" in row and pd.notna(row["Vol_Improve_vs_Baseline"])
                else None
            ),
            "vol_selection_metric": row.get("Vol_Selection_Metric", None),
            "model": MODEL_NAME,
            "score": float(row["Sharpe"]),
            "sharpe": float(row["Sharpe"]),
            "return": float(row["Final_Portfolio"]) if pd.notna(row["Final_Portfolio"]) else None,
            "drawdown": float(row["Drawdown"]) if pd.notna(row["Drawdown"]) else None,
            "trade_count": int(
                row["Executed_Trade_Count"]
                if "Executed_Trade_Count" in row and pd.notna(row["Executed_Trade_Count"])
                else row["Trade_Count"]
            ) if (
                ("Executed_Trade_Count" in row and pd.notna(row["Executed_Trade_Count"])) or
                ("Trade_Count" in row and pd.notna(row["Trade_Count"]))
            ) else None,
            "gate_model": row.get("Gate_Model", None),
            "gate_threshold": float(row["Gate_Threshold"]) if "Gate_Threshold" in row and pd.notna(row["Gate_Threshold"]) else None,
            "gate_pass_count": int(row["Gate_Pass_Count"]) if "Gate_Pass_Count" in row and pd.notna(row["Gate_Pass_Count"]) else None,
            "gate_block_count": int(row["Gate_Block_Count"]) if "Gate_Block_Count" in row and pd.notna(row["Gate_Block_Count"]) else None,
            "artifact": {
                "path": row["artifact_path"],
                "vecnorm": row["vecnorm_path"],
                "exists": bool(os.path.exists(row["artifact_path"])),
                "gate_model_path": row["gate_model_path"] if os.path.exists(row["gate_model_path"]) else None,
                "gate_features_path": row["gate_features_path"] if os.path.exists(row["gate_features_path"]) else None,
                "gate_config_path": row["gate_config_path"] if os.path.exists(row["gate_config_path"]) else None,
            },
        }

    with open(SELECTOR_JSON_PATH, "w") as f:
        json.dump(selected_models, f, indent=2)
    logging.info(f"Selector JSON saved -> {SELECTOR_JSON_PATH}")


# -------------------------
# Main
# -------------------------
if __name__ == "__main__":
    hc = health_check_latest_run(BASE_RESULTS_DIR)
    if hc.get("ok"):
        logging.info(f"[HEALTH] Latest run folder: {hc['latest_run_dir']}")
        logging.info(f"[HEALTH] Summary files: {hc['summary_files']}")
        logging.info(f"[HEALTH] Rows total: {hc['rows_total']} | tickers: {len(hc['tickers'])}")
    else:
        logging.warning(f"[HEALTH] {hc.get('reason')}")

    logging.info(f"RUN_RESULTS_DIR   = {RUN_RESULTS_DIR}")
    logging.info(f"FINAL_MODEL_DIR   = {FINAL_MODEL_DIR}")
    logging.info(f"BASE_RESULTS_DIR  = {BASE_RESULTS_DIR}")
    logging.info(f"DATA_PATH         = {DATA_PATH}")

    min_rows = WINDOW_SIZE + 50
    counts = df["Symbol"].value_counts()
    candidate_symbols = [sym for sym, n in counts.items() if int(n) >= int(min_rows)]

    for sym, n in counts.items():
        if int(n) < int(min_rows):
            logging.warning(f"Skipping {sym}: only {n} rows (< {min_rows}).")

    if not candidate_symbols:
        logging.error("No symbols have enough rows for WINDOW_SIZE. Nothing to train.")
    else:
        logging.info(f"Candidate symbols: {candidate_symbols}")

    needed_cols = ["Close", "Datetime"]
    if ENABLE_WAVELET:
        needed_cols.append("Denoised_Close")
    if ENABLE_SENTIMENT:
        needed_cols.append("SentimentScore")

    valid_symbols = []
    for sym in candidate_symbols:
        cols = set(df.loc[df["Symbol"] == sym].columns)
        missing = [c for c in needed_cols if c not in cols]
        if missing:
            logging.warning(f"Skipping {sym}: missing required cols {missing}")
        else:
            valid_symbols.append(sym)

    if not valid_symbols:
        logging.error("No symbols passed column checks. Nothing to train.")
    else:
        logging.info(f"Final training universe: {valid_symbols}")

    if FORCE_RETRAIN and valid_symbols:
        cleanup_for_force_retrain(
            run_results_dir=RUN_RESULTS_DIR,
            final_model_dir=FINAL_MODEL_DIR,
            qc_top_dir=QC_TOP_DIR,
            tickers=valid_symbols,
            max_windows=50,
        )

        _safe_json_dump(MANIFEST, ARTIFACT_MANIFEST_PATH)
        _safe_json_dump(RUN_CONFIG, RUN_CONFIG_PATH)
        _safe_json_dump({"features_ordered": ORDERED_FEATURES}, FEATURE_MANIFEST_PATH)
        _safe_json_dump(PROVENANCE, DATA_PROVENANCE_PATH)

    all_results = []

    if TEST_MODE:
        test_stocks = ["AAPL", "NVDA", "MSFT"]
        present = [s for s in test_stocks if s in valid_symbols]
        if not present:
            logging.warning("TEST_MODE: none of ['AAPL','NVDA','MSFT'] present after filters.")
        else:
            logging.info(f"TEST_MODE running: {present}")

        for sym in present:
            logging.info(f">>> [TEST_MODE] Processing {sym}")
            res = process_ticker(sym)
            logging.info(f"{sym}: produced {len(res)} window summaries")
            if res:
                all_results.extend(res)

        summary_path = os.path.join(RUN_RESULTS_DIR, "summary_test_mode.csv")
        if all_results:
            pd.DataFrame(all_results).to_csv(summary_path, index=False)
            logging.info(f"Test-mode summary saved -> {summary_path}")
        else:
            logging.warning("Test mode finished but no results were generated.")
    else:
        summary_path = os.path.join(RUN_RESULTS_DIR, "summary.csv")
        all_results = run_parallel_tickers(valid_symbols, out_path=summary_path, max_workers=MAX_WORKERS)

    try:
        build_phase4_reports()
    except Exception as e:
        logging.error(f"build_phase4_reports failed: {e}")

    try:
        build_ppo_selector()
    except Exception as e:
        logging.error(f"build_ppo_selector failed: {e}")
```

```python
import os, glob, re, json
import pandas as pd
from datetime import datetime

BASE = "/content/drive/MyDrive/Results_May_2025"
MASTER_DIR = os.path.join(BASE, "ppo_models_master")
QC_DIR     = os.path.join(BASE, "ppo_models_QC_TOP")

# -------------------------
# Helpers
# -------------------------
def _count(d):
    return {
        "zip": len(glob.glob(os.path.join(d, "*.zip"))),
        "pkl": len(glob.glob(os.path.join(d, "*.pkl"))),
        "features": len(glob.glob(os.path.join(d, "*_features.json"))),
        "prob_cfg": len(glob.glob(os.path.join(d, "*_probability_config.json"))),
        "info": len(glob.glob(os.path.join(d, "*_model_info.json"))),
        "pred": len(glob.glob(os.path.join(d, "*_predictions.csv"))),
        "pred_compat": len(glob.glob(os.path.join(d, "*_predictions_compat.csv"))),
        "trades": len(glob.glob(os.path.join(d, "*_trades.csv"))),
        "summary": len(glob.glob(os.path.join(d, "summary*.csv"))),
        "manifest": int(os.path.exists(os.path.join(d, "artifact_manifest.json"))),
        "config": int(os.path.exists(os.path.join(d, "config.json"))),
        "feature_manifest": int(os.path.exists(os.path.join(d, "feature_manifest.json"))),
        "data_provenance": int(os.path.exists(os.path.join(d, "data_provenance.json"))),
    }

def _prefixes_from_models(folder):
    zips = glob.glob(os.path.join(folder, "ppo_*_model.zip"))
    prefs = sorted(set(re.sub(r"_model\.zip$", "", os.path.basename(p)) for p in zips))
    return prefs

def _latest_run_folder():
    runs = sorted(glob.glob(os.path.join(BASE, "ppo_walkforward_results_*")))
    return runs[-1] if runs else None

def _latest_summary_in_run(run_folder):
    if not run_folder:
        return None
    candidates = sorted(glob.glob(os.path.join(run_folder, "summary*.csv")))
    return candidates[-1] if candidates else None

def _latest_manifest_in_run(run_folder):
    if not run_folder:
        return None
    p = os.path.join(run_folder, "artifact_manifest.json")
    return p if os.path.exists(p) else None

def _check_deploy_bundle(prefix, folder):
    need = ["_model.zip", "_vecnorm.pkl", "_features.json", "_probability_config.json", "_model_info.json"]
    return {s: os.path.exists(os.path.join(folder, prefix + s)) for s in need}

def _check_run_outputs(prefix, run_folder):
    need = ["_predictions.csv", "_predictions_compat.csv", "_trades.csv"]
    return {s: os.path.exists(os.path.join(run_folder, prefix + s)) for s in need}

def _print_bundle_status(label, prefix, status_dict):
    ok = all(status_dict.values())
    print(f"{label} | {prefix} ->", ("OK" if ok else "MISSING"), status_dict)

def _run_state(run_folder):
    if not run_folder or not os.path.exists(run_folder):
        return "missing"
    counts = _count(run_folder)
    if counts["summary"] > 0:
        return "completed_or_partial_outputs_present"
    if counts["manifest"] or counts["config"] or counts["feature_manifest"] or counts["data_provenance"]:
        return "initialized_but_no_summary_yet"
    return "empty_or_uninitialized"

# -------------------------
# Header
# -------------------------
print("\n=== PPO ARTIFACT HEALTH CHECK (Updated) ===")
print("Time:", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
print("BASE:", BASE)
print("MASTER_DIR:", MASTER_DIR, "| exists:", os.path.exists(MASTER_DIR))
print("QC_DIR:", QC_DIR, "| exists:", os.path.exists(QC_DIR))

# -------------------------
# Latest run folder
# -------------------------
latest_run = _latest_run_folder()
latest_summary = _latest_summary_in_run(latest_run)
latest_manifest = _latest_manifest_in_run(latest_run)

print("\n[Latest Run]")
print("Latest run folder:", latest_run)
print("Latest run state:", _run_state(latest_run))
print("Latest summary file:", latest_summary)
print("Latest manifest file:", latest_manifest)

if latest_manifest:
    try:
        with open(latest_manifest, "r") as f:
            man = json.load(f)
        art = man.get("artifacts", {})
        print("Manifest artifact prefixes:", len(art))
        print("Manifest sample prefixes:", list(art.keys())[:10])
    except Exception as e:
        print("Could not read manifest:", e)

if latest_summary and os.path.exists(latest_summary):
    try:
        s = pd.read_csv(latest_summary)
        print("Summary rows:", len(s))
        if "Ticker" in s.columns:
            print("Tickers in summary:", sorted(s["Ticker"].dropna().astype(str).unique().tolist())[:50])
    except Exception as e:
        print("Could not read summary:", e)
else:
    print("No summary file in latest run yet. This can be normal right after FORCE_RETRAIN cleanup or before training finishes.")

# -------------------------
# Folder counts
# -------------------------
print("\n[Counts]")
print("MASTER counts:", _count(MASTER_DIR) if os.path.exists(MASTER_DIR) else "missing folder")
print("QC_TOP counts:", _count(QC_DIR) if os.path.exists(QC_DIR) else "missing folder")
print("RUN counts:", _count(latest_run) if (latest_run and os.path.exists(latest_run)) else "missing run folder")

# -------------------------
# Prefix lists
# -------------------------
master_prefixes = _prefixes_from_models(MASTER_DIR) if os.path.exists(MASTER_DIR) else []
qc_prefixes     = _prefixes_from_models(QC_DIR) if os.path.exists(QC_DIR) else []

print("\n[Prefixes]")
print("MASTER prefixes:", master_prefixes[:15], ("... total " + str(len(master_prefixes)) if len(master_prefixes) > 15 else ""))
print("QC_TOP prefixes:", qc_prefixes[:15], ("... total " + str(len(qc_prefixes)) if len(qc_prefixes) > 15 else ""))

# -------------------------
# QC_TOP integrity (deploy-only)
# -------------------------
print("\n[QC_TOP Deploy Bundle Integrity — should be all True]")
for pref in qc_prefixes[:10]:
    status = _check_deploy_bundle(pref, QC_DIR)
    _print_bundle_status("QC_TOP", pref, status)

# -------------------------
# MASTER integrity (deploy-only)
# -------------------------
print("\n[MASTER Deploy Bundle Integrity — should be all True]")
for pref in master_prefixes[:10]:
    status = _check_deploy_bundle(pref, MASTER_DIR)
    _print_bundle_status("MASTER", pref, status)

# -------------------------
# Run outputs integrity (run-only)
# -------------------------
if latest_run and os.path.exists(latest_run):
    print("\n[RUN Output Integrity — predictions/trades live here]")
    pred_files = glob.glob(os.path.join(latest_run, "ppo_*_predictions.csv"))
    run_prefs = sorted(set(os.path.basename(p).replace("_predictions.csv", "") for p in pred_files))
    if run_prefs:
        for pref in run_prefs[:10]:
            status = _check_run_outputs(pref, latest_run)
            _print_bundle_status("RUN", pref, status)
    else:
        print("No run-level prediction files found in latest run folder.")

# -------------------------
# One-off check function
# -------------------------
def check_prefix(prefix):
    print("\n[Check prefix]", prefix)
    if os.path.exists(os.path.join(QC_DIR, prefix + "_model.zip")):

        where = "QC_TOP"
        folder = QC_DIR
        status = _check_deploy_bundle(prefix, folder)
        print("Found in:", where, "->", folder)
        print(status)
        return

    if os.path.exists(os.path.join(MASTER_DIR, prefix + "_model.zip")):
        where = "MASTER"
        folder = MASTER_DIR
        status = _check_deploy_bundle(prefix, folder)
        print("Found in:", where, "->", folder)
        print(status)
        return

    if latest_run and os.path.exists(os.path.join(latest_run, prefix + "_predictions.csv")):
        where = "RUN"
        folder = latest_run
        status = _check_run_outputs(prefix, folder)
        print("Found in:", where, "->", folder)
        print(status)
        return

    print("Not found in MASTER, QC_TOP, or latest RUN folder.")
```
